## Binding TELs by State

```python
#Establish working directory
workdir='C:/Users/marvinw/Documents/diss/'

#Read in TEL data
tel_limits=pd.read_csv(workdir+'TEL_lim_state.csv').set_index('YEAR')

#Subset to state IDs and summary TEL vars
tel=tel_limits[['FIPST_N','TYPE1','TYPE2','BOTH']].ix[2013]

#Read in FIPS codes
fips_data=pd.read_csv(workdir+'state_fips.csv')

#Create lower case version
fips_data['st']=fips_data['state'].apply(lambda x: x.lower())

#Create FIPS maps
fips=dict(zip(fips_data['fips'],fips_data['state']))
fips_l=dict(zip(fips_data['fips'],fips_data['st']))

#Map abbreviations in with TEL limit data
tel['fips']=tel['FIPST_N'].map(fips)
tel['fips_l']=tel['FIPST_N'].map(fips_l)

#Capture four TEL situations: no TEL, Type 1, Type 2, Both
mask1=(tel['TYPE1']==0) & (tel['TYPE2']==0)
mask2=(tel['TYPE1']==1) & (tel['TYPE2']==0)
mask3=(tel['TYPE1']==0) & (tel['TYPE2']==1)
mask4=(tel['BOTH']==1)
tel['tel']=NaN
tel['tel'].ix[mask1]=1
tel['tel'].ix[mask2]=2
tel['tel'].ix[mask3]=3
tel['tel'].ix[mask4]=4

#Define color map
col=['#c6dbef','#67000d','#cb181d','#fb6a4a']
col_map={1:col[0],
         2:col[1],
         3:col[2],
         4:col[3]}

#Assign colors
tel['col']=tel['tel'].map(col_map)

tel.head()
```
