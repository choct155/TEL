# Local Government Finances from [Censtats](http://censtats.census.gov/usa/usa.shtml)

I need to roll some sort of measure of the relevance of county government finances into the analysis.  The Censtats database hold information on aggregated local finances by county.  The coverage is spotty, it comes in economic census years (1972-2002).  This is less than ideal, but it's an easier fix than determining the proportional allocations of constituent school districts and municipalities within each county (they do not next nicely).  This notebook just cleans up the data from the Censtat Excel sheets and creates a Colorado subset.

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
#Establish working directory
# workdir='C:/Users/marvinw/Documents/diss/CenStat/'

#Read in data
# log_tmp=pd.read_csv(workdir+'CENSTAT_LOG.csv',dtype={'STCOU':'str'})
log_tmp=pd.read_csv('CENSTAT_LOG.csv',dtype={'STCOU':'str'})

#Subset to Colorado counties
log=log_tmp[(log_tmp['STCOU'].str.startswith('08')) & (log_tmp['STCOU'] != '08000')].set_index(['Areaname','STCOU'])

#Construct list of tuples with value concept in first position and year in second
col_tups=[]
##For each value concept...
for vc in ['TOT_REV','TOT_EXP']:
    ##...and for each year...
    for yr in range(1972,2003,5):
        ##...capture the tuple pairing
        col_tups.append((vc,yr))
        
#Convert tuple list into a column MultiIndex         
col_idx=pd.MultiIndex.from_tuples(col_tups)

#Set new column MultiIndex to columns names in data
log.columns=col_idx

#Name column indices
log.columns.names=['VAL','YEAR']

#Stack value concept
log2=log.stack('VAL').reset_index()

#Isolate county name
log2['CTY']=log2['Areaname'].apply(lambda x: x.split(',')[0])

#Drop Areaname
log2.pop('Areaname')

#Set index
log2.set_index(['STCOU','CTY','VAL'],inplace=True)

print log2.to_string()
```

Now we have to do something about the periodicity.  The years in which the data and the study period (1993-2009) approximately intersect are 1992, 1997, and 2002.  That's not great by any measure.  Unfortunately, this means we will have to interpolate between observed points and extrapolate out from 2002.  Clearly a hack, but the alternative is to use Colorado data for each year, and proportionally allocate expenditures and revenues from municipalities and school districts based upon spatial overlap.  Even if I had the school district data, which I currently do not, that's a big lift since this issue has come up only three weeks before the defense.  My advisor has signed off on using this less than ideal proxy.

So, on with the hackery...  It may be easier to perform the interpolation on a transposed set so we can get it going down vectors.

```python
#Set index
logT=log2.T

#Subset to years after 1991 and reindex to last observation
logT92=logT[logT.index>1991].reindex(range(1992,2003))

#Interpolate between known values
for col in logT92.columns:
    logT92[col]=logT92[col].interpolate(method='linear')

#Name index
logT92.index.names=['YEAR']
    
print logT92.diff()

logT92
```

As we can see in the differences DF, the two interpolation periods display quite different properties, and we don't have a good reason to choose one period over the other to inform extrapolation.  We will thus use the "long term" growth profile, that is, the growth between 1992 and 2002 to define the growth for the last 7 years.

```python
#Determine the average growth over the 1992-2002 period
log_grow=DataFrame(logT92.diff().mean()).T

#Create container for year-specific DFs
extrap_dfs=[]

#For each year...
for i,yr in enumerate(range(2003,2010)):
    new_row=DataFrame(logT92.ix[2002]+(i+1)*log_grow.ix[0]).T
    extrap_dfs.append(new_row)
    
#Concatenate new rows
new_rows=pd.concat(extrap_dfs)

#Create new index
new_rows['YEAR']=range(2003,2010)
new_rows.set_index('YEAR',inplace=True)

#Concatenate new rows onto old DF
clog=pd.concat([logT92,new_rows])

#Generate plot figure
fig,ax=plt.subplots(figsize=(15,5))

#Plot expenditures and revenues for Adams
clog[clog.columns[:2]].plot(kind='line',ax=ax,title='Test Plot - Should see average extrap...')

clog
```

Ok, brutish yet effective enough.  Let's get the data in shape for a join with the data contained in `CO_CTY_DATA.ipynb`.  Those data do not carry `STCOU` in the index, years are stacked, and the index names are lower case (`['year','cty']`).
