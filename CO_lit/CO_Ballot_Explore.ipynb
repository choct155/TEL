## County Ballot Data

The ballot data describe votes on spending limit waivers and revenue increases (of the sales, property, and debt variety).  Each vote is carries an identifying county and year, information about the scope of the waivers/changes, and the results of said vote (both as a percentage of yes votes and as a binary pass/fail).  

In the previous Notebook, we looked at the distribution of votes, spending sources, spending/revenue targets, and the magnitude of revenue changes.  Let us take a quick look at these data again since we already have the code.

```python
%%R -w 700 -h 300

print(ggplot(vts,aes(x=results)) + 
      geom_histogram(binwidth=5,colour='white',fill='#B50724') + 
      labs(title='Distribution of County Ballot Initiative Results') +
      xlab('Yes Vote %') +
      ylab('Frequency') +
      theme_bw())

print(ggplot(vts,aes(x=s_source)) + 
      geom_histogram(colour='white',fill='#B50724') + 
      labs(title='Distribution of Explicit Revenue Sources for Spending Waivers') +
      xlab('Source') +
      ylab('Frequency') +
      theme_bw() +
      theme(axis.text.x=element_text(angle=20)))

print(ggplot(vts,aes(x=s_target)) + 
      geom_histogram(colour='white',fill='#B50724') + 
      labs(title='Distribution of Explicit Spending Targets for Spending Waivers') +
      xlab('Target') +
      ylab('Frequency') +
      theme_bw() +
      theme(axis.text.x=element_text(angle=20)))

print(ggplot(vts,aes(x=target)) + 
      geom_histogram(colour='white',fill='#B50724') + 
      labs(title='Distribution of Explicit Revenue Targets for Revenue Changes') +
      xlab('Target') +
      ylab('Frequency') +
      theme_bw() +
      theme(axis.text.x=element_text(angle=20)))

print(ggplot(vts,aes(x=sales_change)) +
      geom_density(fill='#B50724') +
      xlim(0,10) +
      labs(title='Distribution of Sales Rate Changes (%)') +
      xlab('Rate') +
      ylab('Frequency') +
      theme_bw())

print(ggplot(vts,aes(x=prop_change)) +
      geom_density(fill='#B50724') +
      xlim(0,15) +
      labs(title='Distribution of Property Rate Changes (Mills)') +
      xlab('Rate') +
      ylab('Frequency') +
      theme_bw())

print(ggplot(vts,aes(x=debt_change)) +
      geom_density(fill='#B50724') +
      #xlim(0,10) +
      labs(title='Distribution of Debt Changes ($)') +
      xlab('Rate') +
      ylab('Frequency') +
      theme_bw())
```

As can be seen, we have great normal variation in the `results` variable, but substantial bunching elsewhere.

+ *REVENUE SOURCES FOR SPENDING WAIVERS* are dominated by 'open' and 'no waiver', which limits the ability of specific sources being a solid predictive factor for vote outcomes;
+ *SPENDING TARGETS FOR SPENDING WAIVERS* are similarly dominated by 'open' and 'no waiver', which stands to reason given the revenue sources and implies a similar deficiency;
+ *REVENUE TARGETS FOR REVENUE CHANGES* are substantially more varied and may provide a solid predictor, particularly since the distribution above includes combinations of atomistic groupings.  (For example, 'hss' and 'infrastructure' on the same ballot are bundled as 'hss/infrastructure', and this does not add to their individual tallies in the distribution above.)  In practice, the categories are all also split out as dummies, and we can look at this distribution in more detail later;
+ *ALL REVENUE CHANGES* (`sales_change, prop_change, debt_change`) have strong clustering around zero because non-action on a given ballot issue is coded as zero.
+ *SALES CHANGES* show more volatility in the proposed rates than other revenue changes, but they pretty much iron out above approximately 4%;
+ *PROPERTY CHANGES* decline in a far smoother fashion than sales & use, evening out from a frequency standpoint at around 7 mills;
+ *DEBT PROPOSALS* rarely appear to exceed about $2 M.

It would be interesting to see the differences in proposals that passed versus those that did not.  Let's first look at broad action categories.  Essentially, do some types of proposals (`spend_waiver, sales, property, debt`) win more often than others?

```python
#Find sums of each action type within pass/fail subsets
action_cts=DataFrame({'Pass':vts[vts['pass']==1][['spend_waiver','sales','property','debt']].sum(),
                      'Fail':vts[vts['pass']==0][['spend_waiver','sales','property','debt']].sum()})

#Create plot object
fig,ax=plt.subplots(figsize=(15,5))

#Plot comparison
action_cts.plot(kind='barh',title='Comparison of Successful Votes',colormap='RdGy',ax=ax,alpha=.7)

#Set plot parameters
ax.patch.set_facecolor('white')
```

Spending waivers are far more likely to pass, but this may not be all that surprising since they accompany revenue changes in most cases, and often are voted on as standalones.  The near parity in property and sales is interesting, but perhaps not as interesting as the general successful tendency of debt increases compared to the average failures of property and sales.

Do the votes that pass and those that fail look different in terms of the magnitude of the change?

```python
#Generate subsets by pass/fail for scope variables
scope_dist=pd.concat([vts[vts['pass']==1][['pass','sales_change','prop_change','debt_change']],
                      vts[vts['pass']==0][['pass','sales_change','prop_change','debt_change']]]).fillna(0)
#Create plot object
fig,ax=plt.subplots(3,figsize=(15,15))

#Generate violin plots
sb.violinplot(scope_dist['sales_change'],scope_dist['pass'],ax=ax[0])
sb.violinplot(scope_dist['prop_change'],scope_dist['pass'],ax=ax[1])
sb.violinplot(scope_dist['debt_change'],scope_dist['pass'],ax=ax[2])

#Set plot background
ax[0].patch.set_facecolor('white')
ax[1].patch.set_facecolor('white')
ax[2].patch.set_facecolor('white')

#Set titles
ax[0].set_title('Distribution of `sales_change` in Pass/Fail Subsets')     
ax[1].set_title('Distribution of `prop_change` in Pass/Fail Subsets')     
ax[2].set_title('Distribution of `debt_change` in Pass/Fail Subsets')     
```

Well this is interesting.  There is a great deal of variation in the size of the change among the pass subsets of all three types of revenue changes.  The median is a little difficult to gather here dues to the amount of variation.  How about a boxplot view?

```python
#Create plot object
fig,ax=plt.subplots(3,figsize=(15,15))

#Generate violin plots
sb.boxplot(scope_dist['sales_change'],scope_dist['pass'],ax=ax[0])
sb.boxplot(scope_dist['prop_change'],scope_dist['pass'],ax=ax[1])
sb.boxplot(scope_dist['debt_change'],scope_dist['pass'],ax=ax[2])

#Set plot background
ax[0].patch.set_facecolor('white')
ax[1].patch.set_facecolor('white')
ax[2].patch.set_facecolor('white')

#Set titles
ax[0].set_title('Distribution of `sales_change` in Pass/Fail Subsets')     
ax[1].set_title('Distribution of `prop_change` in Pass/Fail Subsets')     
ax[2].set_title('Distribution of `debt_change` in Pass/Fail Subsets')  
```

From this view, it looks like there may be a slight difference in the typical rate differential between passed and failed votes.  It is hard to argue, however, that it is material.  Let's take a quick look at scatters between percentage of yes votes and revenue changes.

```python
#Generate plot object
fig,ax=plt.subplots(1,3,figsize=(18,5))

#Plot revenue changes vs results
ax[0].scatter(vts['results'],vts['sales_change'],s=100,c='#B50724',alpha=.5)
ax[1].scatter(vts['results'],vts['prop_change'],s=100,c='#B50724',alpha=.5)
ax[2].scatter(vts['results'],vts['debt_change'],s=100,c='#B50724',alpha=.5)

#Set background
ax[0].patch.set_facecolor('white')
ax[1].patch.set_facecolor('white')
ax[2].patch.set_facecolor('white')

#Set vertical lines
ax[0].vlines(x=50,ymin=-1,ymax=vts['sales_change'].describe()['max']+1,lw=5,linestyle='--')
ax[1].vlines(x=50,ymin=-1,ymax=vts['prop_change'].describe()['max']+1,lw=5,linestyle='--')
ax[2].vlines(x=50,ymin=-1,ymax=vts['debt_change'].describe()['max']+1,lw=5,linestyle='--')

#Set labels
plt.suptitle('Results vs Revenue Change',fontsize=20)
plt.subplots_adjust(top=.85)
ax[0].set_title('Sales (%)')
ax[1].set_title('Property (Mills)')
ax[2].set_title('Debt ($)');
```

## Distribution of Pass/Fail in the Context of TEL Variables

This whole analysis is attempting to capture the impacts of our TEL variables (`intensity_stock, prop_ratio`).  Let's see where passes and fails sit in the TEL plane.

```python
#Generate subsets
tel_pass=vts[vts['pass']==1][['intensity_stock','prop_ratio']]
tel_fail=vts[vts['pass']==0][['intensity_stock','prop_ratio']]

#Generate plot object
fig,ax=plt.subplots(figsize=(10,10))

#Plot pass/fail on the TEL plane
ax.scatter(tel_pass['intensity_stock'],tel_pass['prop_ratio'],c='b',s=100,alpha=.4)
ax.scatter(tel_fail['intensity_stock'],tel_fail['prop_ratio'],c='r',s=100,alpha=.4)

#Set background
ax.patch.set_facecolor('white')

#Set labels
ax.set_title('Pass/Fail on the TEL Plane',fontsize=20)
ax.set_xlabel('TEL Intensity')
ax.set_ylabel('Gallagher Ratio');
```
