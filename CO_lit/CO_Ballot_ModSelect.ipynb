## Best Subset Regression

Best Subset is a very computationally intensive technique that checks all available subset of regressors to identify the best model.  We actually have a decent number of variables to choose from.  Let's see if this is feasible.  

```python
%%R 

#Generate subset list of variables
var_sub<-names(vts_cc)[!names(vts_cc) %in% c('cty','results','pass','s_target','s_source','target')]

#Generate formula for the models - results dependent
r_formula<-as.formula(paste('results~',paste(var_sub,collapse='+')))

#Fit models
vts_bs<-regsubsets(r_formula,data=vts_cc)
```

Well isn't that interesting.  R is straight up telling us this is a bad idea.  I figured we'd end up with the regularization approach anyway.

## Regularization

We will use the penalization approach with two different penalties:  ridge and lasso.

### Ridge Regression

Unfortunately, the **`glmnet`** package does not use the formula language, so we need to create the `X` matrix and `y` vector used in the model.  We can still leverage a the `model.matrix()` method.  Note that we must mean center the data for regularization techniques.

```python
%%R
#Define mean centering function
center_scale<-function(x){
    scale(x,scale=FALSE)
    }

#Identify numeric columns
num_cols<-sapply(vts_cc,is.numeric)

#Standardize data
vts_cc_mc<-vts_cc
vts_cc_mc[,num_cols]<-scale(vts_cc_mc[,num_cols])

#Drop columns filled with missing values (our complete cases subset had all zeroes)
null_cols<-names(vts_cc_mc)[sapply(vts_cc_mc,function(x)all(is.na(x)))]
vts_reg<-vts_cc_mc[,!(names(vts_cc_mc) %in% null_cols)]

#Generate new subset list of variables
var_sub2<-names(vts_reg)[!names(vts_reg) %in% c('cty','results','pass','s_target','s_source','target')]

#Generate formula for the models - results dependent
r_formula2<-as.formula(paste('results~',paste(var_sub2,collapse='+')))
    
#Generate X matrix
x<-model.matrix(r_formula2,data=vts_reg)

#Generate y vector
#y<-vts_cc_mc$results
y<-vts_reg$results
```

Now we can fit the ridge model.  Note that `alpha=0` corresponds to the ridge penalty, while `alpha=1` corresponds to lasso.

```python
%%R

#Fit ridge model
vts_reg.ridge<-glmnet(x,y,alpha=0)

#Observe coefficients v lambda value
print(plot(vts_reg.ridge,xvar='lambda',label=TRUE))

#Fit train CV
cv.ridge<-cv.glmnet(x,y,alpha=0)

#Observe MSE as function of lambda
print(plot(cv.ridge))

#Find lambda corresponding to min MSE
r_lambda<-cv.ridge$lambda.min

print(paste('LAMBDA MINIMIZES MSE AT:',r_lambda))

print(predict(vts_reg.ridge,type='coefficients',s=r_lambda))
```

Let's evaluate the lasso model.

```python
%%R

#Fit lasso model
vts_reg.lasso<-glmnet(x,y,alpha=1)

#Observe coefficients v lambda value
print(plot(vts_reg.lasso,xvar='lambda',label=TRUE))

#Fit train CV
cv.lasso<-cv.glmnet(x,y,alpha=1)

#Observe MSE as function of lambda
print(plot(cv.lasso))

#Find lambda corresponding to min MSE
l_lambda<-cv.lasso$lambda.min
print(paste('LAMBDA MINIMIZES MSE AT:',l_lambda))

print(predict(vts_reg.lasso,type='coefficients',s=l_lambda))
```

Let's look at the variance in non-zero coefficients as a function of lambda.

```python
%%R -w 700 -h 500
#Generate DF
l_nonzero<-data.frame(lambda=vts_reg.lasso$lambda,nonzero=vts_reg.lasso$df)

#Plot non-zero coefficients vs lambda
print(ggplot(l_nonzero,aes(x=lambda,y=nonzero)) +
      geom_point(alpha=.4,colour='#B50724',size=6) +
      geom_vline(xintercept=l_lambda, linetype='longdash',size=2, colour='grey50') +
      annotate('text',x=l_lambda+.015,y=100, label='Optimal\nLambda') +
      theme_bw() +
      labs(title='Non-Zero Coefficients as f(Lambda)'))
```

Let's check out what lasso chose not to toss out.

```python
%%R

print(str(predict(vts_reg.lasso,type='coefficients',s=l_lambda)))

#Identify position of non-zero coefficients (converted from 0- to 1-index)
lasso_coef_pos<-unlist(predict(vts_reg.lasso,type='coefficients',s=l_lambda)@i)+1

#Identify full list of variables, and the subset of nonzeros
lasso_coefs<-predict(vts_reg.lasso,type='coefficients',s=l_lambda)@Dimnames[[1]]
lasso_nonzero_coefs<-lasso_coefs[lasso_coef_pos]

#Generate DF of nonzero coefficients and values
lasso_coef<-data.frame(var=lasso_nonzero_coefs,coef=unlist(predict(vts_reg.lasso,type='coefficients',s=l_lambda)@x))

#Add variable categories
lasso_coef$cat<-NA
lasso_coef$cat[which(lasso_coef$var %in% id)]<-'id'
lasso_coef$cat[which(lasso_coef$var %in% ballot)]<-'ballot'
lasso_coef$cat[which(lasso_coef$var %in% dem)]<-'dem'
lasso_coef$cat[which(lasso_coef$var %in% tel)]<-'tel'
lasso_coef$cat[which(lasso_coef$var %in% asmt)]<-'asmt'
lasso_coef$cat[which(lasso_coef$var %in% fisc)]<-'fisc'
lasso_coef$cat[which(lasso_coef$var %in% ind)]<-'ind'
lasso_coef$cat[which(lasso_coef$var %in% cycle)]<-'cycle'
lasso_coef$cat[which(lasso_coef$var %in% wts)]<-'wts'

print(lasso_coef)
```

Note that we have both ballot variables and socioeconomic/fiscal variables.  This is encouraging.

```python
%%R -w 800 -h 800


print(ggplot(lasso_coef,aes(x=reorder(var,coef),y=coef,fill=cat)) + 
      geom_bar(stat='identity') +
      coord_flip() +
      theme_bw() +
      labs(title='Determinants of Ballot Initiative Results - Lasso') +
      ylab(paste('Lasso Coefficient at',expression(Lambda),'=',l_lambda)) +
      xlab('Variable'))
```
