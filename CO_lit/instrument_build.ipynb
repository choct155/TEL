## Imputation

Let's extract socioeconomic information.  By this we mean...

1. Population Dynamics
2. Business Cycle Information
3. Housing Information
4. Industrial composition

We will also need to capture identifiers and passage info.  To run the required models requires complete cases.  Let's evaluate the difference in row count if we drop complete cases.

```python
%%R

#Subset to desired variables
vts_sub<-vts[,c(id,c("results"),dem,ind,cycle)]

#Count rows in full set
print(nrow(vts_sub))

#Subset to complete cases
vts_cc<-vts_sub[complete.cases(vts_sub),]

#Count complete cases
print(nrow(vts_cc))

print(names(vts_sub))
```

We actually lose almost 25% of our records, which is clearly non-trivial.  The tempting approach is to drop incomplete cases, find which variables matter, and hope to be able to jettison the biggest missing value offenders.  However, our good friends [Trevor Hastie and Rob Tibshirani](https://class.stanford.edu/courses/HumanitiesScience/StatLearning/Winter2014/about) remind us that we would be partitioning the learning space into two unequal components (thereby introducing bias).  The correct approach is to drop the most invariant variables *before any learning actually occurs*, and again hope that we can jettison the biggest offenders.  To some extent, this is not entirely a bad deal insofar as missing values categorically reduce the capacity for variation in a given variable.

Let us calculate the coefficient of variation for each variable, and the number of missing values.  The graph below shows the coefficient of variation for each variable (length of the bar) and the number of missing values (color of the bar).

```python
%%R
#Calculate coefficient of variation by variable
vts_cov<-sapply(vts_sub,function(x) sd(x,na.rm=TRUE))/
          sapply(vts_sub,function(x) mean(x,na.rm=TRUE))

#Calculate number of missing values by variable              
vts_miss<-sapply(vts_sub,function(x) sum(is.na(x)))

#Combine results in a DF
vts_cm<-data.frame(cbind(vts_cov,vts_miss))
    
#Pull out row names for use in graphing
vts_cm$var<-rownames(vts_cm)
    
#Calculate proportion of all variation
vts_cm$var_prop<-vts_cm$vts_cov/sum(vts_cm$vts_cov,na.rm=TRUE)
print(vts_cm)
    
#Plot variation and missing values    
print(ggplot(vts_cm,aes(x=reorder(var,var_prop),y=var_prop,fill=vts_miss)) + 
  geom_bar(stat='identity') +
  coord_flip() +
  theme_bw() +
  labs(title='Variation and Missing Values') +
  ylab('Coefficient of Variation') +
  xlab('Variable'))
```

Taking out unemployment rate is a no brainer, but some of the other variables that have high amounts of missing values actually vary quite a bit.  Let's take a look at where these missing cases are.  To do this, we can use [Amelia](http://gking.harvard.edu/amelia/) to perform multiple imputation first.  To figure out if `unemp_rate` or others must be removed, we can examine the Amelia output to compare the distributions of the imputed and observed values, as well as observe the pattern of missingness.  

First, [King et al](http://www.jstatsoft.org/v45/i07/paper) warn that if high correlations exist among the variables, it is helpful to use a [ridge](http://tamino.wordpress.com/2011/02/12/ridge-regression/) prior to push covariances towards zero.  As with any regularization technique, this involves the trading of an increase in bias for decrease in variance.  The question is whether or not that bias increase is tolerable, and the comparison plots will help us make the determination.  To this end, let's see how correlated our variables are at the moment.

```python
%%R -w 1000 -h 800

print(corrgram(vts_sub,upper.panel=panel.pie))
```

As one would expect, we are seeing high correlations among demographic and industrial output variables (they are all levels, and will thus scale together).  Remember that we want to retain as many variables as possible and allow our lasso treatment to choose the correct set.  Consequently, employing the ridge prior will help us do that.  

Let's go ahead and generate the imputations.

```python
%%R
#print(?amelia)

vts_mi<-amelia(vts_sub,m=5,ts='year',cs='cty',polytime=2,intercs=TRUE,p2s=2,empri=0.1*nrow(vts_sub))
```

Ok, now we want to see how the distributions (imputed vs. observed) compare for the variables of concern:

+ `r_wage_fin`
+ `r_wage_prof`
+ `r_wage_ed_hlth`
+ `r_wage_cnstr`
+ `unemp_rate`

```python
%%R

print(plot(vts_mi,which.var=c('r_wage_fin','r_wage_prof','r_wage_ed_hlth','r_wage_cnstr','unemp_rate')))
```

What about the pattern of missingness?

```python
%%R -w 1000 -h 1100

print(missmap(vts_mi))
```

I think we have enough evidence to jettison `unemp_rate` altogether.  It doesn't vary much, there are many missing instances, the imputation model has a difficult job even matching central tendency, and it's missing for a number of entire counties in blocks.  It should be noted that the interactions used in the imputation model really stretched the capability of the available degrees of freedom, which hinders effective imputation.  Despite this, the other variables of concern don't look too bad, so we can keep those imputed values.

```python
%%R

#Subset
vts_sub2<-subset(vts_sub,select=-unemp_rate)

#Rerun the imputation
vts_mi2<-amelia(vts_sub2,m=5,ts='year',cs='cty',polytime=2,intercs=TRUE,empri=0.1*nrow(vts_sub2))
```

## Model Selection

We will use lasso regression to identify the most important predictors of passage.  Before we can get there, however, we need to put the data in a form conducive to the model.  It must be remembered, however, that we have performed multiple imputation, and thus need to run the routine for each of the `m=5` datasets.  This calls for a function...

```python
%%R

print(names(vts_sub2))

#Extract id variables and the dependent
lasso_vars<-names(vts_sub2)[!names(vts_sub2) %in% c('year','cty','stcou','results')]

#Generate formula for the model
r_formula<-as.formula(paste('results~',paste(lasso_vars,collapse='+')))

#Define a function to run lasso on each imputed set
lasso_func<-function(df,vars,fmla){
    #Generate X matrix
    x<-model.matrix(fmla,data=df)
    
    #Generate y vector
    y<-df$results
    
    #Fit lasso model
    vts_reg.lasso<-glmnet(x,y,alpha=1)
    
    #Observe coefficients v lambda value
    print(plot(vts_reg.lasso,xvar='lambda',label=TRUE))
    
    #Fit train CV
    cv.lasso<-cv.glmnet(x,y,alpha=1)
    
    #Observe MSE as function of lambda
    print(plot(cv.lasso))
    
    #Find lambda corresponding to min MSE
    l_lambda<-cv.lasso$lambda.min
    print(paste('LAMBDA MINIMIZES MSE AT:',l_lambda))
    
    #Capture results matrix
    res_mat<-predict(vts_reg.lasso,type='coefficients',s=l_lambda)
    
    #Return df version
    return(data.frame(coef=summary(res_mat)$x,
                      row.names=rownames(res_mat)[summary(res_mat)$i]))
    }
```

Now that we have our function and it appears to be operating well, we can roll through the imputed sets, performing lasso regression to find the most significant determinants of increases in the percentage of votes supporting an exemption.

```python
%%R

#Run models
print('***IMPUTATION 1***')
imp1<-lasso_func(vts_mi2$imputations$imp1,lasso_vars,r_formula)
print('***IMPUTATION 2***')
imp2<-lasso_func(vts_mi2$imputations$imp2,lasso_vars,r_formula)
print('***IMPUTATION 3***')
imp3<-lasso_func(vts_mi2$imputations$imp3,lasso_vars,r_formula)
print('***IMPUTATION 4***')
imp4<-lasso_func(vts_mi2$imputations$imp4,lasso_vars,r_formula)
print('***IMPUTATION 5***')
imp5<-lasso_func(vts_mi2$imputations$imp5,lasso_vars,r_formula)
```

Now we can join the imputed results together to get a sense of the most consistent indicators of vote passage.

```python
%%R -w 800
#Adjust column names for each imputed df
names(imp1)<-'imp1'
names(imp2)<-'imp2'
names(imp3)<-'imp3'
names(imp4)<-'imp4'
names(imp5)<-'imp5'

#Define a function to merge all of the DFs together by variable name
mi_lasso<-Reduce(function(a,b){
            temp<-merge(a,b,by='row.names',all=T)
            row.names(temp)<-temp[,'Row.names']
            temp[,!names(temp) %in% 'Row.names']
            },list(imp1,imp2,imp3,imp4,imp5))

#Generate column wise means
mi_lasso$avg<-rowMeans(mi_lasso,na.rm=TRUE)
mi_lasso$avg_na<-rowMeans(mi_lasso)

#Create variable column
mi_lasso$var<-row.names(mi_lasso)

#Remove intercept
mi_lasso2<-mi_lasso[!rownames(mi_lasso) %in% '(Intercept)',]

print(mi_lasso2[with(mi_lasso2,order(-abs(mi_lasso2$avg))),])

#Plot variation and missing values    
```

Interesting.  The biggest factors driving results seem to be government activity and demographic, with basically no influence from the industrial mix.  This is easier to see graphically.

```python
%%R

print(ggplot(mi_lasso2,aes(x=reorder(var,avg),y=avg,fill=avg_na)) + 
  geom_bar(stat='identity') +
  coord_flip() +
  theme_bw() +
  labs(title='Lasso Averages Across Imputation Sets') +
  ylab('Coefficient') +
  xlab('Variable'))
```

Let's see how much variation the top variables explain (as measured by absolute impact).
