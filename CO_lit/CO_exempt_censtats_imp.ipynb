# Predicting Ballot Initiatives with Imputed [CenStats](http://censtats.census.gov/usa/usa.shtml) Data

This Notebook is an extension of CO_exempt_censtats.  The difference is, our CenStats data were imputed with the R library Amelia, instead my own custom interp/extrap procedure.

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
from sklearn import linear_model, datasets
from sklearn.cross_validation import train_test_split
from sklearn import decomposition
import statsmodels.api as sm
```

The goal is to see if a heavy duty impution leaves us with enough variation to differentiate across counties for exemption vote purposes.

## Data Input

```python
#Establish data locations
vt_dir='/home/choct155/dissertation/TEL/CO_lit/'
cs_dir='/home/choct155/dissertation/MiscData/Census/CenStats/'

#Read in vote data
votes_tmp=pd.read_csv(vt_dir+'clean_votes_cov.csv',usecols=['cty','stcou','year','results'])

#Generate a county name map
cty_map=dict(zip(votes_tmp['stcou'],votes_tmp['cty']))

print 'Number of counties:',len(cty_map.keys())

#Drop the county name variable
votes_tmp.pop('cty')

#Convert to float
votes=votes_tmp.astype(float)

#Set index
votes.set_index(['stcou','year'],inplace=True)

#Sort index
votes.sortlevel(0,inplace=True)

votes.head()
```

We also need the imputed sets.

```python
#Read in imputed sets
cs_imp=[]

#For each imputed set...
for i in range(1,6):
    #...capture doctored version in the list
    csi=pd.read_csv(cs_dir+'censtat_imp'+str(i)+'.csv')
    csi2=csi[csi.columns[1:]].set_index(['stcou','year']).sortlevel(0)
    cs_imp.append(csi2)

cs_imp[0].head()
```

Each of these sets must be joined to the results, so that the covariates are sequentially ordered in an appropriate manner.

```python
#Generate container for joined sets
cs_dfs=[]

#For each imputed set....
for df in cs_imp:
    #...join it to the results...
    cs_tmp=votes.join(df,how='left')
    #...add a binary passage variable...
    cs_tmp['pass']=np.where(cs_tmp['results']>50,1,0)
    #...and throw it in the list
    cs_dfs.append(cs_tmp)
    
print (cs_dfs[0].applymap(lambda x: np.where(x<0,1,0)).sum().sum())/float(cs_dfs[0].shape[0]*cs_dfs[0].shape[1])

cs_dfs[0].head()
```

## Principal Components Analysis

Before getting into model selection and prediction, let's set our expectations by viewing the data in space.  With our earlier interp/extrap routine, we captured 96% of the variation in the first two principal components.  What happens in with the Amelia imputations?  First, let us capture our response variables and the design matrix.

```python
#Capture responses (which do not vary across imputed sets)
y_res=cs_dfs[0]['results'].values
y_pass=cs_dfs[0]['pass'].values

#Capture design variable names
X_vars=[var for var in cs_dfs[0].columns if var not in ['results','pass']]

#Create container for design matrices
dmat_list=[]

#For each imputed set...
for df in cs_dfs:
    #...capture the design matrix in the list
    dmat_list.append(df[X_vars].values)
    
dmat_list[0]
```

Now let us design a function that permits us to conduct PCA and view the dynamic for each imputed set.

```python
def imp_pca(dmat,ax,i):
    #Instantiate PCA object
    pca=decomposition.PCA()
    
    #Fit data to PCA object
    pca.fit(dmat)
    
    #Plot explained variance
    ax.plot(pca.explained_variance_ratio_)
    
    #Set range
    ax.set_xlim([0,20])
    ax.set_xticks(range(21))
    
    #Fix titles
    ax.set_title('Explained Variance Ratio vs. Number of Dimensions - '+str(i))
    
    print 'Proportion of variation explained by first two components:',pca.explained_variance_ratio_[:2].sum()
    
    #Generate rotated design matrix
    return np.dot(dmat,pca.components_.T)

#Generate plot object
fig,ax=plt.subplots(3,2,figsize=(16,15))

#Generate container for component output
comp_list=[]

#For each design matrix...
for i,dmat in enumerate(dmat_list):
    comp_list.append(imp_pca(dmat,ax[i//2,i%2],i))
```

Just as before, two components dominate the explained variation for all sets: ~96%.

We are now in a position to search for patterns in the rotated basis.  We will use color to reflect the response dimension.  Further, we will evaluate the continuous `results` data and the binary `pass` data.  Let's see what we have.
