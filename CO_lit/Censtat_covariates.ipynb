I am not clear on why this made sense at the time, but dealing with it is the first purpose of this Notebook.

The Censtats data are observed sparingly.  To preserve some "smoothness" in the annual mapping of this data, the second purpose of this Notebook is to interpolate annual values between observations.

## Converting Data to Long Format

The following years are represented in the data set (in a `'code':year` mapping), though not all variables are observed in each year:

    {'189':1989,
     '190':1990,
     '192':1992,
     '196':1996,
     '199':1999,
     '200':2000,
     '204':2004,
     '209':2007,
     '208':2008,
     '210':2010}

Note that 2007 is actually the mid-year for the data taken from the 2005-2009 [ACS](https://www.census.gov/acs/www/).  The codes are taken from the variable names, which have the following format:

+ Characters 0-2 = High Level Concept (e.g. Age)
+ Characters 3-5 = Low Level Concept (e.g. Median Age)
+ Characters 6-8 = Year Code
+ Character 9 = Data Identifier

I think my approach will be to define a function that stacks different observations of the same measure while attaching the appropriate year identifier.

```python
#Define code:year map
yr_map={'189':1989,
        '190':1990,
        '192':1992,
        '196':1996,
        '199':1999,
        '200':2000,
        '204':2004,
        '209':2007,
        '208':2008,
        '210':2010}

#Grab front portion of indicators
concepts=list(set([x[:6] for x in d if x not in ['Areaname','STCOU']]))

#Define function to stack data observations
def concept_stack(cncpt):
    #Create container to facilitate concatenation of observations
    var_obs=[]
    #For each instance of a concept...
    for var in [x for x in d if x.startswith(cncpt)]:
        #...define a temp DF...
        temp_df=d[['Areaname','STCOU',var]]
        #...add a year identifier...
        temp_df['year']=yr_map[var[6:9]]
        #..rename the variable to just the concept portion...
        temp_df2=temp_df.rename(columns={var:cncpt})
        #...and throw it in the list
        var_obs.append(temp_df2)
    #Concatenate temporary DFs together
    var_series=pd.concat(var_obs).set_index(['Areaname','STCOU','year'])
    return var_series

#Create container for concept specific sets
concept_data=[]

#For each concept...
for concept in concepts:
    #...throw the concept in the container
    concept_data.append(concept_stack(concept))

#Concatenate along the horizontal axis
d2=pd.concat(concept_data,axis=1).reset_index('Areaname')

d2.head(25)
```

## Interpolation

So we are now in ok shape regarding the long format.  However, to interpolate, we need all of the years in the range to be represented.  Thus, area-wise reindexing is in order.

```python
#Create a container for reindexed subsets
reindexed=[]

#For each area...
for code in set(d2.index.get_level_values('STCOU')):
    #...subset via the code...
    code_sub=d2.ix[code]
    #...add the code id back in...
    code_sub['STCOU']=code
    #...reindex to the 1989-2010 time period
    reindexed.append(code_sub.reindex(range(1989,2011)))

#Re-concatenate data
d3=pd.concat(reindexed).reset_index().rename(columns={'index':'year'})
    
#Fill in ID info
d3['Areaname'].fillna(method='ffill',inplace=True)
d3['STCOU'].fillna(method='ffill',inplace=True)

#Create dictionary mapping codes to area names
code_area_map=dict(zip(d3['STCOU'],d3['Areaname']))

#Drop Areaname (it's redundant at the current time)
d3.pop('Areaname')

#Set index
d3.set_index(['STCOU','year'],inplace=True)

    
d3.sortlevel(0)
```

We need to interpolate, but `interpolate()` seems to `ffill()` instead of leaving out of range values alone...

Consequently, we will have to limit the time scope of the interpolation by variable.  The last year of data for each variable is as follows:

    {'EDU685':2007,
     'AGE050':2010,
     'PVY020':2007,
     'ELE025':2008}

```python
#Generate variable to last data year map
var_lstyr_map={'EDU685':2007,
               'AGE050':2010,
               'PVY020':2007,
               'ELE025':2008}

#Create container for county sets
cty_dfs=[]

#For each county code...
for code in sorted(set(d3.index.get_level_values(level='STCOU'))):
    #print code
    #...generate a container for county specific series...
    cty_ser=[]
    #...and for each variable...
    for var in var_lstyr_map.keys():
        #...create an interpolated version of the related series up through the last data year...
        temp_ser=DataFrame(d3.ix[code][var].ix[:var_lstyr_map[var]].interpolate())
        #print temp_ser.head(3)
        #...and capture it...
        cty_ser.append(temp_ser)
    #...consolidate the series into one DF...   
    cty_df=pd.concat(cty_ser,axis=1)
    #...reincorporate the county code...
    cty_df['STCOU']=code
    #...and throw it in the big container of county sets
    cty_dfs.append(cty_df)

#Consolidate into one DF
cty_interp=pd.concat(cty_dfs)

cty_interp
```

To deal with the missing values, we could minimize the impact of these observations would by setting them to the year specific means (thereby capturing the temporal trend while not differentiating one county from another in a given year).  However, the missing values are a function of year.

Alternatively, to retain these observations, we must rely on the county mean across years.  This has the effect of capturing the "fixed effect" of the variable in question, a general propensity.  The problem here is that these averages would be out of whack with other covariates in a given year.

Probably the best alternative is an attempt to extrapolate beyond the first and last values.  We have a tenuous basis for assumption regarding the gradient of the extrapolation, but since we are interpolating anyway, perhaps a linear extrapolation isn't the worst thing in the world.  Let us build an extrapolator.  In a nutshell, if the value exists we will use it.  If the value does not exist, we will increment the previous value.  We will do this by way of a cumulative sum variable (`ret_val`), with the increment (`inc`) defined by the average change in the previous three periods.  Naturally, the index must align with the original data.

```python
def extrapolate(ser,inc,ind):
    '''Function generates a new variable with extrapolated values where the data were previously missing'''
    #Create container for old and extrapolated values
    exp_list=[]
    
    #Initialize cumulative sum variable
    ret_val=NaN
    
    #For each variable in the relevant series...
    for i,val in enumerate(ser):
        #...if the value exists...
        if pd.notnull(val):
            #...set our cumulative sum variable to reflect that value...
            ret_val=val
            #...and throw it in the list
            exp_list.append(ret_val)
        #...if the value does not exist...
        else:
            #...increment the latest value...
            ret_val=ret_val+inc
            #...and throw it in the list
            exp_list.append(ret_val)
            
    return Series(exp_list,index=ind)
```

Now we need to apply our function for each variable/county combination.  Note that if we only proceeded variable wise, we would interpolate over the missing values for counties in the first few years.

```python
#Create a container for all county specific sets
cty_exp_df_list=[]

#Create a dictionary to aid in the construction of individual county specific sets
cty_exp_dict={}

#For each county...
for code in sorted(set(cty_interp['STCOU'])):
    #...and each variable...
    for var in var_lstyr_map.keys():
        #...create a version of the county specific series that extrapolates over missing values on the back end...
        cty_exp_dict.update({var:extrapolate(cty_interp[var][cty_interp['STCOU']==code],\
                                      cty_interp[var].diff()[(cty_interp[var].notnull()) &\
                                                                  (cty_interp['STCOU']==code)].tail(3).mean(),\
                                      cty_interp[cty_interp['STCOU']==code].index)})
    #...convert the dict of those series into DF...
    temp_df=DataFrame(cty_exp_dict)
    #...reintegrate county code...
    temp_df['STCOU']=code
    #...throw the new DF into the container for all county specific sets
    cty_exp_df_list.append(temp_df)
    
#Concatenate counties back together
cty_interp2=pd.concat(cty_exp_df_list)
```

## Reincorporating/Fixing County Name

Now that we have our observations, we can regenerate the county name (which was a bit of a nuisance above).

```python
cty_interp2['cty_temp']=cty_interp2['STCOU'].map(code_area_map)

cty_interp2.head()
```

Let's subset to relevant counties (those in Colorado), and split off the county name.

```python
#Subset to CO
co_cty=cty_interp2[cty_interp2['STCOU'].apply(lambda x: str(x).startswith('8'))]

#Create cty variable
co_cty['cty']=co_cty['cty_temp'].apply(lambda x: x.split(',')[0])

#Drop county totals
co_cty2=co_cty[co_cty['cty'] != 'COLORADO']

#Drop cty_temp
co_cty2.pop('cty_temp')

#Rename columns
# co_cty2.columns=['edu_yrs','med_age','pov_pop','perc_dem','stcou','cty']
co_cty2.columns=['med_age','edu_yrs','perc_dem','pov_pop','stcou','cty']

co_cty2.head(100)
```
