# Tree-Based Exploration of the County Ballot Initiatives in Colorado

In *CO_Ballot_Initiatives_Prep*, we cleaned the aggregated data collected by [Colorado Counties, Inc.](http://ccionline.org/) capturing all county-level ballot initiatives for sales/property/debt increases and spending limit waivers in Colorado over the 1993-2013 time period.  In this script, we are going to explore this data to try to uncover patterns in successful ballot initiatives.

The nature of this data makes useful the ability to uncover sequential chains/conditional subgrouping.  For example, passage of a debt increase is unlikely to depend on a non-existent sales rate change.  Given that a ballot issue is for a debt increase, only then does the size of the increase come into play.  This makes trees useful because the interior nodes offer decision points that uncover the utility of certain variables.  Again, if the interior node is whether or not a debt increase exists, downstream nodes are likely to involve the size of the debt increase.

Let's load `rmagic` and push our newly cleaned data set into the R namespace.

```python
%load_ext rmagic
```

```python
%%R 
#Establish workind directory ... again
workdir<-'/home/choct155/dissertation/TEL/CO_lit/'

#Read in data
vts<-read.csv(paste(workdir,'clean_votes.csv',sep=''))
attach(vts)

#Inspect head
print(head(vts))

#Load desired packages
library(tree)
library(ggplot2)

#Check out length of data compared to length of complete case data
print('NUMBER OF ROWS IN DATASET')
print(nrow(vts))
print('NUMBER OF COMPLETE CASES IN DATASET')
print(length(complete.cases(vts)))
```

Let's take a look at the distribution of `results`, the variable capturing the percentage of yes votes in each initative.

```python
%%R -w 900 -h 500

print(ggplot(vts,aes(x=results)) + 
      geom_histogram(binwidth=5,colour='white',fill='#B50724') + 
      labs(title='Distribution of County Ballot Initiative Results') +
      xlab('Yes Vote %') +
      ylab('Frequency') +
      theme_bw())
```

## Exploratory Trees (within Vote data)

As it turns out, we actually have pretty good variance in the vote distribution. Let's check out a tree seeking to predict the binary pass/fail indicator (pass) based solely on the type of action considered.

```python
%%R
#Fit tree to data
tree.pass=tree(pass~spend_waiver+sales+property+debt,data=vts)

#View tree fit results
print(summary(tree.pass))
print(plot(tree.pass))
print(text(tree.pass,pretty=0))
```

Ok, so what is this telling us?  It says that the two most important binary action predictors `sales` and `property`.  The tree also shows that, on average, sales and property tax increase proposals are losers.  We know this because the corresponding averages of `pass` at the terminal nodes are both below 50%.  What happens if we run the same model on the interval version of the outcomes, `results`?

```python
%%R
#Fit tree to data
tree.results=tree(results~spend_waiver+sales+property+debt,data=vts)

#View results of tree fit
print(summary(tree.results))
print(plot(tree.results))
print(text(tree.results,pretty=0))
```

As expected, the tree fit on `results` corroborates the tree fit on `pass`.  This is no surprise because the `pass` value is derived from the `results` value.  It is, however, interesting to see that while non-property/sales initiatives are far more likely to pass, the majorities could not be described as overwhelming.  

Let us fit the entire model, less the unused outcome variable (`pass` | `results`) and the county identifier (`cty`).  We can also exclude the variables used to generate the target and source dummies (`s_target`,`s_source`,`target`).

```python
%%R
#Generate subset list of variables
var_sub<-names(vts)[!names(vts) %in% c('cty','results','pass','s_target','s_source','target')]

#Generate formulae for the tree function - pass evaluation
p_formula<-as.formula(paste('pass~',paste(var_sub,collapse='+')))

#Fit tree to data
tree.pfull=tree(p_formula,data=vts)

#View tree fit results
print(summary(tree.pfull))
print(plot(tree.pfull))
print(text(tree.pfull,pretty=20))
print(tree.pfull)
```

These results are interesting for a few reasons:

1. The threshold for sales & use rate changes is actually quite high at 18.5%.  Initiatives below this threshold succeed 61.7% fo the time.  It must be kept in mind, however, that this actually includes instances in which a sales action is not being taken (which are scored as `sales_change = 0`).  For this reason, it may be useful to run trees on subsets that include only one type of revenue action.
2. Non-property tax related measures are more likely to pass (71% vs 46%), and within property measures, no levy increases over 6.95 mills are passed.
3. Public safety appears to be a difficult rationale to swallow.

## Exploratory Trees (inclusive of covariate data)

Ultimately, these votes are not made in a vacuum.  Let's run the same exploratory trees with covariate data included.

```python
%%R

#Read in data
vts_cov<-read.csv(paste(workdir,'clean_votes_cov.csv',sep=''))

#Detach small set and attach new set
#detach(vts_cov)
attach(vts_cov)

print(search())
```

This view ought to give us insight into the strength of the ballot versus the strength of environmental factors.

```python
%%R
#Generate subset list of variables
var_sub2<-names(vts_cov)[!names(vts_cov) %in% c('cty','results','pass','s_target','s_source','target')]

#Generate formulae for the tree function - pass evaluation
p_formula2<-as.formula(paste('pass~',paste(var_sub2,collapse='+')))

#Fit tree to data
tree.pfull2=tree(p_formula2,data=vts_cov)

#View tree fit results
print(summary(tree.pfull2))
print(plot(tree.pfull2))
print(text(tree.pfull2,pretty=20))
print(tree.pfull2)
```

Wow, that got a little out of control in a hurry with the addition of all the new variables.  Perhaps the most interesting part of this, however, is that the individual ballot issues dropped completely off the tree.  All of the interior nodes now refer to the covariates.  This may be a case of overfitting however.  We know that standalone trees have less predictive power than we'd like, so rather than mess around with pruning this tree, let's employ a more powerful technique.

## Random Forest Exploration

With Random Forests, trees are built from orthogonal subsets of the available predictors and then combined into a composite predictor.

```python
%%R

print(nrow(vts_cov[complete.cases(vts_cov),]))
```

```python
%%R

library(randomForest)

#Identify complete cases
vts_cov2=vts_cov[complete.cases(vts_cov),]

#Generate training set
train = sample(1:nrow(vts_cov2),300)

#Fit random forest
rf.pass=randomForest(p_formula2,data=vts_cov2,subset=train,mtry=10,ntree=5000)
print(rf.pass)
```

Hmmm, can this really be correct?  It's one thing if the explained variance is not as high as desired, but this is abysmal...

Let's try varying up the number of predictors considered for each tree.

```python
%%R

oob.err = double(12)
test.err = double(12)
for (mtry in seq(5,60,5)) {
    fit = randomForest(p_formula2,data=vts_cov2,subset=train,mtry=mtry,ntree=5000)
    oob.err[mtry] = fit$mse[400]
    pred = predict(fit, vts_cov2[-train, ])
    test.err[mtry] = with(vts_cov2[-train, ], mean((pass - pred)^2))
    cat(mtry, " ")
}
```

```python
%%R

print(oob.err)
print(test.err)
```

```python
%%R

matplot(1:mtry, cbind(test.err, oob.err), pch = 19, col = c("red", "blue"), 
    type = "b", ylab = "Mean Squared Error")
legend("topright", legend = c("OOB", "Test"), pch = 19, col = c("red", "blue"))
```
