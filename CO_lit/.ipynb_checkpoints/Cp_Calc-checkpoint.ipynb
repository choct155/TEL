## [Variance](http://en.wikipedia.org/wiki/Variance)

In a general way, variation is just when things differ from each other.  From a statistical point of view, variation is when things we measure cover more than one value.  For example, we could take the temperature in each room of a house to understand the variation in temperature within said house.  To get more specific, when we talk about variance, we try to capture this general variation relative to a specific reference point.  This point is usually the average.  To use the house example, in the course of calculating the variance, we would subtract the average from each of our individual readings.

The explicit definition of variance is as follows:

$\sigma^2=\sum_i^n(x_i-\bar x)^2$

where

+ $\sigma^2$ = variance ($\sigma$ = standard deviation)
+ $x_i$ = the value of observation $i$
+ $\bar x$ = the average value of all observations
+ $x_i-\bar x$ = deviation from the mean

In words, the variance of a given collection of data is *the sum of squared deviations from the mean*.  We use squared deviations (as opposed to simple deviations) because we do not care whether the value of a given observation is above or below the mean.  We only care about how far it is from the mean, and squaring does away with the sign.

We can use our house example to get an idea of how it's calculated.  Let's say we have a four room house with horrific irregularities in insulation.

Room ($i$)|Temperature ($x_i$)
:--:|:---:
1|50
2|60
3|70
4|80

We know that the average temperature ($\bar x$) is 65 degrees.  We can calculate the variance ($\sigma^2$) as follows:

$\sigma^2=\sum_i^n(x_i-\bar x)^2=(50-65)^2+(60-65)^2+(70-65)^2(80-65)^2=225+25+25+225=500$

Since we know that variance is the standard deviation squared, we can also find the standard deviation ($\sigma$):

$\sqrt{\sigma^2}=\sqrt{500}=22.360679774997898=\sigma$

Both standard deviation and variance are measures of spread that provide basically the same information.  Given the Six Sigma focus on $\sigma$, we will just talk about standard deviation from here on out.

#### Please Tell Me There is a Reason for All of This

A reasonable person might ask ... why the heck we care about calculating the variance and/or standard deviation?  What does this tell me about my data?  It turns out that the standard deviation is a discriminating characteristic of a *data generating process* (DGP), and a DGP is nothing more than whatever process has given rise to the data.  In our house example, whatever process governs the heat in the house is the DGP.  If we were looking at the amounts of insurance claims, the DGP would be the process that gives rise to insurance claim amounts we see.  Defining the drivers of the claim amounts (in statistical terms, defining the parameters of the claim amount function) is the real trick, and the point of doing this kind of analysis.

Admittedly, this is a little abstract, but the reason for bringing up DGP is that we need to distinguish between samples and populations.  A sample is a subset of observations collected from an exhaustive group.  If we want to say something about a million claims, we might draw a subset of 1000 claims to run our analysis on for one reason or another.  These 1000 claims are a sample from the entire population of one million claims.  This population was generated directly by our DGP.  If we pull multiple samples from our million-strong data set and calculate their standard deviations, they will probably differ from each other to some extent.  This would occur just because of the claims we happened to get in our sample being different.  *Even though the same DGP generated all of our million claims, the standard deviations of the samples can be different.*  

Why don't we just calculate the standard deviation of the whole group?  Wouldn't that make all of this nonsense about DGP and sampling irrelevant?  Indeed it would, but this is not always a feasible option.  In fact, it is this limitation that makes statistics useful.  It helps us describe data in a way that we can digest, when we can only see part of the data at any given time.

This was a long way of saying that the standard deviation helps us distinguish one DGP from another, but the intuition is important.  Just like we use hair color, height, and weight to distinguish people from one another, we use distributional parameters (like average and standard deviation) to distinguish DGPs.  Among other things, this helps us understand if the DGP has changed in response to something.  If we know that the variation in claim amounts jumped during the recession, we have *partial* evidence that people changed the way they thought about making claims.

Getting the standard deviation also allows us to make claims about one sample.  In particular, it helps us understand how likely certain values are to be present in the sample.  In the context of a normal distribution (as we will expand on below), a value that is one standard deviation from the mean is much more likely to occur than a value that is three standard deviations from the mean.

## [Normal (Gaussian)](http://en.wikipedia.org/wiki/Normal_distribution) Data

To solidify this DGP business and talk a bit about distributions, let's look at some data.  From my limited exposure, it seems like the Six Sigma approach relies heavily on the [Bell Curve](http://en.wikipedia.org/wiki/The_Bell_Curve), which concerns data generated from a normal (or Gaussian) distribution.  Looking at data generated from this distribution is probably a good place to start.  

The normal distribution is a very commonly used distribution because it is actually often a pretty decent approximation of real world DGPs when enough observations are collected.  It is defined by two characteristics:  mean and our good friend,standard deviation.

Let's say we wanted to simulate some data from this distribution.  We need to specify which average and standard deviation we want to use.  (The normal distribution is defined by *a relationship* between the mean and standard deviation, *not by specific values for either characteristic*.)  Why don't we check out a distribution that centers on zero ($\bar x=0$) and has a standard deviation of one ($\sigma=1$).  Here are 100 data points generated by the normal DGP defined by these parameters.

```python
#Define parameters
mean=0
sigma=1

#Generate data
x=np.random.normal(mean,sigma,100)

x
```

It's kind of a lot to absorb, and it's only 100 values!  We already know the mean and standard deviation (since we defined them), but what does this data actually look like?

```python
#Generate plot object
fig,ax=plt.subplots()

#Plot distribution
sb.kdeplot(x,shade=True,ax=ax)

#Set face color
ax.patch.set_facecolor('white')

#Set title
ax.set_title('Distribution of 100 Random Normal Values');
```

Well that kind of looks like a Bell Curve, but also kind of not like a Bell Curve.  What's going on here?  We only have a sample of 100 values.  It turns out that if we add more observations, it will look more like the textbook curve.

```python
#Generate plot object
fig,ax=plt.subplots(1,3,figsize=(18,5))

#Plot distribution
sb.kdeplot(np.random.normal(mean,sigma,100),shade=True,ax=ax[0])
sb.kdeplot(np.random.normal(mean,sigma,1000),shade=True,ax=ax[1])
sb.kdeplot(np.random.normal(mean,sigma,10000),shade=True,ax=ax[2])

#Set face color
ax[0].patch.set_facecolor('white')
ax[1].patch.set_facecolor('white')
ax[2].patch.set_facecolor('white')

#Set title
ax[0].set_title('Distribution of 100 Random Normal Values')
ax[1].set_title('Distribution of 1,000 Random Normal Values')
ax[2].set_title('Distribution of 10,000 Random Normal Values');
```

Just as an aside, we call this behavior consistency.  As we add more observations, our sample distribution converges with (becomes more like) the population distribution.

In any event, you can think of the height of the curve as a representation of how many data points have the value shown on the $x$ axis.  Notice how most of the values are clustered around the middle.  This goes back to the earlier point, as we increase the distance from the center of the normal distribution we are less likely to see those values.

How does $\sigma$ come into play explicitly?  We can get a better sense by evaluating three combinations of our key characteristics for the normal distribution ($\bar x$ & $\sigma$).

```python
#Define parameters
mean2=0
sigma2=1

mean3=0
sigma3=3

mean4=1
sigma4=1

#Generate data
x2=np.random.normal(mean2,sigma2,100000)
x3=np.random.normal(mean3,sigma3,100000)
x4=np.random.normal(mean4,sigma4,100000)

#Generate plot object
fig,ax=plt.subplots(3,figsize=(15,20),sharex=True,sharey=True)

#Plot distribution
sb.kdeplot(x2,shade=True,ax=ax[0])
sb.kdeplot(x3,shade=True,ax=ax[1])
sb.kdeplot(x4,shade=True,ax=ax[2])

#Set face color
ax[0].patch.set_facecolor('white')
ax[1].patch.set_facecolor('white')
ax[2].patch.set_facecolor('white')

#Set title
ax[0].set_title(r'Normal Distribution $\bar x$=0, $\sigma$=1',fontsize=22)
ax[1].set_title(r'Normal Distribution $\bar x$=0, $\sigma$=3',fontsize=22)
ax[2].set_title(r'Normal Distribution $\bar x$=1, $\sigma$=1',fontsize=22)

#Set mean line
ax[0].vlines(x2.mean(),0,.5,linewidth=3)
ax[1].vlines(x3.mean(),0,.22,linewidth=3)
ax[2].vlines(x4.mean(),0,.5,linewidth=3)

#Set sigma count lines
ax[0].vlines(x2.mean()+x2.std(),0,.5,linewidth=.5)
ax[0].vlines(x2.mean()+2*x2.std(),0,.5,linewidth=.5)
ax[0].vlines(x2.mean()+3*x2.std(),0,.5,linewidth=.5)

ax[0].vlines(x2.mean()-x2.std(),0,.5,linewidth=.5)
ax[0].vlines(x2.mean()-2*x2.std(),0,.5,linewidth=.5)
ax[0].vlines(x2.mean()-3*x2.std(),0,.5,linewidth=.5)

ax[1].vlines(x3.mean()+x3.std(),0,.22,linewidth=.5)
ax[1].vlines(x3.mean()+2*x3.std(),0,.22,linewidth=.5)
ax[1].vlines(x3.mean()+3*x3.std(),0,.22,linewidth=.5)

ax[1].vlines(x3.mean()-x3.std(),0,.22,linewidth=.5)
ax[1].vlines(x3.mean()-2*x3.std(),0,.22,linewidth=.5)
ax[1].vlines(x3.mean()-3*x3.std(),0,.22,linewidth=.5)

ax[2].vlines(x4.mean()+x4.std(),0,.5,linewidth=.5)
ax[2].vlines(x4.mean()+2*x4.std(),0,.5,linewidth=.5)
ax[2].vlines(x4.mean()+3*x4.std(),0,.5,linewidth=.5)

ax[2].vlines(x4.mean()-x4.std(),0,.5,linewidth=.5)
ax[2].vlines(x4.mean()-2*x4.std(),0,.5,linewidth=.5)
ax[2].vlines(x4.mean()-3*x4.std(),0,.5,linewidth=.5)

#Annotate distribution points
ax[0].annotate('MEAN',xy=(-.4,.47),rotation=90,fontsize=12)
ax[0].annotate('-1$\sigma$',xy=(-1.4,.45),rotation=90,fontsize=12)
ax[0].annotate('-2$\sigma$',xy=(-2.4,.45),rotation=90,fontsize=12)
ax[0].annotate('-3$\sigma$',xy=(-3.4,.45),rotation=90,fontsize=12)
ax[0].annotate('1$\sigma$',xy=(.6,.45),rotation=90,fontsize=12)
ax[0].annotate('2$\sigma$',xy=(1.6,.45),rotation=90,fontsize=12)
ax[0].annotate('3$\sigma$',xy=(2.6,.45),rotation=90,fontsize=12)

ax[1].annotate('MEAN',xy=(-.4,.19),rotation=90,fontsize=12)
ax[1].annotate('-1$\sigma$',xy=(-3.4,.18),rotation=90,fontsize=12)
ax[1].annotate('-2$\sigma$',xy=(-6.4,.18),rotation=90,fontsize=12)
ax[1].annotate('-3$\sigma$',xy=(-9.4,.18),rotation=90,fontsize=12)
ax[1].annotate('1$\sigma$',xy=(2.6,.18),rotation=90,fontsize=12)
ax[1].annotate('2$\sigma$',xy=(5.6,.18),rotation=90,fontsize=12)
ax[1].annotate('3$\sigma$',xy=(8.6,.18),rotation=90,fontsize=12)

ax[2].annotate('MEAN',xy=(.6,.47),rotation=90,fontsize=12)
ax[2].annotate('-1$\sigma$',xy=(-.4,.45),rotation=90,fontsize=12)
ax[2].annotate('-2$\sigma$',xy=(-1.4,.45),rotation=90,fontsize=12)
ax[2].annotate('-3$\sigma$',xy=(-2.4,.45),rotation=90,fontsize=12)
ax[2].annotate('1$\sigma$',xy=(1.6,.45),rotation=90,fontsize=12)
ax[2].annotate('2$\sigma$',xy=(2.6,.45),rotation=90,fontsize=12)
ax[2].annotate('3$\sigma$',xy=(3.6,.45),rotation=90,fontsize=12);
```

So what can we see here?  For starters, we can see that changing the mean ($\bar x$) changes the location of the *center* of the distribution.  Secondly, we see that changing the standard deviation ($\sigma$) changes the *width* of the distribution.  

Hmmm, that's interesting.  In all cases, we see six standard deviation markers, but we have differences in the widths of the distribution.  Again, standard deviation is calculated directly from the data.  The length of one standard deviation depends on the data in question.  In fact, this property is what makes it useful for comparing distributions.  It tells us that *a larger standard deviation indicates a wider distribution.*  Said differently, since we know that the standard deviation is equal to the sum of squared deviations from the mean, *larger standard deviations mean that we have more points that are far from the mean compared to distributions with smaller standard deviations.*  This larger incidence of variation may be cause for concern depending on the process we are examining.  If we think the process should be highly repeatable, high amounts of variation are probably something we want to fix.

## Visualizing Distributions

Up until now, the plots have been examples of a [kernel density plot](http://en.wikipedia.org/wiki/Kernel_density_estimation).  The details of its construction are not important right now, but suffice it to say, it gives us a view of where the values in the data are.  Again, the higher the curve gets, the more data points have that value.  

A similar view can be given by a [histogram](http://en.wikipedia.org/wiki/Histogram).

```python
#Generate plot object
fig,ax=plt.subplots()

#Plot histogram
ax.hist(x2)

#Set facecolor
ax.patch.set_facecolor('white')

#Set title
ax.set_title(r'Histogram Version of Normal Distribution $\bar x$=0, $\sigma$=1',fontsize=16);
```

Basically this view just counts all of the values in a given bin (e.g. between 0 and .5) and uses this value as the height of the bar.

An alternative view that we talked about on Saturday is the [box plot](http://en.wikipedia.org/wiki/Box_plot).  This is a very popular tool because it is so compact that it lends itself to use as a tool to compare multiple groups.  In effect, it is sharing a subset of the information provided in the kernel density plots and histograms.  To see how this can be, we can relate the two directly.

```python
#Generate plot object
fig,ax=plt.subplots(2,figsize=(15,10),sharex=True)

#Plot distribution
sb.kdeplot(x2,shade=True,ax=ax[0],color='#CC0033')
sb.boxplot(x2,ax=ax[1],vert=False,color='#CC0033')

#Set facecolor
ax[0].patch.set_facecolor('white')
ax[1].patch.set_facecolor('white')

#Set lines
ax[1].axvline(np.median(x2),.1,2.2,clip_on=False,linewidth=3,color='k')
ax[1].axvline(np.median(x2)+stats.scoreatpercentile(x2,75),.9,2.2,clip_on=False,linestyle='--',color='k')
ax[1].axvline(np.median(x2)+stats.scoreatpercentile(x2,25),.9,2.2,clip_on=False,linestyle='--',color='k')

#Annotate
ax[0].set_ylim([0,.5])
ax[0].annotate('MEDIAN',xy=(-.15,.47),rotation=90,fontsize=12)
ax[0].annotate('P=.25',xy=(-.15++stats.scoreatpercentile(x2,25),.47),rotation=90,fontsize=12)
ax[0].annotate('P=.75',xy=(-.15++stats.scoreatpercentile(x2,75),.47),rotation=90,fontsize=12)

#Set title
ax[0].set_title('Comparison of Kernel Density and Box Plots',fontsize=22);
```

The box captures the [median](http://en.wikipedia.org/wiki/Median) with its centerline, and the 25th and 75th percentiles with the outer edges.  The "whiskers" (the vertical lines at the end of the horizontal lines) capture the minimum and maximum values *that are not considered outliers*.  All of those points beyond the whiskers are outliers.

In this case, since we are using the normal distribution and we have many observations (which means we are converging on the pure normal distribution), the median and mean overlap.  If we used a different distribution, we could see the median and mean diverge.

The plot below demonstrates this concept with the [Beta distribution](http://en.wikipedia.org/wiki/Beta_distribution).

```python
#Generate plot object
fig,ax=plt.subplots(2,figsize=(15,10),sharex=True)

b=np.random.beta(1,.5,1000)

#Plot distribution
sb.kdeplot(b,shade=True,ax=ax[0],color='#CC0033')
sb.boxplot(b,ax=ax[1],vert=False,color='#CC0033')

#Set facecolor
ax[0].patch.set_facecolor('white')
ax[1].patch.set_facecolor('white')

#Set lines
ax[1].axvline(np.median(b),.1,2.2,clip_on=False,linewidth=3,color='k')
ax[1].axvline(np.mean(b),.9,2.2,clip_on=False,linestyle='--',color='k')

#Annotate
ax[0].annotate('MEDIAN',xy=(np.median(b)-.03,1.6),rotation=90,fontsize=12)
ax[0].annotate('MEAN',xy=(np.mean(b)-.03,1.6),rotation=90,fontsize=12)

#Set title
ax[0].set_title('Mean-Median Divergence',fontsize=22);
```

When we have several distributions to compare, boxplots are cleaner alternatives to kernel density plots or histograms.

```python
#Generate plot object
fig,ax=plt.subplots(2,figsize=(15,10),sharex=True)

#Generate more random data
x5=np.random.normal(-1,2,100000)
x6=np.random.normal(-4,1,100000)
x7=np.random.normal(3.5,4,100000)

#Define colors
cols=['#b2182b','#ef8a62','#fddbc7','#d1e5f0','#67a9cf','#2166ac']

#Plot distribution
sb.kdeplot(x2,shade=True,ax=ax[0],color=cols[0])
sb.kdeplot(x3,shade=True,ax=ax[0],color=cols[1])
sb.kdeplot(x4,shade=True,ax=ax[0],color=cols[2])
sb.kdeplot(x5,shade=True,ax=ax[0],color=cols[3])
sb.kdeplot(x6,shade=True,ax=ax[0],color=cols[4])
sb.kdeplot(x7,shade=True,ax=ax[0],color=cols[5])

sb.boxplot([x2,x3,x4,x5,x6,x7],ax=ax[1],vert=False,color=cols)


#Set facecolor
ax[0].patch.set_facecolor('white')
ax[1].patch.set_facecolor('white')
```

## Example Problem

Here is the application you shared with me on Saturday.  $C_p$ is a variance metric that seeks to describe the "width" of a process.  It is defined as follows:

$C_p=\frac{USL-LSL}{6\sigma}$

where:

$USL$ = Upper Specification Limit
$LSL$ = Lower Specification Limit

The problem is as follows:
>The Upper Specification Limit equals 145 minutes and the Lower Specification Limit equals 115 minutes.  After taking a sample of process cycle times, ou find the average process cycle time equals 128 minutes with a standard deviation of 12 minutes.  What is the $C_p$ of this process?

```python
#Define parameters of the equation
USL=140
LSL=115
sigma=12

#Calculate Cp
Cp=(USL-LSL)/float((6*sigma))

print 'THE Cp INDEX VALUE IS',Cp
```

Given what we now know about variation, we can think about this measure in a more abstract way.  It appears that you are given specification limits and then $Cp$ is used to compare what the actual data reveals about the process to these specs.  While it may be that the global specifications ($USL$ & $LSL$) were once defined based upon the performance of the process (and thus based upon previous data), at this stage we want to know how well our process sample compares with the global specifications.  In other words, given the current state of the process, we want to get an idea of how often the cycle times will be within the time range defined by the $[USL,LSL]$ interval.

If the numerator ($USL-LSL$) is large compared to the denominator ($6\sigma$), then we have comparatively low variation in our process and it's easier to for our process to fit inside the specification range (as long as the average time is close to the middle of the range like it is in the problem).  If the numerator is small compared to the denominator, the opposite is true.  We have so much variation  in our process that even if the average time is in the specification range, we will have a comparatively large number of times that are not.  In the problem above, we can see that the specification range is about a third of the time spanned by six standard deviations. 

Here is a depiction of the data identified in the problem.

```python
#Generate data
cp_data=np.random.normal(128,sigma,10000)

#Generate plot object
fig,ax=plt.subplots(2,figsize=(15,10),sharex=True)

#Plot distribution
sb.kdeplot(cp_data,shade=True,ax=ax[0],color=cols[0])
sb.boxplot(cp_data,ax=ax[1],vert=False,color=cols[0])

#Set facecolor
ax[0].patch.set_facecolor('white')
ax[1].patch.set_facecolor('white')

#Set lines
ax[1].axvline(USL,.1,2.2,clip_on=False,linewidth=3,color='k')
ax[1].axvline(LSL,.1,2.2,clip_on=False,linewidth=3,color='k')

#Annotate
ax[0].annotate('LSL',xy=(LSL-3,.03),rotation=90,fontsize=14)
ax[0].annotate('USL',xy=(USL-3,.03),rotation=90,fontsize=14)

#Set title
ax[0].set_title(r'$C_p$ Plot',fontsize=22);
```

If the standard deviation were higher, less of the values would fall in the specification range.  If it were lower, more of the values would fall in the specification range.

To wrap it up, we can look at these hypothetical situations.  For the second sample (`cp_data2`) we will set $\sigma$=5 minutes, and for the third sample (`cp_data3`), we will set $\sigma$=25 minutes.

```python
#Generate data
cp_data2=np.random.normal(128,5,10000)
cp_data3=np.random.normal(128,25,10000)

#Generate plot object
fig,ax=plt.subplots(2,figsize=(15,10),sharex=True)

#Plot distribution
sb.kdeplot(cp_data2,shade=True,ax=ax[0],color=cols[0],label=r'Low $\sigma$')
sb.kdeplot(cp_data3,shade=True,ax=ax[0],color=cols[5],label=r'High $\sigma$')
sb.boxplot([cp_data2,cp_data3],ax=ax[1],vert=False,color=[cols[0],cols[5]])

#Set facecolor
ax[0].patch.set_facecolor('white')
ax[1].patch.set_facecolor('white')

#Set lines
ax[1].axvline(USL,.1,2.2,clip_on=False,linewidth=3,color='k')
ax[1].axvline(LSL,.1,2.2,clip_on=False,linewidth=3,color='k')

#Annotate
ax[0].annotate('LSL',xy=(LSL-3,.03),rotation=90,fontsize=14)
ax[0].annotate('USL',xy=(USL-3,.03),rotation=90,fontsize=14)

#Set title
ax[0].set_title(r'$C_p$ Plot - Low and High $\sigma$',fontsize=22)

#Set legend
ax[0].legend(loc='best');
```

As can be seen, even though both processes have the same average value, a much higher amount of the times in the 'Low $\sigma$' process fall within the specification window.
