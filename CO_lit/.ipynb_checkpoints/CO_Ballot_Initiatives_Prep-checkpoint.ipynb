# Ballot Initiatives in Colorado - Data Preparation

In an effort to establish a better understanding of the constraints imposed by the requirement to vote on exemptions from TELs in Colorado, this script seeks to provide the means to explore the dynamics of individual votes and the socioeconomic context in which they occurred.  Specifically, this script processes the data for analysis.  We are using data collected by the [Colorado Counties, Inc.](http://ccionline.org/) organization.  These data capture relatively high-level descriptions of the matters on the ballot and voting results by county-year.

```python
#Basic data manipulation
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

#Text manipulation
import re
from fuzzywuzzy import process

#Display
from IPython.display import HTML
```

The data were originally housed in an Excel spreadsheet, and arranged for a different purpose than data analysis. Limited pre-processing has been conducted. The less trivial activity was pulling the time limits of individual issues out of the text for the `Sales & Use Tax/Mill Levy Increase` description.  This ended up being more time efficient because the time limits are described in a range of formats that complicate a simple logical extract.  (For example, in some situations a straightforward limit is described as X years, while in others it's provided as a time range between two dates of varying granularity.)

Capturing the percent changes/extensions for sales taxes is a tricky business.  We need to define a first pass function that captures the substring containing `%`, and report that string in a new variable.  We then need a function that will convert these strings to float, but will return missing values when applicable.  This should cover most cases, and then we can clean up the misses afterwards.  Note that if there is no action on sales taxes, then `sales_change==0`.  The same will be true of property and debt.

```python
def perc_extract(string):
    try:
        return ''.join([x[:x.index('%')] if '%' in x else '' for x in string.split(' ')])
    except:
        return ''

#Create string version of change in sales rate
votes['sales_dstr']=votes['Sales & Use Tax/Mill Levy Increase'].apply(lambda x: perc_extract(x))

def float_convert(x):
    try:
        return float(x)
    except:
        return 0.0

#Create float version of change in sales rate
votes['sales_change']=votes['sales_dstr'].apply(lambda x: float_convert(x))

#Drop interim string version
votes.pop('sales_dstr')

#Capture columns of interest
sales_change=['cty','Action','Sales & Use Tax/Mill Levy Increase','sales','sales_change']

#Change misses
votes['sales_change'].ix[11]=.05
votes['sales_change'].ix[12]=.5
votes['sales_change'].ix[44]=2
votes['sales_change'].ix[47]=0
votes['sales_change'].ix[74]=.25
votes['sales_change'].ix[102]=0
votes['sales_change'].ix[111]=.25
votes['sales_change'].ix[219]=.23
votes['sales_change'].ix[223]=.5
votes['sales_change'].ix[227]=.5
votes['sales_change'].ix[237]=.25
votes['sales_change'].ix[251]=1
votes['sales_change'].ix[252]=.23
votes['sales_change'].ix[282]=.5
votes['sales_change'].ix[283]=1
votes['sales_change'].ix[290]=.6
votes['sales_change'].ix[291]=.25
votes['sales_change'].ix[336]=1
votes['sales_change'].ix[384]=.25
votes['sales_change'].ix[390]=.15
votes['sales_change'].ix[392]=.4
votes['sales_change'].ix[629]=1
votes['sales_change'].ix[659]=0

#Remove sales score for sales exemptions and reductions
votes['sales'].ix[16]=0
votes['sales'].ix[177]=0
votes['sales'].ix[386]=0
votes['sales'].ix[396]=0
votes['sales'].ix[426]=0
votes['sales'].ix[434]=0
votes['sales'].ix[436]=0

# HTML(votes[votes['sales']==1][sales_change].to_html())
# votes['sales_change'].head()
```

Now we need a conceptually similar operation for property information.  This time, however, we need to extract the the string that precedes *mill* in the `Sales & Use ...` variable.  This will tell us, in most cases, how much the property rate is changing.  There is a wrinkle with the property information.  While most increases are expressed in mill terms, some are expressed in levy terms.  We will need a new variable (`prop_levy_change`) to capture the different format.

```python
def mill_extract(string):
    try: 
        return ''.join([string.split(' ')[i-1] if 'mil' in x or 'Mil' in x else '' for i,x in enumerate(string.split(' '))])
    except:
        return ''
    
#Extract string version of mill rate change
votes['prop_dstr']=votes['Sales & Use Tax/Mill Levy Increase'].apply(lambda x: mill_extract(x))

#Convert string version to float
votes['prop_change']=votes['prop_dstr'].apply(lambda x: float_convert(x))

def vol_extract(string):
    try:
        return ''.join([x if '$' in x else '' for x in string.split(' ')])
    except:
        return ''

#Extract string version of debt volumes
votes['prop_levy_dstr']=votes[votes['property']==1]['Sales & Use Tax/Mill Levy Increase'].apply(lambda x: vol_extract(x)).str.replace('[$,]','')

#Convert to float
votes['prop_levy_change_temp']=votes['prop_levy_dstr'].apply(lambda x: float_convert(x))

#Fill missing values with zero
votes['prop_levy_change']=votes['prop_levy_change_temp'].fillna(0)

#Drop string version
votes.pop('prop_dstr')
votes.pop('prop_levy_dstr')

#Remove liability reductions
votes['property'].ix[16]=0
votes['property'].ix[269]=0

#Fix misses
votes['prop_change'].ix[33]=.0394
votes['prop_levy_change'].ix[33]=0
votes['prop_change'].ix[121]=3
votes['prop_change'].ix[155]=9.8
votes['prop_change'].ix[231]=1.5
votes['prop_levy_change'].ix[235]=2800000
votes['prop_levy_change'].ix[236]=5100000
votes['prop_change'].ix[302]=2
votes['prop_change'].ix[340]=2
votes['prop_change'].ix[342]=5
votes['prop_change'].ix[344]=.575
votes['prop_levy_change'].ix[370]=3130000
votes['prop_change'].ix[388]=1
votes['prop_change'].ix[391]=.3
votes['prop_change'].ix[444]=.25
votes['prop_change'].ix[500]=1
votes['prop_change'].ix[560]=1.38
votes['prop_change'].ix[565]=3.75
votes['prop_levy_change'].ix[565]=0
votes['prop_levy_change'].ix[604]=1070000
votes['prop_change'].ix[610]=1
votes['prop_levy_change'].ix[634]=150000
votes['prop_change'].ix[647]=2.321
votes['prop_change'].ix[657]=2.321
votes['prop_change'].ix[657]=.5
votes['prop_change'].ix[681]=np.NaN

#Capture columns of interest
prop_change=['cty','Action','Sales & Use Tax/Mill Levy Increase','property','prop_change','prop_levy_change']

# HTML(votes[(votes['property']==1) & (votes['prop_change'].isnull())][prop_change].to_html())
votes[votes['prop_levy_change'].isnull()][prop_change]
```

Extracting the target for the expenditure requires functional aggregation.  We can define these functions, and then use keywords in the description to allocate each ballot initiative.  The approach used here will be to define dummies for each of our functional categories, and populate them conditional on a function being associated with given keywords in a dictionary mapping.

**UPDATE**:  After trying several associative strategies including the one above, it became apparent that the post-processing required was sufficiently labor-intensive as to offset the efficiency gains of the programmatic solutions of which I am aware.  Consequently, this was remedied by pre-processing the input set.  All specified revenue allocations were assigned the following categories under the `target` variable:

+ **Administrative** captures the business of government (studies, legislative tasks, etc.)
+ **Agriculture** captures explicitly identified agriculture expenditures;
+ **Debt** captures levy increases to pay down debt;
+ **Economic Development** captures explicitly defined economic development initiatives;
+ **Education** captures all school-related and library expenditures;
+ **Environment** captures all expenditure targets associated with open space, parks, trails, and water conservation;
+ **HSS** (health & social services) captures animal services, childcare, cultural functions, and housing;
+ **Infrastructure** captures all expenditures related to energy, roads & bridges, telecommunications, and transportation;
+ **Open** captures explicit statements indicating that the revenues were for *general* use;
+ **Public Safety** captures all expenditures related to emergency services and law enforcement; 
+ **Unknown** captures a single instance in which the target was specified, but the function was unclear; and,
+ **Unspecified** captures all instances which lacked specific direction about how the revenue was to be spent.

In general, the above categories are an attempt to match the revenue and debt increases/extensions to broad governmental service functions.  

Time limits embedded in the revenue/debt text were also problematic because they were expressed in different ways (distance between dates of differing formats, outright year statements, etc.).  They were also pulled out into a separate variable, `time_limit`, expressing the length of the revenue/debt provision in years when applicable.  A new variable, `rev_time` will take the values from `time_limit` and replace missing values with 99.

```python
#Fill missing values for time_limit in with 99
votes['rev_time']=votes['time_limit'].astype(float).fillna(99.)
```

An analagous procedure was followed for the spending waiver information, parsing out the information into three variables: `spend_source`, `spend_target`, and `spend_time`.  Unless a specified source, target, or time limit was explicitly provided, these values were left missing in the data.  All of the missing values for the source and target variables will be treated as though no restriction exists (NaN >> 'open), with the time variable similarly coded as 99.  In non-spending waiver observations, the variables should reflect no increased flexibility.  Thus, the source and target variables will be coded as 'no waiver', and the time variable as 0.

```python
#Replace missing values in spending waiver rows with 'open' and 99
votes['s_source']=np.where((votes['spend_waiver']==1) & (votes['spend_source'].isnull()),'open',votes['spend_source'])
votes['s_time']=np.where((votes['spend_waiver']==1) & (votes['spend_time'].isnull()),99,votes['spend_time'])
votes['s_target']=np.where((votes['spend_waiver']==1) & (votes['spend_target'].isnull()),'open',votes['spend_target'])

#Replace missing values in non-spending waiver rows with 'no waiver' and 0
votes['s_source']=np.where(votes['s_source'].isnull(),'no waiver',votes['s_source'])
votes['s_time']=np.where(votes['s_time'].isnull(),0,votes['s_time'])
votes['s_target']=np.where(votes['s_target'].isnull(),'no waiver',votes['s_target'])

# HTML(votes[votes['spend_waiver']==1][['cty','Action','Spending Limit Waiver','spend_source','spend_target','spend_time',\
#                                       's_source','s_target','s_time']].to_html())
```

## Dummy Generation

Now that we have trimmed back the variables, we are going to add many more back into the set.  The reason is that sometimes ballot issues have multiple expenditure targets.  In our current target variables `s_target` and `target`, multiple objectives are sometimes combined in one value instance (e.g. 'environment/hss').  While this was useful as a recording device, it muddies the water for modeling because 'hss', for example, may appear alone or in combination with another target.  This makes it difficult to interpret the impact of targeting 'hss'.  The same issue arises with `s_source`.

The fix is to make binary dummies for each of these targets.

```python
#Establish lists of possible targets
s_target_list=['open','no waiver','infrastructure','public safety','environment','education','unspecified','hss',\
               'debt','economic development','tourism']
s_source_list=['open','no waiver','sales','property','nontax','value','transfer','nonproperty','tourism','unknown',\
               'library','gaming','development']
r_target_list=['infrastructure','public safety','hss','environment','unspecified','education','open','tourism',\
               'administrative','economic development','agriculture','debt','unknown','limited','choice']


#For each spending waiver target...
for st in s_target_list:
    #...generate a new dummy variable indicating the presence of the target
    vts['st_'+st]=np.where(vts['s_target'].apply(lambda x: st in x),1,0)
    
#For each spending waiver target...
for ss in s_source_list:
    print 'ss_'+ss
    #...generate a new dummy variable indicating the presence of the target
    vts['ss_'+ss]=np.where(vts['s_source'].apply(lambda x: ss in x),1,0)

#For each spending waiver target...
for rt in r_target_list:
    #...generate a new dummy variable indicating the presence of the target
    vts['rt_'+rt]=np.where(vts['target'].apply(lambda x: rt in str(x)),1,0)    
    
#Write to disk
vts.to_csv(workdir+'clean_votes.csv')
    
vts[vts['target'].apply(lambda x: '/' in str(x))].head().T
```

The goal is to keep as many variables in the set as possible.  The theory rests on the `tels` group, but we want to control for everything else and allow the data to tell us what is relevant.  Nevertheless, some of these variables need to go, because they are redundant.  For example, there are four different population measures in the set (three total and one group quarters).  There are also three measures of industrial composition (we will stick with wages over employees and establishments).  Once these are removed, we will join the set to the votes data.

```python
#Identify redundant variables
weights=[x for x in cov.columns if x.endswith('_ap')]
industry=[x for x in cov.columns if x.startswith('est') or x.startswith('emp')]+['labforce']
misc=['num_hh','vac_hu','july_pop','july_hh_pop','gq_pop','defl','r_mhv']

#Build composite list of redundant variables
redundant=weights+industry+misc

#Identify subset of interest
var_sub=[x for x in cov.columns if x not in redundant]

#Subset covariates
cov_sub=cov[var_sub].set_index(['year','cty'])

#Join sets
vts_cov=vts.join(cov_sub).ix[1993:2009]

#Replace missing values with zero
for var in ['r_wage_info','r_wage_mfg','r_wage_nat_res','r_wage_other','r_wage_unclass']:
    vts_cov[var]=np.where(vts_cov[var].isnull(),0,vts_cov[var])


#Write to disk
vts_cov.to_csv(workdir+'clean_votes_cov.csv')

vts_cov.info()
```

## Slimming Down

The dataset written to disk above is constructed in a way that allows for some level of tailoring to the specific research task.  This is the reason that targets, for example, are split.  If the target information was listed in spending limit waiver column, it is `st_...` and if it's mentioned in the revenue change column, it is `rt_...`.  For most purposes, it is enough to simply know that earmarking had occurred.  For modeling purposes, we also don't need the variables upon which these are based (e.g. `spend_waiver`) or the grouping variables that simply indicate the type of change that has occured (e.g. `sales`).  The latter can be inferred by a non-zero change value.  For these reasons, we will provide reduce this set somewhat to provide an easier modeling input.  Let's start by consolidating the target variables

```python
#Create copy
vts_cov2=vts_cov.copy(deep=True)

#Identify set of concepts in s_target_list and r_target_list
target_common=set(s_target_list) & set(r_target_list)
target_s=set(s_target_list) - set(r_target_list)
target_r=set(r_target_list) - set(s_target_list)
target_rs=list(target_common)+list(target_s)+list(target_r)

#Instantiate new variables
for tgt in target_rs:
    vts_cov2[tgt]=0

#Set values from spending targets where possible
#For each target in the consolidated target list...
for tgt in target_rs:
    #...if the spend target version of it exists...
    if ('st_'+tgt in vts_cov2.columns):
        #...copy the values over
        vts_cov2[tgt]=vts_cov2['st_'+tgt]
        
# print set(target_rs)-set(list(target_s) + list(target_common))==target_r

#Set values from rev targets where possible, unless a 1 already exists from the spend side.
#In that case, leave it alone.
#For each target in the consolidated target list...
for tgt in target_rs:
    #...if the rev target version of it exists...
    if ('rt_'+tgt in vts_cov2.columns):
        #...leave a 1 value alone, otherwise copy the values over...
        vts_cov2[tgt]=np.where(vts_cov2[tgt]==1,1,vts_cov2['rt_'+tgt])

#For each target variable...
# for tgt in target_rs:
#     print vts_cov2[[var for var in vts_cov2 if tgt in var]].head(40)

#Drop old spend target variables
for var in list(target_s) + list(target_common):
    vts_cov2.pop('st_'+var)
    
#Drop old rev target variables
for var in list(target_r) + list(target_common):
    vts_cov2.pop('rt_'+var)
    
#Create list of upstream variables
upstream_vars=['spend_waiver','s_source','s_target','lim5_5','sales','property','debt','target','rev_change']

#Drop upstream variables
for var in upstream_vars:
    vts_cov2.pop(var)
```
