We need to define a few things:

+ The color maps we will use.  If a variable contains a p-value below the threshold in any given year, it will be colored with a diverging color scale.  If it does not, the color scheme will be monochromatic.
+ To set a color break point at a given p-value, we need to calculate the percentile value associated with said break point (aka - threshold) for each vector of variable-specific p-values.
+ We need general functions that will produce our plots in a consistent manner.  One will parse the data by model, and one by year.

```python
#Create color mapping with mpltools
bcr_rgba = [(0.02, 0.2, 0.4, 1),    # grayish blue, opaque
            (0.02, 0.2, 0.4, 0.3),  # grayish blue, transparent
            (0.4,  0.0, 0.1, 0.3),  # dark red, transparent
            (0.4,  0.0, 0.1, 1)]    # dark red, opaque
red_rgba= [(0.4,  0.0, 0.1, 0.3),  # dark red, transparent
           (0.4,  0.0, 0.1, 1)]    # dark red, opaque

#Define percentile position calculator
def perc_calc(seq,thresh):
    if thresh < seq.min():
        return 0
    elif thresh > seq.max():
        return 1
    else:
        perc=(thresh - seq.min())/(seq.max() - seq.min())
        return perc

#Create plotting object
fig,ax=plt.subplots(figsize=(18,4))
    
#Define function to plot coefficient estimates by model
def coef_mod(df_in,model,p,fig=fig,ax=ax):
    '''Function isolates a model and plots variable impacts on the model dependent.  
    The estimates are colored by p-value and sized by R-squared.'''
    #Subset to variable of interest
    df=df_in[df_in['model']==model].reset_index()
    #Identify percentile of the desired p-value
    thresh=perc_calc(df['pval'].values,p)
    #If the desired p-value is in the set...
    if thresh>0:
        #...create a diverging color scale that splits at the p-value...
        blue_clear_red = color.LinearColormap('blue_clear_red', bcr_rgba,
                                          index=[0, thresh, thresh, 1])
        #...and plot
        plot=ax.scatter(df.index,df['est'],c=df['pval'],cmap=blue_clear_red,s=5000*df['rsq'],alpha=.4)
    else:
        #...otherwise create a monochromatic scale...
        reds=color.LinearColormap('blue_clear_red', red_rgba,
                                          index=[0, 1])
        #...and plot
        plot=ax.scatter(df.index,df['est'],c=df['pval'],cmap=reds,s=5000*df['rsq'],alpha=.4)
    #Set title for plot
    ax.set_title(model+' (R^2 = '+str('{0:.4f}'.format(df['rsq'][0]))+')',fontsize=18)
    #Set x-axis scope
    ax.set_xlim(-.5,len(df['var'])+.5)
    #Set xticks
    ax.set_xticks(range(len(df)))
#     fig.subplots_adjust(hspace=10)
    #Set xtick labels
    ax.set_xticklabels(list(df['var']),rotation=20)
    #Generate an axis-specific colorbar
    fig.colorbar(plot,ax=ax,shrink=.9)
    #Adjust padding
    plt.tight_layout()
    
coef_mod(ols_res,'ols_rev',0.1,ax=ax)
```
