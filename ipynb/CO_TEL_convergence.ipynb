# Analysis of Fiscal Convergence in Colorado

Testing whether or not the constraints imposed by Tax and Expenditure Limitations (TELs) lead to local fiscal convergence is difficult to achieve with the prevailing models in this area.  Most of the empirical work attempts to study TELs across states, and as such, focuses on statewide definitions of TELs.  Obviously this covers up local variation, but perhaps more interestingly, this choice in geographic scope encourages certain questions.  What is the impact of TELs on revenue volatility ([St. Clair 2012](http://papers.ssrn.com/sol3/papers.cfm?abstract_id=2151690))?  Do TELs constrain property taxes ([Dye & McGuire 1997](http://www.sciencedirect.com/science/article/pii/S0047272797000479))?  Do TELs constrain growth in employment and wages ([Poterba & Rueben 1995](http://www.jstor.org/discover/10.2307/2117953?uid=3739256&uid=2&uid=4&sid=21102688756161))? Exploration of such questions can rely on binary indicators of TEL presence, or indices that incorporate information about legal basis, ease of circumvention, and scope of application among other characteristics.

Comparatively little research examines the within state dynamic.  Once one asks the question about local differences, modeling the parameter of interest is a different game altogether.  In this case, the researcher must tease out variation among fiscal circumstances among jurisdictions for whom the statewide legislation is a common denominator.  Once this issue comes to fore, one is immediately confronted with the following question: if the statewide legislation does not vary cross-sectionally, do we expect uniform impact across all local jurisdictions at time $t$?  The unlikely nature of this situation is the motivating factor for this study.  Fiscal behavior is a function of economic circumstance, and given the variation in economic bases across a given state, one would expect that TELs have different impacts in different jurisdiction. In identifying patterns in the variation of local revenues and expenditures, [Mullins (2004)](http://onlinelibrary.wiley.com/doi/10.1111/j.0275-1100.2004.00350.x/abstract?deniedAccessCustomisedMessage=&userIsAuthenticated=false) stands out as an important example of why higher resolution is needed.

A second, and related, question is whether or not we expect a temporal element in the impact of TELs.  Is it more constraining as a jurisdiction spends more time operating within the TEL environment?  Furthermore, given the fact that the TEL environment in Colorado is composed of overlapping policies, do we expect interaction effects to play a significant role?  To shed light on such questions, this study employs spatial techniques to examine panel data of county-level fiscal and economic data in Colorado over the 1987-2009 time period.

## Strategy

This analysis seeks to identify spatial clustering of fiscal behavior.  To do so we will examine the fiscal clustering over the given time period, and seek to identify whether or not it has changed as a consequence of the four primary elements of Colorado TEL structure:

1. General Statewide Limit on Property Tax Revenue (SLPTR):  Local property tax revenues may not increase by more than 5.5% in a given year (the limit was 7% before 1988)
2. Gallagher Amendment (GA): Assessment rates for residential property are set statewide, calculated to keep historical ratios of residential and non-residential property tax revenue consistent (with some minor space for modification)
3. Taxpayer Bill of Rights (TABOR): Limits annual growth in revenues to inflation plus a measure of growth (new construction for local governments and enrollment for school districts)
4. Amendment 23 (A23): Mandates minimum increases in school funding and removes a portion of revenue from the TABOR base

There are three main stages to this inquiry: verification of variation, non-spatial modeling, spatial modeling.

#### Verification of Variation

The hypothesis rests on the idea that not only does spatial clustering of fiscal behavior occur, the nature of the clustering must vary over time.  To capture this we will take snapshots of fiscal and economic hot spots every five years, starting with 1987.  These hotspots will be identified using Local Indicators of Spatial Autocorrelation (LISAs) developed initially in Getis & Ord (1992) and Anselin (1995).  These are the $G_i^*$ and Moran's $I_i$ statistics, respectively.

#### Non-Spatial Modeling

The non-spatial modeling will employ pooled Ordinary Least Squares (OLS), pursued with the knowledge that ignorance of dependency structures across time and space will disrupt the capacity to identify a clear statistical signal.  More than anything, this component of the analysis serves to establish a baseline, against which spatial analyses may be compared.

#### Spatial Modeling

Once variation across time and space has been established, a series of spatial models of increasing complexity will be used to measure the impact of TEL intensity on the revenue capacity of a given jurisdiction.  The goal is to demonstrate, over the course of examining all the evidence, that TELs create wedges between jurisdictions that feature differing economic starting positions.

## Conceptual Infrastructure

The two conceptual drivers of this analysis will be measures of TEL intensity, and fiscal convergence.  These two concepts are discussed separately for 1) their primary importance, and 2) to provide modularity to the analysis.  With respect to the latter, the choice of indicators used for intensity and convergence is not final.  Any choice is likely to exhibit some unfavorable characteristics.  As this research program continues, it is useful to allow for new indicators that may complement or improve upon those that have come before.

#### TEL Intensity

An ideal intensity score would capture three characteristics of any given county at time $t$:

1. The accumulation of statewide legislation over time;
2. The proliferation of locally derived exemptions from TABOR and SLPTR; and,
3. The local economic dynamics that may trigger a breach in the ceilings imposed by the aforementioned legislation.

The initial indicator to be used is an ordinal score that regrettably captures only the first two.  Effectively, the score adds a point for each additional constraint.  For example, all counties in 1987 will carry a score of two given the general application of both SLPTR and GA.  These scores are subsequently modified by De-Brucing, which 47 of 64 counties chose to do in the 1993-2009 time period.  The exemptions from these De-Brucing efforts are for TABOR and sometimes the SLPTR.  If only TABOR is waived, one point will be removed.  Two will be removed if both are waived.

#### Fiscal Convergence

In a broad sense, we can think of fiscal capacity as a measure of the resources available to a given jurisdiction.  If two jurisdictions capture the same amount of revenue per constituent, it stands to reason that they will make expenditure decisions that yield similar levels of utility.  From this perspective, revenues per capita emerges as a reasonable indicator to serve as the basis for a determination of fiscal convergence.  Indeed, *this will be the starting point in this research program.*

That being said, fiscal decisions are not made in a vacuum - initial conditions matter.  It is therefore desirable to capture a more robust characterization of the fiscal behavior of a given county.  A property base composed of primarily residential housing can and should be economically characterized in a different fashion than a base characterized by heavy industry.  The problem that the researcher must confront is how to deal with the increased dimensionality of multivariate descriptions of a base.  Comparisons of economic bases broken down by NAICS codes are far less straightforward than the differences in point values.

One way fairly robust way to reduce the number of variables one deals with is to utilize [principal component analysis](http://www.youtube.com/watch?v=UUxIXU_Ob6E&list=WLCAC16A321C6C5555), which in effect attempts to capture information about the latent process driving the variation in the directly observed variables.  

Another approach is to use a similarity score, in particular a metric that captures information about the size of given industries and their relative importance within a given jurisdiction.  The author proposes a measure comprised of two distance metrics: Kendall's $\tau$ and Euclidean Distance.  The former captures the distance between two 1-d ranking arrays, and the latter captures the distance between two 1-d value arrays.  The interpretation of Kendall's $\tau$ is as follows:

+**1** indicates the same rank order;
+**-1** indicates the reverse order; and,
+**0** indicates no relation.

Relationships are thus indicated by larger absolute magnitude, in constrast to Euclidean Distance, which decreases with similarity.  Consequently, the following statistic grows as the similarity between two compositional arrays grows:

$$\frac{\tau}{ED}$$

#### Creation of Spatial Weight Matrices

To evaluate clustering, LISAs require a definition of the linkages of importance between jurisdictions.  Said differently, they require an explicitly defined neighborhood.  Spatial analysis is generally sensitive to choice in weight matrices (cite GA), so we will test multiple neighborhood criteria and look for corroboration among the tests.

We will define neighborhoods by the following metrics:

1. Rook Contiguity
2. Queen Contiguity
3. Distance Band - Binary
4. Distance Band - Continuous (Inverse Distance Decay)
5. Kernel (Gaussian Decay)

Once defined, these weight matrices will be of general use throughout the rest of the analysis.  The Rook and Queen weights are straightforward to generate.

```python
#Generate rook weights
w_rook=pysal.rook_from_shapefile(workdir+'co_county.shp')

#Generate queen weights
w_queen=pysal.queen_from_shapefile(workdir+'co_county.shp')
```

The Distance Band weight generation is a multi-step process. This weight matrix will be defined so as to ensure that each county has at least one neighbor.  Consequently, the minimum distance at which this occurs must be extracted from the data.

Once the threshold is identified, two weight matrices are generated.  The first is binary (like Rook and Queen contiguity).  All neighbors in the neighborhood have equivalent weights.  The second is continuous, and will decay in inverse proportion to the distance.

```python
#Calculate minimum distance at which each county has at least one neighbor
thresh=pysal.min_threshold_dist_from_shapefile(workdir+'co_county.shp')

#Calculate binary distance band weights
w_db_b=pysal.threshold_binaryW_from_shapefile(workdir+'co_county.shp',thresh)

#Calculate continous distance band weights
w_db_c=pysal.threshold_continuousW_from_shapefile(workdir+'co_county.shp',thresh)
```

Kernel weights are much like continuous Distance Band weight matrices, except they feature more complicated decay functions.  In this instance, the decay function is Gaussian.

```python
#Generate kernel based weights
w_kern=pysal.kernelW_from_shapefile(workdir+'co_county.shp',function='gaussian')
```

It is useful to get a sense of the variance in neighborhood size.  The following plots provide a view of cardinality (or neighborhood size) distribution for each weighting scheme.  The *histogram* method returns a list of tuples identifying a given neighborhood size (first position) and the number of jurisdictions, $i$, that have neighborhoods of said size (second position).

```python
#Define function for generating cardinality series
def card(w):
    data=[y for x,y in w.histogram]
    ind=[x for x,y in w.histogram]
    w_series=Series(data,index=ind)
    return w_series

#Set plot plot size
plt.rcParams['figure.figsize']=15,10

#Construct list of weight matrices
weights=[w_rook,w_queen,w_db_b,w_kern]

#Construct list of titles
titles=['w_rook','w_queen','w_db_b','w_kern']

#Plot neighborhood size by county
fig,axes=plt.subplots(2,2)
plt.suptitle('Distribution of Neighborhood Sizes by Weight Scheme',fontsize=20)
i=0
for w in weights:
    card(w).plot(kind='bar',title=titles[i],ax=axes[i//2,i%2])
    i+=1
```

#### Evaluating Local Moran's I

Moran's I measures the global spatial autocorrelation in an attribute $y$ over a given neighborhood defined by $w_{ij}$.

$$I=\frac{n}{S_0}\frac{(\sum_i\sum_jz_iw_{i,j}z_j)}{(\sum_iz_i^2)}=\frac{n}{(\sum_i\sum_jw_{i,j})}\frac{\sum_i\sum_j(y_i-\bar y)(y_j-\bar y)}{(y_i-\bar y)^2}$$  

The local version is implemented as follows:

$$I_i=\frac{\sum_j z_i w_{i,j} z_j}{\sum_i z_i^2}=\frac{\sum_j (y_i-\bar y)w_{i,j}(y_j-\bar y)}{\sum_i(y_i-\bar y)^2}$$

which is then calculated for each of $i$ observations.

Why do we care about local Moran's I?  It allows us to get a kind of topological view of autocorrelation.  Just as providing a single moment of a data set masks distributional content, providing a global parameter masks regional dynamics.  In this way, we are able to identify clusters of counties with similar per capita revenue values.

Thus, the statistic is calculated for each year-specific subset of the fiscal data, and each of the considered weight schemes.

```python
wname=['w_rook_i','w_queen_i','w_db_b_i','w_db_c_i','w_kern_i']
for key in co_sp_dict.keys():
    i=0
    for w in [w_rook,w_queen,w_db_b,w_db_c,w_kern]:
        lm=pysal.Moran_Local(co_sp_dict[key]['pcrev'].values,w,permutations=9999)
        co_sp_dict[key][wname[i]]=lm.z_sim
        i+=1    
```

#### Getis & Ord's $G_i^*$

As noted by Anselin (1995), the interpretation of Moran's $I_i$ is insufficient for a comprehensive knowledge of the clustering behavior of spatially related activities.

$G_i^*$ is a natural complement in this regard, because it explains a fundamentally different kind of spatial association. $I_i$ returns high values when *similar* values are clustered and low values when *disimilar* values are clustered.  Whether or not they are high or low does not impact the magnitude of the statistic.  By contrast, $G_i^*$ returns high values when *high* values are clustered and low values when *low* values are clustered.  If one were to rely solely on the former, it would not be possible to determine the distinction between fundamentally different clusters.  For example, clustering of low revenue capacity would be indistinguishable from clustering of high revenue capacity.  If one were to rely solely on the latter, it would not be possible to detect unusual patterns of discontinuity arose (discontinuities that were beyond that which would occur with random shuffling of values).  In this instance local anomoly detection would be made more difficult, and the measure does not lend itself to the identification of spatial non-stationarity.

These measures will also be integrated into the new shapefiles soon to be created.

## Estimating the Impact of Tax and Expenditure Limitations on Fiscal Clustering

Having established that clustering occurs within a given year and the clustering dynamic changes over time, it remains to be seen whether or not TELs have a significant role to play in determining the extent of clustering activity.  The impact of TELs will be estimated in four ways.

+ **Ordinary Least Squares**  - OLS will serve as a non-spatial pooled estimate

   $$y=X\beta+\epsilon$$


+ **Spatial Error Model** - The error term in this model seeks to mitigate spatial dependency across counties

   $y=X\beta+\epsilon$  and  
   
   $\epsilon=\gamma W \epsilon + u$  
   
   becomes  $y=\gamma Wy + X\beta-\gamma WX\beta + \epsilon$


+ **Spatial Lag Model** - This model explicitly incorporates the weighted average of neighborhood values

   $y = \rho Wy + X\beta + \epsilon$
   
   becomes $y=(I - \rho W)^{-1} X\beta + (I - \rho W)^{-1} \epsilon$

In the first model, the dependent variable ($y$) will be the clustering values generated above in the first run and the second run will utilize per capita revenue (the main fiscal convergence measure).  In the last two models, $y$ represent revenue capacity.  The focus on per capita revenue facilitates estimation of the impact of the neighborhood directly, as opposed to the stringency of the TEL.  

+ **Fixed Effect Spatial Lag Model** - This model both incorporates the weighted average of neighborhood values and the value of the observation in the preceding time period

   $$y=\lambda (I_T \otimes W_N)y + (\iota_T \otimes I_N)\mu + X\beta + \epsilon$$ 


The independent variables ($X$) are as follows:

+ **TEL Stringency** - The policy variable of interest
+ **Relative County Per Capita Output** - Provides a sense of placement in the statewide distribution of economic capacity without being associated with Gross State Product
+ **Gross State Product** - Provides information about the business cycle

#### Ordinary Least Squares

The first model to be run ignores spatial association and the panel structure of the data.  It is meant to serve purely as a baseline to better understand the impact of incorporating these features into the estimation process.  

A few items of note regarding modeling with [PySAL](http://pythonhosted.org/PySAL/index.html):

+ The dependent variable, $y$, must be structured as an $n \times 1$ numpy array
+ The regressor matrix, $X$, must be structured as a $n \times j$ numpy array, where $j$ is the number of independent variables
+ As a consequence of working with numpy arrays, labeling information must be explicitly entered as parameters to the model

In keeping with the above practice, all weight matrices will be tested.

```python
def ols_by_w(w):
    """This function is utility to facilitate looping through all weighting schemes.
    Note: data are subset to provide only non-null values in relative annual payroll per capita.
    The collection of annual payroll data in one county did not begin until 2002."""
    #Set dependent variable
    y=co_sp_cbp[w][co_sp_cbp['pcap_rel'].notnull()].values
    
    #Adhere to the structure constraints imposed by PySAL
    y.shape=(len(co_sp_cbp[w][co_sp_cbp['pcap_rel'].notnull()]),1)
    
    #Create container for regressors
    X=[]
    
    #Add regressors to container in turn
    X.append(co_sp_cbp['intensity'][co_sp_cbp['pcap_rel'].notnull()].values)
    X.append(co_sp_cbp['pcap_rel'][co_sp_cbp['pcap_rel'].notnull()].values)
    X.append(co_sp_cbp['gsp'][co_sp_cbp['pcap_rel'].notnull()].values)
    
    #Convert from j x n to n x j (again to adhere to PySAL structure constraints
    X=np.array(X).T
    
    #Run model
    ols=pysal.spreg.ols.OLS(y,X,name_y=w,name_x=['TEL_intensity','pc_payroll','gsp'],name_ds='co_spatial_fiscal')
    return ols
print '***TEST RUN - Dependent = Local Morans I with Kernel Neighborhood Definition***\n'
print ols_by_w('w_kern_i').summary
```

The diagnostics indicate non-normality and heteroskedasticity, as would be expected where both spatial and temporal dependency are factors.  Nevertheless, it is interesting to note that higher output per capita is associated with lower spatial clustering for the kernel weighting scheme.

#### Spatial Error Model

The use of a Spatial Error Model (SEM) is employed here as an attempt to incorporate some of the dependency structure that has plagued the above OLS models.  In particular, the model integrates the weight matrix into the error term.  In this way, the model seeks to remove the dependency complication, but not actally explain its impact.  The latter is the goal of the Spatial Lag Model.  SEM is "a special case of a regression with a non-spherical error term, in which the off-diagonal elements of the covariance matrix express the structure of spatial dependence." (Rey & Anselin )  While OLS estimation of coefficients remains unbiased, efficiency is diminished and estimates of standard errors are biased.

From a practical standpoint, SEM is implemented via the two-stage least squares method developed by Kelejian and Prucha (1998).  Estimation of the model cannot be done in a pooled fashion because the weight matrix applies to a given set of geography, with each member of that set holding a single record.  Consequently, the analysis must be run for each year in the dataset.

A further complication arises due to the missing data in Broomfield County alluded to above.  Missing data exists until 2001.  Thus, if we restrict the dataset to only non-null values, those first 15 years carry only 63 geographic observations.  The weight matrices, on the other hand, were constructed on the entire set of 64.  Recall the structure of the model:

$y=\gamma Wy + X\beta-\gamma WX\beta + \epsilon$

In this model, $y$ is $n \times 1$ and $X$ is $n \times j$ where $j$ is the number of regressors.  If $W$ is $(n+1) \times (n+1)$, the model cannot be estimated due to the matrix dimensionality mismatch preventing the calculation of the product.  To mitigate this issue, the chosen strategy is to leverage the fact that the model seeks to estimate the average effect by inserting the year-specific mean.  In this way, we seek to reduce the impact of these observations by allowing them only to affect the observation count.  Since **pcap_rel** was calculated to be the ratio of the measured value to the year-specific mean of **pcap**, the mean is simply 1.

#### Spatial Lag Model

To reiterate, while Spatial Error Models (SEMs) seek to diminish the impacts of a "nuisance" spatial dependency structure, Spatial Lag Models (SLMs) seek to model the neighborhood impacts directly.  The time series equivalent of this task would be an autoregressive specification seeking to model the impact of preceding observations $y_{t-1},y_{t-2},...,y_{t-n}$ on the current observation $y_t$.  

$$Y_t=\alpha + \sum_{i=1}^p \rho_i Y_{t-i} + X\beta + \epsilon_t$$

In the spatial context, this "autoregressive" impact captures the weighted average of the values for all neighboring geographies, $y_j \forall j \neq i$, in the neighborhood (defined by $W$) of geography unit $y_i$.

$$y =\alpha + \rho Wy + X\beta + \epsilon$$

Much like the SEM, the SLM is a cross-sectional model that must be run for each year.

```python
import pysal.spreg.diagnostics as D

def slm_by_w_yr(ydep,lag,w,yr):
    """This function is utility to facilitate looping through all years and weighting schemes.
    Note: data are subset to provide only non-null values in relative annual payroll per capita.
    The collection of annual payroll data in one county did not begin until 2002."""
    #Set dependent variable
    y=co_sp_cbp[ydep][co_sp_cbp['AUDIT_YEAR']==yr].values
    
    #Adhere to the structure constraints imposed by PySAL
    y.shape=(len(co_sp_cbp[ydep][co_sp_cbp['AUDIT_YEAR']==yr]),1)
    
    #Create container for regressors
    X=[]
    
    #Add regressors to container in turn
    X.append(co_sp_cbp['intensity'][co_sp_cbp['AUDIT_YEAR']==yr].values)
    X.append(co_sp_cbp['pcap_rel2'][co_sp_cbp['AUDIT_YEAR']==yr].values)
    
    #Convert from j x n to n x j (again to adhere to PySAL structure constraints
    X=np.array(X).T
    
    #Row standardize weight matrix
    w.transform='r'
    
    #Run model
    slm=pysal.spreg.twosls_sp.GM_Lag(y,X,w=w,w_lags=lag,name_y=ydep,name_x=['TEL_intensity','pc_payroll'],name_ds='co_spatial_fiscal')
    return slm
print slm_by_w_yr('pcrev',2,w_rook,2000).summary
```

While the SLM does appear to explain more of the variation relative to the SEM (and OLS estimation), it is clear the current specification has limited explanatory power.  Nevertheless, observing the impact of TEL intensity over time can yet again yield some insight.  

```python
#Create container for SLMs
slm_list=[]

#Create containers for year and weight matrix labels
yr_list=[]
wlab_list=[]
lag_list=[]

#For each weight matrix...
for w in wm_list:
    #...and each year within each weight matrix...
    for yr in range(1993,2010):
        #...and each of four lags within each year, within each weight matrix...
        for lag in range(1,5):
            #...do your best to run the SLM (subject to structural constraints on the weight matrix) and throw it in a list.
            try:
                slm_list.append(slm_by_w_yr('pcrev',lag,w,yr))
                yr_list.append(yr)
                wlab_list.append(w)
                lag_list.append(lag)
            #If it doesn't work out, let me know where the problem occurred.
            except:
                print 'THERE WAS A PROBLEM WITH THE',w,'MATRIX IN',yr,'WITH LAG',lag
```

#### Fixed Effects Spatial Lag Model

 Much like the Spatial Lag Model (SLM) above, the Fixed Effects Spatial Lag Model (FESLM) captures the neighborhood impact explicitly in the model specification.  The difference is in the ability to also capture the value from the previous time period.  Recall the stacked form of the model (Millo & Piras 2012).
 
 $$y=\lambda (I_T \otimes W_N)y + (\iota_T \otimes I_N)\mu + X\beta + \epsilon$$  
 
 Unpacking this model illuminates how it extends the earlier SLM.  
 
 + $\lambda$ = spatial autoregressive coefficient
 + $I_T$ = $T \times T$ identity matrix
 + $W_N$ = spatial weights matrix
 + $\iota_T$ = vector of ones of dimension $T$
 + $I_N$ = $N \times N$ identity matrix
 + $\mu$ = individual-specific effect
 + $\epsilon_i$ = spherical errors
 
 The first term, $\lambda (I_T \otimes W_N)y$, captures both the temporal ($\lambda I_T y$) and spatial ($\lambda W_N y$) autocorrelation factors.  The fixed effects $\mu$ are distributed across time and space by $(\iota_T \otimes I_N)$.  As a reminder, the [Kronecker product](http://en.wikipedia.org/wiki/Kronecker_product) ($\otimes$) generates a block matrix in which every element of one matrix is multiplied with every element of a second matrix.  In practice, this model drops the individual-specific effect $\mu$ (a.k.a. eliminating the time invariant effects), by subtracting the average for each cross-section over time (much like the operation calculating relative per capita revenue above).  This removes the intercept and fixed effect component all together, resulting in the following form:
 
 $$y^*=\lambda(I_T \otimes W_N)y^* + X*\beta + \epsilon^*$$
 
$\lambda$ is determined via numerical optimization, and then estimates and standard errors are obtained from first order conditions.  Unfortunately, the PySAL module does not yet support spatial analysis of panel data.  The following model is implemented with the **spml** function of the [splm](http://www.google.com/url?sa=t&rct=j&q=&esrc=s&source=web&cd=5&cad=rja&ved=0CFYQFjAE&url=http%3A%2F%2Fwww.jstatsoft.org%2Fv47%2Fi01%2Fpaper&ei=DWJLUqyyOdbe4AP41YDoDA&usg=AFQjCNHP4_c3cPXquuPDrt9EZQwZt4lygA&sig2=KRz-5u9dZeLXcg_Q_aneaQ&bvm=bv.53371865,d.dmg) package in R.
