# Exploring the County Government Fiscal Share

It is useful to understand the county share of fiscal activity, relative to the fiscal activity of all governments situated in a given county.  In a first pass at this concept, we have used [CenStat](http://censtats.census.gov/) data for revenues and expenditures on all governments in the county.  Observations only overlap (approximately) with the study period (1993-2009) in three years:  1992, 1997, 2002.  We have interpolated between observations for years up to 2002, and extrapolated with the 1992-2002 annual average change to get data up to 2009.  

The county fiscal data just comes from our old friends at the [Colorado Department of Local Affairs](http://www.colorado.gov/cs/Satellite/DOLA-Main/CBON/1251589672852).  These data we have every year.  

This is just a quick Notebook to explore the county share, and maybe even generate some useful plots.
