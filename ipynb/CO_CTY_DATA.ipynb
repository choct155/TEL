```python
#Create new year variables
county_gg['year']=county_gg.AUDIT_YEAR
co_dem['year']=co_dem['Population Year']

#Create new county name variable that has only the first part
county_gg['cty']=county_gg.NAME.apply(lambda x: re.sub(' County','',x))

#Fix misses (combined city/county jurisdictions)
county_gg.cty.ix[525:559]='Denver'
county_gg.cty.ix[2205:2239]='Broomfield'

#Create new county name variable
co_dem['cty']=co_dem[' Area']

#Set indices for joins
cgg=county_gg.set_index(['year','cty'])
dem=co_dem.set_index(['year','cty'])
asm=asmt.set_index(['year','cty'])

#Drop unused original variables
cgg.pop('AUDIT_YEAR')
cgg.pop('NAME')
dem.pop('Population Year')
dem.pop(' Area')

print cgg.head()
print dem.head()
asm.head()
```

Now we need to associate the fiscal data using the hierarchical strategies used in the past.  Basically we are just building dictionaries so that we can understand how the components roll up.

```python
#Construct hierarchical dictionaries    
c_rev_dict={'REV_TOTAL':['REV_TOTAL_TAX','REV_LICENSES','REV_INTGOVT','REV_CHARGES','REV_FINES','REV_MISC','REV_TRANSFER_IN'],
          'REV_TOTAL_TAX':['REV_PROPERTY_TAX','REV_SO_TAX','REV_SALES_USE_TAX','REV_FRANCHISE_TAX','REV_OCCUPATION_TAX','REV_OTHER_TAX','REV_UNCLASS_TAX'],
          'REV_INTGOVT':['REV_HUT','REV_CIGARETTE_TAX','REV_MOTOR_VEH_FEE','REV_CTF','REV_SOCIAL_SERVICE','ALL_OTHER_INTGOVT']}
c_exp_dict={'EXP_TOTAL':['EXP_TOTAL_OPERATING','EXP_TRANSFER_OUT','EXP_CAPITAL_OUTLAY','EXP_DEBT_SERVICE_GEN'],
          'EXP_TOTAL_OPERATING':['EXP_GEN_GOVT','EXP_JUDICIAL','EXP_TOTAL_PUBLIC_SAFETY','EXP_TOTAL_PUBLIC_WORKS','EXP_HEALTH','EXP_RECREATION','EXP_SOCIAL_SERVICE','EXP_MISC'],
          'EXP_TOTAL_PUBLIC_SAFETY':['EXP_POLICE','EXP_FIRE','EXP_OTHER_PUBLIC_SAFETY'],
          'EXP_TOTAL_PUBLIC_WORKS':['EXP_STREET','EXP_TRASH','EXP_OTHER_PUBLIC_WORKS'],
          'EXP_DEBT_SERVICE_GEN':['EXP_PRINCIPAL_GEN','EXP_INTEREST_GEN']}
c_debt_dict={'TOTAL_DEBT':['GO_DEBT_GEN','REVENUE_DEBT_GEN','OTHER_DEBT_GEN']}

#Create TOTAL_DEBT
cgg['TOTAL_DEBT']=cgg[['GO_DEBT_GEN','REVENUE_DEBT_GEN','OTHER_DEBT_GEN']].sum(axis=1)

#Generate Adams County subset
adams=cgg.xs('Adams',level='cty')

#For each fiscal mapping...
for dct in [c_rev_dict,c_exp_dict,c_debt_dict]:
    #...and for each grouping within a given mapping...
    for key in dct:
        #...check if the group total value equals the sum of its parts for all years...
        same=adams[key]==adams[dct[key]].sum(axis=1)
        for yr in same.index:
            #...if any of the years has a group that doesn't add up...
            if same.ix[yr]==False:
                #...tell me about it...
                print 'KEY (',key,'): THE FOLLOWING COMPONENTS ARE NOT FULLY SPECIFIED\n'
                print same.ix[yr]
                print adams[dct[key]][i:i+1]
            else:
                pass
```

Let's start with prepping the QCEW data.  We need to subset to Colorado and map `fips` to county names.  Then we need to unstack the industries so we can merge on `year` and `cty`.

```python
#Subset to Colorado
cew_co=cew[cew.cty.apply(lambda x: ', Colorado' in x)].rename(columns={'cty':'cty_temp',
                                                                       'estabs':'est',
                                                                       'tot_wages':'wage'})

#Create new cty variable
cew_co['cty']=cew_co.cty_temp.apply(lambda x: re.sub(' County, Colorado','',x))
# print dict(zip(cew_co.cty_temp,cew_co.cty))

#Create dictionary to map long industry names to shorter ones
ind_short=['cnstr','ed_hlth','fin','info','lsr_hosp','mfg','nat_res','other','prof','pub_ad','trd','unclass']
ind_dict=dict(zip(sorted(set(cew_co.industry)),ind_short))

#Create new industry variable
cew_co['ind']=cew_co['industry'].map(ind_dict)

#Subset, set index, and unstack industry info
cew_co2=cew_co.set_index(['year','cty','ind','fed_wage','st_wage','loc_wage'])[['est','emp','wage']].unstack('ind')

#Rename columns
cew_co2.columns=[x+'_'+y for (x,y) in cew_co2.columns]

#Reset gov contribution columns
cew_co3=cew_co2.reset_index(['fed_wage','st_wage','loc_wage'])

#Check for set overlap
print set(cew_co3.ix[2005].index)-set(co_local.index.get_level_values(level='cty'))
print set(co_local.index.get_level_values(level='cty'))-set(cew_co3.ix[2005].index)


cew_co3.head()
```

##### LATE EDITION: Total Expenditures and Revenues for All Govs by County

In an update, we need to join in summary finance data for all governments in the county.  At the current time, these data are from the Censtats database, which means they have been interpolated between observed data points.  The data stop at 2002, so all subsequent data points are grown at the 1992-2002 annual average growth level.

Note that the CenStat data are in $000s.

```python
#Read in all gov data
agd=pd.read_csv(all_gov_dir+'tot_cty_fin.csv')

#Set index
agd.set_index(['year','cty'],inplace=True)

#Join in data
for col in agd.columns:
    cty_full[col]=agd[col].ix[cty_full.index].values*1000.

cty_full[cty_full['all_gov_exp'].notnull()][list(agd.columns)+['EXP_TOTAL','REV_TOTAL']].head(20)
```

We also need to deal with Broomfield, which did not exist until 2002.  We will replace the missing values for each variable with its year-specific mean value.

```python
#Define variables to be imputed
imp_vars=[var for var in cty_full if var != 'stcou']

#For each year up to 2002...
for yr in range(1975,2002):
    #...and each variable...
    for var in imp_vars:
        #...define mask...
        mask=(cty_full.index.get_level_values('cty')=='Broomfield') & \
             (cty_full.index.get_level_values('year')==yr)
        #...capture year-specific mean for the variable...
        yr_mean=cty_full.ix[yr][var].mean()
        #...and replace the Broomfield value with the year-specific mean
        cty_full.loc[mask,var]=yr_mean

print cty_full.ix[1975]['REV_TOTAL'].mean()        
cty_full.xs('Broomfield',level='cty').head(20)
```

Now we can calculate the county proportions of total fiscal activity from all governments in the county.

```python
#Calculate county proportions of fiscal activity
cty_full['cty_rev_prop']=cty_full['REV_TOTAL']/cty_full['all_gov_rev']
cty_full['cty_exp_prop']=cty_full['EXP_TOTAL']/cty_full['all_gov_exp']

cty_full[cty_full['all_gov_exp'].notnull()][list(agd.columns)+['EXP_TOTAL','REV_TOTAL','cty_rev_prop','cty_exp_prop']].head(20)
```

Hmmm, Costilla is causing some problems...

```python
cty_full.xs('Costilla',level='cty')[list(agd.columns)+['EXP_TOTAL','REV_TOTAL','cty_rev_prop','cty_exp_prop']]
```

The problem is that the expenditures and revenues in the county actually feature a downward trajectory as an artifact of extrapolating from 2002, when in fact the county finances (observed each year) generally start to go back up after 2002.  We have no basis for assigning a growth rate that makes sense, so we will effectively drop these observations by again assiging the year-specific mean.

```python
#For each year after 2002...
for yr in range(2002,2010):
    #...and each variable...
    for var in list(agd.columns)+['cty_rev_prop','cty_exp_prop']:
        #...define mask...
        mask=(cty_full.index.get_level_values('cty')=='Costilla') & \
             (cty_full.index.get_level_values('year')==yr)
        #...capture year-specific mean for the variable...
        yr_mean=cty_full.ix[yr][var].mean()
        #...and replace the Broomfield value with the year-specific mean
        cty_full.loc[mask,var]=yr_mean
        
cty_full.xs('Costilla',level='cty')[list(agd.columns)+['EXP_TOTAL','REV_TOTAL','cty_rev_prop','cty_exp_prop']]
```

Finally, we have two observations with a ratio of county expenditures to all gov expenditures that exceeds one.  This is clearly an artifact of our imputation procedure, but we have little else to go on at the current time.  We will just cap these.

```python
cty_full['cty_exp_prop']=cty_full['cty_exp_prop'].apply(lambda x: min(x,1.))
cty_full[cty_full['cty_exp_prop']>1][list(agd.columns)+['EXP_TOTAL','REV_TOTAL','cty_rev_prop','cty_exp_prop']]
```

### `Inflation Adjustment`

The financial component of the above data are in nominal terms.  We need to adjust them en masse to ensure a common base year.  To perform this task, we will utilize a basic module called *`toolbox`*, which is simply a collection of functions that I find myself writing over and over again.  In particular, we will use a function `defl()`, which builds upon [pandas](https://www.google.com/search?q=pandas&oq=pandas&aqs=chrome..69i57j69i60l2j69i59l3.1664j0j4&sourceid=chrome&espv=2&es_sm=94&ie=UTF-8) to pull CPI information from the [FRED database](http://research.stlouisfed.org/fred2/).  Let's get deflators for the 1975-2009 period, using 2009 as the base year.

```python
import toolbox

print help(toolbox.defl)

#Grab deflators
defl=toolbox.defl(1975,2012,2009)

defl.head()
```

We need to identify which columns should be deflated.  These will be nearly all of the fiscal, wage, and real property variables (assessment and home value).  As a practical matter, we will build a list of these variables, and then simply name their real counterparts by prepending `r_` to the original name.  This will make for some longer variable names, but they will be unambiguously linked to their raw origins.

```python
other_fisc_list=['GO_DEBT_GEN','REVENUE_DEBT_GEN','OTHER_DEBT_GEN','ASSETS','LIABILITIES','RETAIL_SALES',\
                 'ST_SALES_TAX_PAID','TOTAL_DEBT']

print agd.columns
other_fisc_list+list(agd.columns)
```

```python
#Establish component lists
rev_list=[x for x in cty_full.columns if x.startswith('REV')] + ['ALL_OTHER_INTGOVT']
exp_list=[x for x in cty_full.columns if x.startswith('EXP')]
other_fisc_list=['GO_DEBT_GEN','REVENUE_DEBT_GEN','OTHER_DEBT_GEN','ASSETS','LIABILITIES','RETAIL_SALES',\
                 'ST_SALES_TAX_PAID','TOTAL_DEBT']+list(agd.columns)
rp_list=['total','resid','non_resid','res_val','nonres_val','mhv']
wage_list=[x for x in cty_full.columns if x.startswith('wage')]

#Concatenate lists
financial_list=rev_list+exp_list+other_fisc_list+rp_list+wage_list

#Check for misses
# print financial_list
# print sorted(set(cty_full.columns)-set(financial_list))

#Create copy
cty_real_temp=cty_full.copy(deep=True)

#Join deflators to data
cty_real_temp['defl']=defl['defl'].ix[cty_real_temp.index.get_level_values(level='year')].values

#For variable in financial_list...
for var in financial_list:
    #...create an inflation adjusted version of the variable
    cty_real_temp['r_'+var]=cty_real_temp[var].div(cty_real_temp['defl'])
    
#Identify list of inflation adjusted variables and variables that do not need adjustment
real_vars=sorted(set(cty_real_temp.columns) - set(financial_list))

#Create inflation adjusted subset
cty_real=cty_real_temp[real_vars+['cty_rev_prop','cty_exp_prop']]

#Write to disk
cty_real.to_csv(g_dir+'CO_CTY_REAL.csv')

print sorted(list(cty_real.columns))
```

## Variable Lookup

With each new input and transformation, the size of this set continues to grow.  We need some way of identifying variables for similar concepts.  For example, if we want industrial classification, we might want to represent this with all real wages by industry.  Instead of typing each and every variable out each time, it's useful to call them in groups.

Since the lists of variables are different lengths for different groups, we should store this in a dictionary and write to json.

```python
import pprint
import json

#Instantiate PrettyPrinter
pp = pprint.PrettyPrinter(indent=1)

#Generate variable mapping
nominal_vars= {'fiscal':
                   {'revenue':rev_list,
                    'expenditures':exp_list,
                    'other_fiscal':other_fisc_list,
                    'rate':['SALES_TAX_RATE']},
               'industrial':
                   {'employees':[x for x in cty_full.columns if x.startswith('emp')],
                    'establishment':[x for x in cty_full.columns if x.startswith('est')],
                    'wages':[x for x in cty_full.columns if x.startswith('wage')],
                    'gov_contribution':['loc_wage','st_wage','fed_wage']},
               'base':
                   {'people':['pop','labforce','july_pop','july_hh_pop','births','deaths','migration',\
                              'nat_pop_inc','gq_pop'],
                    'housing':['num_hh','h_units','hh_size','vac_rate','permits','vac_hu'],
                    'real_prop':[x for x in asm.columns if x != 'Unnamed: 0']+['mhv']},
               'tels':['intensity_flow','intensity_stock','prop_ratio'],
               'lisa':[x for x in cty_full.columns if x.startswith('w_')],
               'economy':['unemp_rate']
               }

#Create container for validation check of variables
var_check=[]

#For each primary key in the dictionary...
for key in nominal_vars.keys():
    #...if the value is another dictionary...
    if isinstance(nominal_vars[key],dict):
        #...go one level deeper and for each secondary key...
        for subkey in nominal_vars[key]:
            #...add the associated values to the validation container
            var_check=var_check+nominal_vars[key][subkey]
    else:
        #...otherwise add the values from the primary key
        var_check=var_check+nominal_vars[key]
   
print set(cty_full.columns) - set(var_check)
                  
pp.pprint(nominal_vars)
```

Now we can write this mapping to disk.  It can be updated as needed.

```python
#Serialize to JSON
nom_vars=json.dumps(nominal_vars)

#Open file
nv_file=open(g_dir+'nom_vars.json','w')

#Write to the file
print >> nv_file,nom_vars

#Close file
nv_file.close()
```
