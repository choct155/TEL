#### TEL Intensity

We also need to develop a method of evaluating TEL intensity on a given county.  To borrow from the *tabor_score* Notebook, measuring the impact of TELs requires four components:


1. Comparison between the growth in the total property tax base (residential and commercial) with the statewide property tax growth limitation of 5.5% annually.  We can measure the rate of growth of the assessment base.
    + Total Property Tax Base Growth = $\frac{resid_t + non\_resid_t}{resid_{t-1} + non\_resid_{t-1}} - 1$
2. Comparison between the growth in total property base and the percentage growth value attributable to inflation plus new construction. (Note that the growth factor for school districts is actually enrollment figures, so this measure would be imperfect pending greater jurisdictional resolution.)  This is quite similar to the first component, except that the limit is variable instead of fixed at 5.5%.  In practice, the composite limit (Statewide Property Tax Growth Limitation and TABOR) is governed by this relationship below. This limit is modified only when De-Brucing occurs.  De-Brucing may lead the limit to reduce to one of the two arguments in the minimum function, or to disappear entirely.  Either way, the property tax growth is calculated in the same way, and then compared with whichever version of the limit exists for a given county.

   $$PRlimit = min[.055,inflation + construction]$$
   

3. Comparison the market value growth in residential and non-residential property tax base. This is needed for both the state and county levels. The former is needed only to understand the shift in assessment rates (which are set statewide), and thus could be circumvented by just getting the assessment rates by year. The latter is critical so that we may be able to see the impact of drops in the residential assessment rate by county.  One way to get at this concept would be to compare the ratio of residential to non-residential assessment bases across years.  An increase in the ratio would indicate greater growth in residential assessment value relative to non-residential.  

$$ASMT\_RATIO = \frac{Res\_ASMT}{NonRes\_ASMT} = \frac{Res\_VALUE * Res\_RATE}{NonRes\_VALUE * NonRes\_RATE}$$

Maybe I just want the ratio because it tells me about residential concentration as a characteristic of the county.  A residential concentration increases the difficulty of matching revenue growth to the increase in the property-based activity that might benefit from public provision of infrastructure and other services.

Since equivalent increases in market value growth yield slower growth for residential assessment value

4. State aid allocations by county. These should be available from budgetary information.

To validate these as measures, it would be useful to explore the dynamics of these components over time.

#### Assessment Base Growth

The first step is calculating the base growth for residential, non-residential, and total property assessment levels.  Note that we want to avoid the percentage change calculation at the boundary between counties.  Enter the old split-apply-combine.  To keep within county integrity, we will do the following:

1. Split the data by county
2. Calculate the percentage change
3. Replace any *inf* with *NaN* (*inf* occurs when a zero value transitions to a non-zero value)
4. Concatenate the DF pieces back together into one large DF

We will need housing unit growth later in the script as well, so we might as well calculate that now.

```python
sorted(co_sfdal2.columns)
```

```python
#Subset to after 1992 (note that all change variables, including assessment, will be NaN in 1992)
co_92=co_sfdal2[co_sfdal2.index.get_level_values('AUDIT_YEAR')>1991]

#Create containers for DF components
co_DF_list=[]

#For every county...
for cty in set(co_92.reset_index()['FNAME']):
    #Create a county subset
    co_cty=co_92.ix[cty]
    #Create percentage change variables
    co_cty['d_resid']=co_cty['resid'].pct_change()
    co_cty['d_non_resid']=co_cty['non_resid'].pct_change()
    co_cty['d_total']=co_cty['total'].pct_change()
    co_cty['hu_growth']=co_cty['hu_num'].pct_change()
    co_cty['pop_growth']=co_cty['CTY_POP'].pct_change()
    #Replace any resultant 'inf' with 'NaN'
    co_cty2=co_cty.replace(inf,NaN)
    #Reset county name
    co_cty2['FNAME']=cty
    #Throw it back in the container
    co_DF_list.append(co_cty2)
    
#Concatenate pieces back together
co92_temp=pd.concat(co_DF_list).reset_index()

#Set index again
co93=co92_temp[co92_temp['AUDIT_YEAR']>1992].set_index(['FNAME','AUDIT_YEAR']).sortlevel(0)
    
print co93[['resid','d_resid','non_resid','d_non_resid','total','d_total']].head(10)
co93[['resid','d_resid','non_resid','d_non_resid','total','d_total']].describe()
```

There are two components to the limit dynamic.  Within year, revenue growth (which can be proxied by assessment growth in a single year) cannot exceed the composite limit.  But, there is also this persistent effect, that is a function of the distance between potential and realized revenue.  Assessment growth increases this revenue potential quite independently of revenue growth *once the composite limit has been breached*.  If we were to measure the revenue potential (rev_pot) as the product of the tax rate and the assessment base, we could determine whether or not the revenue potential is greater than the composite limit distance from the prior year's realized revenue.  If it is, then the limit_diff must then be calculated as the difference between the composite limit and percentage growth in revenue required to exhaust the assessment base in a given year.  If it isn't, then the composite limit is not binding.  Assessment growth is not the right comparison, because it only tells the flow component of the story.  The above method must be modified to also take account of the stock.

One problem is, if we were to index assessment potential to compare it realized revenue, we would need a start year. That start year should probably be when TABOR was implemented, so ultimately I will need more assessment data.

I can initialize the series with the median rate in the start year.  If I were to just start the clock at zero, all counties would be in the same boat from an accumulated effect standpoint.  Employing the median rate to simulate revenue potential permits variation in starting conditions. Note that the [annual reports](http://www.colorado.gov/cs/Satellite/DOLA-Main/CBON/1251591547549) have average mill rates by county, but the reports available online only go back to 2004.  

With the update, we can go back to 1993, but we still fall victim to the staggered start because the statewide growth limitation was in place prior to 1993. Consequently, we will assume uniform starting points at the inception of TABOR.  This is less than ideal, but it can be asserted that there was a categorical change at the inception of TABOR.  It masks previous deficits, but the bias is less than it might have been with a start year of 1998.

To operationalize this information, 1993 serves as $YEAR_0$.  The 1993 assessment levels are treated as indicative of allowable revenues, an assumption we hope to remedy once the relevant data become available.  In $YEAR_1$, we must calculate the distance between the growth in the assessment base and the baseline imposed by the composite limit.  At this point, assessment value serves as our proxy for revenue yield.  If the assessment value in $YEAR_1$ exceeds the sum of $YEAR_0$ assessment *and the growth permitted by the composite limit*, the TEL is actively constraining the county in question at a magnitude defined by said distance.  If the assessment value growth does not exceed that which is allowable, then there is no active constraint imposed by the TEL.

So far we have addressed only the flow portion of the equation.  The stock portion becomes material in $YEAR_2$.  At this point we again calculate the annual distance between the assessment growth and the growth allowed by the composite limit.  However, to capture the stock effect, we employ the cumulative sum of these annual distances in an additional variable.  This cumulative sum measures the extent to which fiscal capacity is hindered by the growth limitations imposed by TABOR and the Statewide Property Tax Revenue Limit. 

The implicit assumption that drives this model is that prior to TABOR, each countied levied taxes at rates that reflected the preferences of the county in question.  Once TABOR was enacted,forced reductions in levies are actually wedges imposed by state law rather than county-specific preference.  

De-Brucing enters as a pressure release, eliminating one or both growth constraints.  In the previous analysis (*CO_TEL_convergence*), TEL intensity was measured as a score, so the impact of De-Brucing was captured as a positive integer subtracted from the number of overlapping policies.  In this analysis, De-Brucing will determine whether or not a given limit is present (TABOR or Statewide Limit on Property Tax Revenue) in the calculation of the composite limit.  From a programmatic standpoint, the De-Brucing information must be entered upstream of the cumulative impact.  We will leverage the same De-Brucing dictionary used in *CO_TEL_convergence*.

We can now calculate the following values:

1. The composite constraint (*comp_constraint*), which is just min(*S_constraint,T_constraint*)
2. The annual distance (*intensity_flow*) between total assessment growth (*d_total*) and the composite constraint (*comp_constraint*)
3. The cumulative distance (*intensity_stock*) between total assessment growth (*d_total*) and the composite constraint (*comp_constraint*)

Note that each of these is a conditional assignment.  *comp_constraint* is the minimum of two pre-existing Series.  *intensity_flow* and *intensity_stock* only have non-zero values only if said values are positive.  The latter is a bit tricky because it must reset whenever the potential is non-existent, ignoring any subsequent negative values.  To spell this out more explicitly, there are three behavioral components for the stock values:

1. If stock is already zero, negative flow values are irrelevant.
2. If stock is positive, negative flow values must reduce said stock.
3. Positive flow values always increase stock.

```python
#Calculate composite constraint
co93['comp_constraint']=co93['S_constraint'].where(co93['S_constraint'] < co93['T_constraint'],co93['T_constraint'])

#Calculate annual distance
co93['intensity_flow']=Series(co93['d_total']-co93['comp_constraint']).where(co93['d_total']>co93['comp_constraint'],0)

#Initialize cumulative distance
co93['intensity_stock']=co93['intensity_flow']

##Calculate cumulative distance (split, apply, combine)
#Create simple intensity difference
co93['intensity_diff']=Series(co93['d_total']-co93['comp_constraint']).replace(NaN,0)

#Create container for processed DF subsets
cs_df_list=[]

#Create list of counties
cty_list=sorted(list(set(co93.index.get_level_values(level='FNAME'))))

#For each county...
for cty in cty_list:
    #...generate county-specific subset...
    cty_df=co93.ix[cty]
    #...for each year within county...
    for i in range(1993,2010):
        if (i==1993):
            cty_df.loc[i,'intensity_stock']=0
        else:
            if (cty_df.ix[i-1]['intensity_stock']==0) & (cty_df.ix[i]['intensity_diff']<0):
               cty_df.loc[i,'intensity_stock']=0
            elif (cty_df.ix[i-1]['intensity_stock']>0) & (cty_df.ix[i]['intensity_diff']<0): #Check this component
                cty_df.loc[i,'intensity_stock']=np.where(cty_df.ix[i-1]['intensity_stock']+cty_df.ix[i]['intensity_diff']>0,\
                                                         cty_df.ix[i-1]['intensity_stock']+cty_df.ix[i]['intensity_diff'],0)
            else:
                cty_df.loc[i,'intensity_stock']=cty_df.ix[i-1]['intensity_stock'] + cty_df.ix[i]['intensity_diff']
    #...re-integrate county name...
    cty_df['FNAME']=cty
    #...and throw processed DF subset into list
    cs_df_list.append(cty_df.reset_index())

#Concatenate sets back together
co_tel=pd.concat(cs_df_list).set_index(['FNAME','AUDIT_YEAR'])

```

## Estimating the Impact of Tax and Expenditure Limitation Constraints

This section will implement the actual models of concern.  Following the example set by *CO_TEL_convergence*, we will start by implementing the same models with different specifications.  In this case, we will be using the models described at the beginning of the **Model Specification** section above.

+ **Ordinary Least Squares**  - OLS will serve as a non-spatial pooled estimate

   $$y=X\beta+\epsilon$$


+ **Spatial Error Model** - The error term in this model seeks to mitigate spatial dependency across counties

   $y=X\beta+\epsilon$  and  
   
   $\epsilon=\gamma W \epsilon + u$  
   
   becomes  $y=\gamma Wy + X\beta-\gamma WX\beta + \epsilon$


+ **Spatial Lag Model** - This model explicitly incorporates the weighted average of neighborhood values

   $y = \rho Wy + X\beta + \epsilon$
   
   becomes $y=(I - \rho W)^{-1} X\beta + (I - \rho W)^{-1} \epsilon$
   

+ **Fixed Effect Spatial Lag Model** - This model both incorporates the weighted average of neighborhood values and the value of the observation in the preceding time period

   $$y=\lambda (I_T \otimes W_N)y + (\iota_T \otimes I_N)\mu + X\beta + \epsilon$$

### Spatial Lag Model

Spatial Lag Models (SLMs) seek to model the neighborhood impacts directly.  The time series equivalent of this task would be an autoregressive specification seeking to model the impact of preceding observations $y_{t-1},y_{t-2},...,y_{t-n}$ on the current observation $y_t$.  

$$Y_t=\alpha + \sum_{i=1}^p \rho_i Y_{t-i} + X\beta + \epsilon_t$$

In the spatial context, this "autoregressive" impact captures the weighted average of the values for all neighboring geographies, $y_j \forall j \neq i$, in the neighborhood (defined by $W$) of geography unit $y_i$.

$$y =\alpha + \rho Wy + X\beta + \epsilon$$

The SLM is a cross-sectional model that must be run for each year.  First, however, we need to generate some weight matrices.  We will use the same approach taken in *CO_TEL_convergence*.  Note that the distance based thresholds, *distance band (db)* and *kernel (kern)*, required a minimum threshold distance.  That threshold will be set to ensure all geographies have at least one neighbor.

```python
#Establish shapefile location
shp_dir='../shp/'

#Generate rook weights
w_rook=pysal.rook_from_shapefile(shp_dir+'co_county.shp')

#Generate queen weights
w_queen=pysal.queen_from_shapefile(shp_dir+'co_county.shp')

#Calculate minimum distance at which each county has at least one neighbor
thresh=pysal.min_threshold_dist_from_shapefile(shp_dir+'co_county.shp')

#Calculate binary distance band weights
w_db_b=pysal.threshold_binaryW_from_shapefile(shp_dir+'co_county.shp',thresh)

#Calculate continous distance band weights
w_db_c=pysal.threshold_continuousW_from_shapefile(shp_dir+'co_county.shp',thresh)

#Generate kernel based weights
w_kern=pysal.kernelW_from_shapefile(shp_dir+'co_county.shp',function='gaussian')

#Create list of weight matrices
w_list=[w_rook,w_queen,w_db_b,w_db_c,w_kern]
```

Now we have the components required for an SLM, so we can define a function to facilitate the execution of multiple models.  First, however, we must deal with Broomfield County.  The county did not have assessment nor demographic profile information prior to 2001.  This has led to the existence of missing data that are confounding the matrix shape requirements for this estimator.  This is a less than ideal solution, but we are replacing missing data with within county means for each specific column.

```python
for col in co_TEL95.ix['Broomfield'].columns:
    co_TEL95.ix['Broomfield'][col].replace(NaN,co_TEL95.ix['Broomfield'][col].mean(),inplace=True)
```

Now we can develop the SLM function.  Recall that its cross-sectional nature renders Gross State Product invariant.  Thus, it must be removed.

```python
def slm_yr(df,df_name,yvar,x_list,yr,w,lag,null_check):
    """This function is utility to facilitate looping through all weighting schemes.
    Note: data are subset to provide only non-null values in relative annual payroll per capita.
    The collection of annual payroll data in one county did not begin until 2002."""
    #Set dependent variable
    #print '\nSETTING DEPENDENT:',yvar
    y=df[df[null_check].notnull()].xs(yr,level='AUDIT_YEAR')[yvar].values
    
    #Adhere to the structure constraints imposed by PySAL
    y.shape=(len(y),1)
    
    #Create container for regressors
    X=[]
    
    #Add regressors to container in turn
    #print '\nBUILDING REGRESSOR MATRIX'
    for reg in x_list:
#         print '\t',reg
        X.append(df[df[null_check].notnull()].xs(yr,level='AUDIT_YEAR')[reg].values)
    
    #Check for null values
    #print DataFrame(X)
    #null_series=DataFrame(X).groupby(DataFrame(X).index).apply(count_missing)
    #print null_series
    #print null_series[null_series>0]
        
    #Convert from j x n to n x j (again to adhere to PySAL structure constraints
    #print '\nTRANSPOSING REGRESSOR MATRIX'
    X=np.array(X).T
    
    #Row standardize weight matrix
    w.transform='r'
    
    #Run model
    slm=pysal.spreg.twosls_sp.GM_Lag(y,X,w=w,w_lags=lag,name_y=yvar,name_x=x_list,name_ds=df_name)
    return slm

#Define subset of cross-sectionally varying columns
cross_cty_econ=sorted(list(set(X_econ) - set(['gsp','st_unempr'])))
cross_cty_rev=sorted(list(set(X_rev) - set(['gsp','st_unempr'])))

slm_econ95=slm_yr(co_TEL95,'co_TEL',y_econ,cross_cty_econ[::-1],1995,w_list[0],2,null_var)

print slm_econ95.summary
```

With the specification modified to accommodate cross-sectional invariance in statewide indicators, we can attempt to extract the summary information into a DataFrame (much like the OLS summary process above).

```python
def slm_df(model,model_name,yr):
    print '***RUNNING',model_name,'***\n'
    
    #Create containers for column names, estimates and p-values
    col_list=[]
    est_list=[]
    pval_list=[]
    
    #Populate estimate and p-value containers
    for col in model.name_x:
        col_list.append(col)
    for est in model.betas.tolist():
        est_list.append(est[0])
    for zstat_tuple in model.z_stat:
        pval_list.append(zstat_tuple[1])
        
    #Generate model DF (note that while W does not show up in the column names, it is in the estimates/pvals)
    slm_df=DataFrame({'var':col_list[1:],
                      'est':est_list[1:-1],
                      'pval':pval_list[1:-1]})
    
    #Include R^2
    slm_df['pr2']=model.pr2
    
    #Include model name
    slm_df['model']=model_name
    
    #Include model year
    slm_df['year']=yr
    
    return slm_df

slm_df(slm_econ95,'slm_econ95',1995)
```

Now we can execute this model for each year, within each model specification.

```python
slm_list=[]
for year in range(1995,2010):
    slm_econ=slm_yr(co_TEL95,'co_TEL',y_econ,cross_cty_econ[::-1],year,w_list[0],2,null_var)
    slm_rev=slm_yr(co_TEL95,'co_TEL',y_rev,cross_cty_rev[::-1],year,w_list[0],2,null_var)
    slm_list.append(slm_df(slm_econ,'slm_econ',year))
    slm_list.append(slm_df(slm_rev,'slm_rev',year))
    
print slm_list[5]
```

### Panel Analysis

To round out the analysis, we can run fixed effect models seeking to tease out the impact of cumulative TEL intensity while avoiding county-specific unobservables.  Furthermore, this should help to isolate the impact on changes within county.  First we need to get the data in a panel structure.

```python
#Create container for county specific subsets
cty_list=[]

#For each county...
for cty in co_TEL95[X_rev].index.get_level_values(level='FNAME'):
    #...throw the county-specific subset in the list
    cty_list.append(co_TEL[X_rev+['pcrev']].ix[cty])
    
#Create a dict of county subsets
TEL_panel_dict=dict(zip(co_TEL95[X_rev].index.get_level_values(level='FNAME'),cty_list))

#Create panel set
TEL_panel=pd.Panel(TEL_panel_dict)

TEL_panel['Adams']
```

Now we can define our model inputs and run our fixed effect model.

```python
#Define dependents
ypan_rev=TEL_panel.ix[:,:,'pcrev']
ypan_econ=TEL_panel.ix[:,:,'pcap']

#Define independents
Xpan_rev=TEL_panel.ix[:,:,X_rev].swapaxes('items','minor')
Xpan_econ=TEL_panel.ix[:,:,X_econ].swapaxes('items','minor')

#Run models
fe_rev=pd.ols(y=ypan_rev,x=Xpan_rev,entity_effects=True)
fe_econ=pd.ols(y=ypan_econ,x=Xpan_econ,entity_effects=True)

#Capture results
fe_rev_res=fe_rev.summary_as_matrix.T
fe_econ_res=fe_econ.summary_as_matrix.T

#Capture mode fit
fe_rev_res['rsq']=fe_rev.r2
fe_econ_res['rsq']=fe_econ.r2

print fe_rev_res.head(8).to_latex()
```

Let's write these guys to disk.

```python
fe_rev_res.to_csv(wd2+'fe_rev_results.csv')
fe_econ_res.to_csv(wd2+'fe_econ_results.csv')
```

```python
#Grab residuals
rev_resid=DataFrame(fe_rev.resid.stack()).rename(columns={0:'resid'})
econ_resid=DataFrame(fe_econ.resid.stack()).rename(columns={0:'resid'})

#Grab observations
rev_obs=DataFrame(ypan_rev.stack()).rename(columns={0:'obs'})
econ_obs=DataFrame(ypan_econ.stack()).rename(columns={0:'obs'})

#Sync index level names
for df in [rev_obs,rev_resid,econ_obs,econ_resid]:
    df.index.names=['AUDIT_YEAR','CTY']

#Join residuals and observations
rev_resVobs_temp=rev_obs.join(rev_resid)
econ_resVobs_temp=econ_obs.join(econ_resid)

#Discard missing values
rev_resVobs=rev_resVobs_temp[pd.notnull(rev_resVobs_temp).all(axis=1)]
econ_resVobs=econ_resVobs_temp[pd.notnull(econ_resVobs_temp).all(axis=1)]

#Generate scatters
sb.lmplot('obs','resid',rev_resVobs)
sb.lmplot('obs','resid',econ_resVobs)
```
