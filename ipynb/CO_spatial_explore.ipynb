## Acquiring Census Data

Moving on, let's couple some data with the fortified shapefile.  To grab the data, we will be using the Census API.  Here is my key:  b62798704de5e04d638987171ec291a6889c16bf

We are going to utilize a quick wrapper to pull this data more efficiently.  

**Note**:  To use the *pandas.read_json()* method below, you must be using pandas-0.12+

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import urllib2

def census(year,product,indicator,geo,location,prnt=False):
    #Establish request string
    req='http://api.census.gov/data/'+year+'/'+product+'?key=b62798704de5e04d638987171ec291a6889c16bf&get='+indicator+',NAME&for='+geo+'&in='+location
    
    #Print option
    if prnt:
        print req
    else:
        pass
    #Open connection to Census API
    request = urllib2.urlopen(req)
    
    #Read data as json and use convert to dataframe
    req = pd.read_json(request.read())
    
    #First row of 'data' values are really column headings, so subset to real data
    req_df=DataFrame(req[1:])
    
    #Use that first row to set column labels
    req_df.columns=list(req[:1].values)
    return req_df

HTML(census('2010','sf1','P0010001','tract:*','state:08+county:001')[:20].to_html())
```

So, we have discovered something interesting in experimenting with the requests above.  I initially tried to pull all of the tract data in Colorado, which seemed reasonable.  When it didn't work, I beat my head on the desk for a while trying to understand what was wrong with my syntax.  As pointed out by a helpful [post](http://apiforum.ideascale.com/a/dtd/wildcarding-geo-values/168057-18378) on the [Census Developer Forum](http://apiforum.ideascale.com/), when you run a portal to such large data sources, you have to be careful about worst case loads.  For this reason, there are request configurations that are restricted.  The worst case scenario is a request like the following:

    [...]get=P0010001&for=block:*&in=state:*+county:*+tract:*

Pulling all blocks in all states, counties, and tracts is a request that is 11 million records deep.  Imagine many people doing this within a given time window, and it's a recipe for disaster.  For that reason, smaller requests are enforced via allowable parameter combinations.  I imagine there is also a limit on the number of requests that can be made per day, because one could just wrap these requests in loop structures (like we are going to do). Ultimately, this may end up limiting the usefulness of the API for comprehensive analysis, but we might as well find out.

```python
#Grab data
co_cty_pop=census('2010','sf1','P0010001','county:*','state:08')

#Print summary
print co_cty_pop

#Grab list of county IDs
co_cty_id=list(co_cty_pop['county'].values)

#Grab list of county names and create a dictionary for mapping a new variable
co_cty_name=list(co_cty_pop['NAME'].values)
cty_name_dict=dict(zip(co_cty_id,co_cty_name))

#Look at first 10 records for confirmation
HTML(co_cty_pop[:10].to_html())
```

With all county numerical identifiers in place, we can grab tracts by county and throw them into one large data set.

```python
def subcty(cty_list,year,product,indicator,geo,location,cty_name):
    #Generate container for county level data sets
    cty_data=[]
    
    #For each county ...
    for cty in cty_list:
        #Create a new, composite location parameter
        loc_cty=location+'+county:'+cty
        
        #Use inputs from warapper function and substitute location
        data=census(year,product,indicator,geo,loc_cty)
        
        #Throw county level result in container
        cty_data.append(data)
    
    #Concatenate into one set after all county data is read in
    all_cty=pd.concat(cty_data)
    
    #Include count names
    all_cty['cty_name']=all_cty['county'].map(cty_name)
    
    return all_cty

co_tract_pop=subcty(co_cty_id,'2010','sf1','P0010001','tract:*','state:08',cty_name_dict)
print co_tract_pop
HTML(co_tract_pop[:10].to_html())
```

We will be opening a socket to access the document in place, and then parsing it to pull information out by the appropriate tags.

```python
from lxml import etree

#Open socket
request=urllib2.urlopen('http://www.census.gov/developers/data/sf1.xml')

#Read in data
census_var=request.read()

#Use XML method to parse file
census_2010var=etree.XML(census_var)

#View concepts available
for element in census_2010var.iter("concept"):
    print("%s - %s" % (element.tag, element.get('name')))
```

For our purposes, we care little for race.  We need to know population (P1), while information housing tenure (H4) and housing size (H16) would be nice.  For each of these elements, we need to pull out the relevant codes.  Diving a bit deeper we are going to cycle through each variable *within* each concept.  If the concept tag includes the indicators in var_list, we want to see all of the variables under that concept.  Further, we want to grab the codes and descriptions of said variables.

```python
var_list=['P1.','H4.','H16.']

#View variables available
codes=[]
desc=[]
for element in census_2010var.iter("concept"):
    for var in var_list:
        if var in element.get('name'):
            for i in range(len(element)):
                print element[i].xpath("string()"),'|',element[i].get('name')
                codes.append(element[i].get('name'))
                desc.append(element[i].xpath("string()"))

#Create indicator input
code_in=(",".join(codes))

#Create dictionary
desc_short=['tot_pop','occ_hu','own_mort','own_clear','gen_rent_occ','occ_hu_size','own_occ','own_hh1','own_hh2','own_hh3','own_hh4',\
            'own_hh5','own_hh6','own_hh7','rent_occ','ren_hh1','ren_hh2','ren_hh3','ren_hh4','ren_hh5','ren_hh6','ren_hh7']
cens_info=dict(zip(codes,desc_short))
desc_dict=dict(zip(desc,desc_short))

#View reference
for item in cens_info.items():
    print item
```

With indicators in hand, let's rerun the tract request.

```python
#Pull data via API
co_tract_sf1=subcty(co_cty_id,'2010','sf1',str(code_in),'tract:*','state:08',cty_name_dict)

#Rename some of the columns
co_tract_sf1.rename(columns=cens_info,inplace=True)

print co_tract_sf1
```

Ultimately, since the spatial objects we are using have their origins in R, it's probably a safer bet to perform this merge in R.

```python
import pandas.rpy.common as com

#Convert pandas DataFrames to R data.frames
co_tract_sf1_r=com.convert_to_r_dataframe(co_tract_sf1.reset_index())

#Push new data.frame into the R namespace
%Rpush co_tract_sf1_r
```

It turns out that merging data into a Spatial object is not as straightforward as a typical merge.  Evidently this command can cause sorting issues and object coercion.  The following was taken from Roger Bivand (co-author of Applied Spatial Data Analysis with R) in a response to a question about joining data to Spatial objects on the r-geo mailing list.

> My experience is that merge() is often a false friend, because of object
> coercion and sorting, as you suggest. The rules for SpatialLinesDataFram
> and SpatialPolygonsDataFrame are simple, and SpatialPointsDataFrame (and
> by extension SpatialPixelsDataFrame) can be made as simple. The first two
> constructor functions take a match.ID= argument, default TRUE, which
> matches the geometry ID - the ID slot in the component Lines or Polygons
> objects - to the data.frame row.name. For SpatialPointsDataFrame(), the
> same argument exists, but is only used if the coordinate matrix has row
> names to be matched to the data.frame row names. If it has been used, the
> data slot row names are the geometry IDs.

Given that he was an architect of the *sp* package, I am going to go ahead and take what he says seriously.  In general, we need to extract the data slot for a merge, and then reconstruct the Spatial object.  To make this happen, it is critical that we match the geo IDs to the row.names of the appropriate rows in the data.  Perhaps we should start with verifying the IDs.

```python
%%R

#Load needed packages
require(ggplot2)
require(plyr)

IDs<-sapply(slot(c,'polygons'),
             function(x) slot(x,'ID'))
print(IDs)
```

Now let's extract the data and check the row names.  To make sure that we actually have an identifier in each row to match the geo IDs, we will take the precautionary step of adding a new variable for this purpose.  Note that we are actively using the the geo IDs in the creation of this new variable.

This appears to be redundant because the row.names for the data slot *should* already match the geo IDs.  However, better safe than sorry.

```python
%%R

#Create sorting variable by assigning the list populated by polygon IDs
c@data$sorting_id<-sapply(slot(c,'polygons'),
                          function(x) slot(x,'ID'))

#Pull data from SPDF
c_df<-as(c,'data.frame')
print(head(c_df))
```

Now we can actually perform the merge on the data component (with our sorting variable included).

```python
%%R

#Merge shape data with Census data
co_cens<-merge(x=c_df,y=co_tract_sf1_r,by.x='GEOID',by.y='GEOID',sort=FALSE)

print(head(co_cens))
cat('\n')
print('********************')
cat('\n')
print(tail(co_cens))
```

Judging by the comparison of NAMELSAD (from SPDF) and NAME.y (from Census tract data), it looks like we have a successful match!

We should now sort the merged set by sorting_id, and then check to see if row names of both the merged table and the table originally pulled from the SPDF are identical.

```python
%%R

#Reorder by sorting_id
co_cens.o<-co_cens[order(as.numeric(co_cens$sorting_id)),]

#Check equality of objects
print(paste('Equality of Row Names (ordered vs original):',identical(row.names(co_cens.o),row.names(c_df))))

#Check equality of object lengths
print(paste('Equality of Row Name Length (ordered vs original):',identical(length(row.names(co_cens.o)),length(row.names(c_df)))))
cat('\n')

#Compare ordered set to original data pulled from SPDF (head,tail,spot check in the middle)
print(head(co_cens.o[,1:6]))
cat('\n')
print(head(c_df[,1:6]))
print('********************')
cat('\n')
print(tail(co_cens.o[,1:6]))
cat('\n')
print(tail(c_df[,1:6]))
print('********************')
cat('\n')
print(co_cens.o[600:606,1:6])
cat('\n')
print(c_df[600:606,1:6])
```

It turns out that the row names *are not* identical.  Somehow, during the ordering process, sorting_id shifted from the interval [0,1248] to [1,1249].  I am not clear on why this happened, but it does appear that the appropriate order has been set when compared to the data originally pulled from the SPDF.  The spot checks at the beginning, middle, and end of the data.frames support this theory.  Consequently, we can move forward with just assigning the row names of the original table to the new ordered table, so that we may match the geo IDs in the SPDF.  (Note that I did confirm that both row name vectors were of class character, so the row names were not recast to factor during these operations).

```python
%%R

row.names(co_cens.o)<-row.names(c_df)

print(head(co_cens.o))
```

Looking good.  This data can be safely inserted into the SPDF.  However, rather than overwrite the data in the original object, per Roger Bivand's [advice](https://stat.ethz.ch/pipermail/r-sig-geo/2008-January/003064.html) I will create a new SPDF from the original and insert the data there.  Post hoc debugging is facilitated by preserving linear code flow.

Once the *data* slot has been replaced, we can transform/fortify yet again.

```python
%%R

#Create copy of SPDF
c2<-c
print(getClass(c2))

#Insert new data
c2@data<-co_cens.o

#Convert to WGS84
co_ll2<-spTransform(c2,CRS('+proj=longlat +datum=WGS84'))

#Convert to R df
co2<-fortify(co_ll2,data=c2@data)

#Check data
print(head(co2))
```

Even though it feels like we are running in circles, we need to merge the fortified information (**co2**) back in with the attribute data (**c2@data**).

```python
%%R

#Check equality of row names and sorting_id
print(identical(row.names(c2@data),c2@data$sorting_id))

#Merge spatial information to attributes
co_mapd<-merge(co2,c2@data,by.x='id',by.y='sorting_id')

#Check merge
print(head(co_mapd))
```
