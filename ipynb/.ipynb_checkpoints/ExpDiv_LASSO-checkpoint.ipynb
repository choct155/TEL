## Data Input

In `ExpDiv_DistClust_Lasso.R`, we created five imputed versions of the Chapter 5 data.  We will use one of these sets to perform our LASSO regression (they are quite similar).  If anyone has beef, we can automate the process, run it on the other four sets as well and integrate the results.

```python
#Establish working directory
workdir='/home/choct155/dissertation/TEL/ExpDiv/'

#Read in data
imp1=pd.read_csv(workdir+'dist_clust_lasso_imp1.csv',index_col=['cty','year'])

#Drop old index
imp1.pop('Unnamed: 0')

#Identify weight variables
w_vars=[var for var in imp1.columns if var.startswith('w')]

#Identify variables that should not be compared with any single measure of cluster distance
no_eval=['dist_clust','dist_clust_bf','dist_clust_pca','dc_diff','dc_diff_bf','dc_diff_pca','stcou']

#Identify variables that do not vary
no_vary=['nonres_rate','r_REV_REAL_ESTATE_TRANSFER_TAX']

#Create design subset
imp1_dm=imp1[[var for var in imp1.columns if var not in w_vars+no_eval+no_vary]]

imp1_dm.head()
```

## LASSO Regression

We can now define a function that takes any distance measure and couples it with the design matrix to fit and estimate the model.

```python
def lcv(resp,design):
    #Define response
    y=resp.values
    #Define design matrix
    X=design.values
    #Initialize estimator
    lasso=linear_model.LassoCV(normalize=True)
    #Fit data
    lfit=lasso.fit(X,y)
    #Return model
    return lasso

#Fit model to base definition of cluster distance
dc1=lcv(imp1['dist_clust'],imp1_dm)
```

```python
dc1.score(imp1_dm,imp1['dist_clust'])
```
