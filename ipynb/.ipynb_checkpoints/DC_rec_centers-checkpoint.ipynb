# DC Recreation Centers (Fitness Facilities)

Today I learned of the fitness facilities that are run by the District.  There is a pretty low annual fee of $125, and apparently that gets you access to all facilities in the city.  I already knew about the one close to my office, but naturally I wondered about how convenient other locations would be.  Sure, I could have gone the [Google Maps](https://www.google.com/maps/preview) route, but where's the fun in that?

Let's make something that looks cooler.  Besides, I could certainly use the practice and I am not entirely sure how to feed Google Maps a vector of locations anyway.  First, let's enable R and read in the data (which I cleanly copied from the [DPR website](http://dpr.dc.gov/page/dpr-fitness-centers) and pasted into a .csv).

```python
#Enable R capability
%load_ext rmagic 
```

```python
%%R

library(ggmap)

r<-read.csv('/home/choct155/Downloads/DC_rec_centers.csv')
print(r)
```

As can be seen, we have only addresses for location.  This can easily be fixed by [Google's geocoding API](https://developers.google.com/maps/documentation/geocoding/) (which is embedded in the [ggmap](http://cran.r-project.org/web/packages/ggmap/index.html) package).

```python
%%R

#Geocode coordinates
coords<-geocode(as.character(r$Address))

#Split matrix into long/lat, and attach to the data.frame
r$lon<-coords[,1]
r$lat<-coords[,2]

#Verify success
print(r)
```

Just like that, we have enough to plot.  Note that I am centering the map on my house to identify what is convenient for me.  Screw anyone else who needs this info...

```python
%%R -w 1000 -h 1000

#Generate map tile, upon which we will plot our data
dc_map<-qmap('22 Bryant St. NE Washington DC',zoom=14,maptype = 98997,api_key='439cf1240ef4468ebb4277d0d31a6570',source = "cloudmade")

#Plot rec center locations, colored by Ward
print(dc_map + geom_point(aes(x = lon, y = lat,colour = as.character(Ward)),size=10,data = r))
```

Tidy indeed, and even colored by Ward (just because I can).  However, while this map is cool, I kind of want one with better markers.
