# Testing Spatial Dependence in Cluster Distances

This is just a quick script to execute some global evaluations of spatial autocorrelation on various calculations of distance to county type cluster.  The measures were created in service of the analysis in Chapter 5.

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import pysal as ps
from mpltools import color

def simpleaxis(ax):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()
```

```python
#Establish Google drive directory
g_dir='/home/choct155/Google Drive/Dissertation/Data/'

#Read in censtat data to capture county codes
censtat=pd.read_csv(g_dir+'censtats.csv')

#Establish working directory
workdir='/home/choct155/dissertation/TEL/ipynb/'

#Read in cluster distance data
cd_tmp=pd.read_csv(workdir+'h1_exp_model_in.csv')

#Establish shapefile location
shp_dir='/home/choct155/Google Drive/Dissertation/Data/spatial/US/'

#Generate queen weights
w_queen=ps.queen_from_shapefile(shp_dir+'co_county.shp',idVariable='GEOID')

#Row standardize weights
w_queen.transform='r'
```

We should quickly map in county codes to make sure our data is in an order that accords with our weights.

```python
#Map county to codes
cty_map=dict(zip(censtat['cty'],censtat['stcou'].apply(lambda x: '0'+str(int(x)))))

#Map codes into cluster distance set
cd_tmp['stcou']=cd_tmp['FNAME'].map(cty_map)

#Check that same counties are in the cd data and the weights data
print 'County Correspondence:',set(cd_tmp['stcou'])==set(w_queen._W__get_id_order())

#Set index to year and county
cd=cd_tmp.set_index(['AUDIT_YEAR','stcou']).sortlevel(0)

cd.head()
```

## Testing the 2000 Cross-Section 

We can start by sectioning off values from 2000, and using them to test spatial dependence.  Once this is done, if needed when can test all years. We can manually reset the weights order to reflect the within year order of counties.

```python
#Set order
w_queen._W__set_id_order(list(cd.ix[1993].index))

print 'Orders Match:',list(cd.ix[1993].index)==w_queen._W__get_id_order()
```

Now we are in a position to check Moran's I.

```python
print ps.esda.moran.Moran(cd.ix[2000]['dist_clust'],w_queen).I
print ps.esda.moran.Moran(cd.ix[2000]['dist_clust'],w_queen).p_norm
print ps.esda.moran.Moran(cd.ix[2000]['dist_clust'],w_queen).p_rand
```

## Testing All Measures and All Years

```python
#Establish list of distance measures
dmeas=['dist_clust','dist_clust_bf','dist_clust_pca','dc_diff','dc_diff_bf','dc_diff_pca']

#Create containers for year-specific sets
yr_dfs=[]

#For each year...
for yr in range(1993,2010):
    #...generate containers for values
    mi=[]
    pn=[]
    pr=[]
    dm=[]
    
    #For each distance measure...
    for d in dmeas:
        #...instantiate Moran's I stat object...
        mi_obj=ps.esda.moran.Moran(cd.ix[yr][d],w_queen)
        #...and capture value and p-values
        mi.append(mi_obj.I)
        pn.append(mi_obj.p_norm)
        pr.append(mi_obj.p_rand)
        dm.append(d)
    
    #Capture in DF    
    df_tmp=DataFrame({'mi':mi,
                      'pnorm':pn,
                      'prand':pr,
                      'measure':dm})
    
    #Add year back in
    df_tmp['year']=yr
    
    #Throw it in the container
    yr_dfs.append(df_tmp)

#Concatenate into a single set
moran=pd.concat(yr_dfs).set_index(['measure','year']).sortlevel(0)

moran.head()
```

To visualize this data, we will need a color scheme.  We can use the full interval, [0,1], to show this.  Just to be robust, we will check p-values with the normality assumption...

```python
#Create color mapping with mpltools
bcr_rgba = [(0.02, 0.2, 0.4, 1),    # grayish blue, opaque
            (0.02, 0.2, 0.4, 0.3),  # grayish blue, transparent
            (0.4,  0.0, 0.1, 0.3),  # dark red, transparent
            (0.4,  0.0, 0.1, 1)]    # dark red, opaque

#Define colormap
blue_clear_red = color.LinearColormap('blue_clear_red', bcr_rgba,index=[0,0.1,0.1, 1])

#Create measure to description mapping
desc=['Distance to Cluster Centroid - Full Portfolio',\
      'Distance to Cluster Centroid - Regression Filtered',\
      'Distance to Cluster Centroid - Principal Components',\
      'Change in Distance to Cluster Centroid in Next Year - Full Portfolio',\
      'Change in Distance to Cluster Centroid in Next Year - Regression Filtered',\
      'Change in Distance to Cluster Centroid in Next Year - Principal Components']
meas_desc=dict(zip(dmeas,desc))
      

#Generate plot object
fig,ax=plt.subplots(6,figsize=(14,25),sharex=True)

#For each measure...
for i,d in enumerate(dmeas):
    #...plot the Moran scores, colored by normal p-value
    plot=ax[i].scatter(moran.ix[d].index,moran.ix[d]['mi'],c=moran.ix[d]['pnorm'].values,cmap=blue_clear_red,
                       s=1500,lw=0)
    ax[i].set_title(meas_desc[d])
    simpleaxis(ax[i])
    ax[i].axhline(0,lw=2,c='k',linestyle='--')

cbar_ax = fig.add_axes([0.91, 0.135, 0.03, 0.75])
fig.colorbar(plot, cax=cbar_ax)
```
