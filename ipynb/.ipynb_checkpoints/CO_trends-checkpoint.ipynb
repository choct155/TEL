```python
#Spatial IO
import fiona

#Spatial geometric operations
from shapely.geometry import shape,Polygon,MultiPolygon
from shapely.ops import unary_union
from descartes import PolygonPatch
from rtree import index

#Visualization
from matplotlib.collections import PatchCollection
from mpltools import color
```

First we need to read in the data from a CO county shapefile.

```python
#Establish shapefile directory
shp_dir='/home/choct155/projects/gdrive_data/Google Drive/Dissertation/Data/spatial/US/'

#Open county shapefile
shp=fiona.open(shp_dir+'co_county.shp')

#Initiate iterator
cty=shp.next()

#Create a container for features and names
cty_list=[]
cty_name_list=[]

#For each county...
for c in shp:
    cty_list.append(shape(c['geometry']))               #...throw the county shape in the list...
    cty_name_list.append(c['properties']['NAME'])       #...throw the county name in the list
    
#Map names to each shape
cty_dict=dict(zip(cty_name_list,cty_list))

print 'THERE ARE ',len(cty_list),' COUNTIES IN THE DATA SET'

shp.meta
```

Let us confirm that all elements are polygons...

```python
mps=[cty for cty in cty_dict.items() if cty[1].geom_type != 'Polygon']
mps
```

It seems we do have `MultiPolygons`.  This means we need to pull the affected counties out of the dictionary, but it up into `Polygon` objects and put the dict back together with the constituent parts.

```python
#Make copies of the lists and dict
cty_list2=cty_list
cty_name_list2=cty_name_list
cty_dict2=cty_dict

#Define function to remove MP from dict
def removekey(d, key):
    r = dict(d)
    del r[key]
    return r

#Define function to bust up MP
def mpfix(mp,l,d):
    #Extract Polygon components
    parts=[poly for poly in mp[1]]
    #Create new names for Polygons
    names=[mp[0]+str(i) for i in range(len(mp[1]))]
    #Generate sub-dict with MP component Polygons
    mp_parts_dict={}
    for i,poly in enumerate(parts):
        mp_parts_dict.update({names[i]:poly})
    #Update county list
    l=[cty for cty in l if cty.geom_type == 'Polygon'] + parts
    #Update dictionary
    d2=removekey(d,mp[0])
    d2.update(mp_parts_dict)
    #Return county list and dict
    return [l,d2]
    
#Initialize modification with first MP
init_mp=mpfix(mps[0],cty_list2,cty_dict2)

# print init_mp[0]
init_mp[1]

#For each remaining MP...
for mp in mps[1:]:
    #...recursively update the list
    init_mp=mpfix(mp,init_mp[0],init_mp[1])
    
#Update lists
cty_list3=init_mp[0]
cty_name_list3=[item[0] for item in init_mp[1].items()]
cty_dict3=init_mp[1]
             
print 'THERE ARE ',len(cty_list3),' POLYGONS IN THE DATA SET'
    
cty_dict3
```

Now we can move on to associating each county with a color and map that noise.  First, let's create a `MultiPolygon` to capture the macro bounds.

```python
#Generate single MultiPolygon containing all Polygons
ctys=MultiPolygon(cty_list3)

ctys
```

Now we need to assign colors.  Recall that there are several `MultiPolygons`.  We need to assign the same color to all of components of the same county.  We can do this in two steps.

1. Assign colors to each county; and,
2. Use that mapping to assign colors to MP components.

```python
#Create subset
co_map_sub=co_map[['HC01_VC85','NAME']].reset_index().set_index('NAME')

#Create color mapper
cm=color.color_mapper((co_map['HC01_VC85'].min(),co_map['HC01_VC85'].max()),cmap='Blues',start=.2)

#Assign colors to each county
co_map_sub['color']=co_map_sub['HC01_VC85'].apply(lambda x: cm(x))

#Create mapping from base counties to colors
cty_col_base=dict(zip(co_map_sub.index,co_map_sub['color']))

#Create color map for disaggregated counties
cty_col={}

#For each disaggregated county...
for cpoly in sorted(cty_name_list3):
    #...and for each base county...
    for c in sorted(cty_col_base.items()):
        #...if the base county contains or is equal to the disaggregated county...
        if c[0] in cpoly:
            cty_col.update({cpoly:cty_col_base[c[0]]})  #...assign the base county's color to the disaggregated component

cty_col
```

```python
#Generate plot object
fig,ax=plt.subplots(figsize=(16,14))

#Set scope of the plot
minx, miny, maxx, maxy = ctys.bounds
w, h = maxx - minx, maxy - miny
ax.set_xlim(minx - 0.2 * w, maxx + 0.2 * w)
ax.set_ylim(miny - 0.2 * h, maxy + 0.2 * h)

#Set aspect ratio
ax.set_aspect(1)

#Create container for patches (neighborhoods in this context) and locations
patches = []

#Define corresponding lists for shapely objects and names
cty_objects=[x[1] for x in sorted(cty_dict3.items())]
cty_names=[x[0] for x in sorted(cty_dict3.items())]

#For each county...
for c in cty_dict3.items():
    #...assign a color based upon median income
    colour=cty_col[c[0]]
    #...and generate a patch representing the county, then throw it in the list
    patches.append(PolygonPatch(c[1],fc=colour,ec='#555555',lw=0.2,alpha=0.8,zorder=1))
    
#Add the patches to the plot object en masse
mhv_map=ax.add_collection(PatchCollection(patches, match_original=True))

#Remove ticks
ax.set_xticks([])
ax.set_yticks([])

#Set legend
plt.legend()

#Maximize plot usage by minimizing margins
plt.tight_layout()

#Turn axes off
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)


#Generate color bar
##Calculate normalized median home values
norm=mpl.colors.Normalize(vmin=co_map['HC01_VC85'].min(),vmax=co_map['HC01_VC85'].max())
##Build figure component for colorbar
fig.subplots_adjust(right=.8)
cbar_ax=fig.add_axes([.75,.325,.03,.35])
##Generate colorbar using normalized values
cbar=mpl.colorbar.ColorbarBase(cbar_ax,cmap=plt.get_cmap('Blues'),norm=norm)
cbar.set_label('Median Income ($)')

plt.savefig('../out/figures/CO_med_inc_cty_map.svg')
```
