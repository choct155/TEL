```python
%%R

library(maptools)
library(gpclib)
library(sp)
library(ggmap)
library(rgdal)
library(raster)

#Read in Ward shapefile
w<-readShapeSpatial('/home/choct155/Google Drive/Dissertation/Data/spatial/DC/WardPly.shp')
projection(w) <- '+proj=lcc +lat_1=38.3 +lat_2=39.45 +lat_0=37.66666666666666 +lon_0=-77.0 +x_0=400000.0 +y_0=0.0 +datum=NAD83 +units=m +no_defs'
print(summary(w))
```

```python
%%R

#Convert to r data.frame
ward<-fortify(w)
print(head(ward))
```

And, now we can actually map the Wards.

```python
%%R

#Create base map
dcmap<-qmap('washington dc',zoom=11,maptype='toner',source='stamen')

#Combine ward data with base map
print(dcmap + geom_polygon(aes(x=long,y=lat,group=group),
                           data=ward,colour='white',fill='red',alpha=.5,size=.2))
```
