The goal here is to get everything into NAICS 2007, so we need to use the mappings provided in the Census [correspondence](http://www.census.gov/eos/www/naics/concordances/concordances.html) information.  We need two of the available bridges:  1987 SIC to 2002 NAICS & 2002 NAICS to 2007 NAICS.  They are located here:

    /home/choct155/data/cbp/

```python
#Set directory location
cor_dir='/home/choct155/data/cbp/'

#Read in SIC1987 to NAICS2002 information
s87_n02=pd.read_csv(cor_dir+'1987_SIC_to_2002_NAICS.csv')

#Create dictionary mapping from SIC1987 to NAICS2002
s87_n02_dict=dict(zip(list(s87_n02['SIC']),list(s87_n02['2002 NAICS'])))

#Read in NAICS2002 to NAICS2007 information
n02_n07=pd.read_csv(cor_dir+'2002_to_2007_NAICS.csv')

#Create dictionary mapping from SIC1987 to NAICS2002
n02_n07_dict=dict(zip(list(n02_n07['2002 NAICS Code']),list(n02_n07['2007 NAICS Code'])))

print s87_n02[:10]
HTML(n02_n07[:10].to_html())
```

So there are few problematic elements with with these mappings.
>1. Some of the SIC codes have been split across multiple NAICS codes.  Furthermore, some of these are in different super groups.

>2. Some of the NAICS codes correspond with more than one SIC code, and elements within the same super code group correspond with more than one SIC super code.

>3. Some of the figures in the County Business Patterns data are only provided at an aggregated level.  In many cases, no direct mapping to NAICS exists because the summary group captures SIC codes that go to different summary groups in the NAICS framework.

I am thinking that for one-to-one mappings, I will have to opt (somewhat arbitrarily) for the first encountered NAICS code.  For summary groups, I will choose the NAICS super group that is most represented (the mode).  This all assumes the Census folks don't have a better allocation scheme...

```python
#Create to column sub-lists that account for the change in industrial classification
sub_cols98_=['fipstate','fipscty','naics','emp','qp1','ap','est','n1_4','n5_9','n10_19','n20_49',\
          'n50_99','n100_249','n250_499','n500_999','n1000','n1000_1','n1000_2','n1000_3','n1000_4']
sub_cols_97=['fipstate','fipscty','sic','emp','qp1','ap','est','n1_4','n5_9','n10_19','n20_49',\
          'n50_99','n100_249','n250_499','n500_999','n1000','n1000_1','n1000_2','n1000_3','n1000_4']

#Create compression function
def compress_cbp(path,cols,ind_id):
    """Reads in expanded CBP file, and returns a compressed version (to super codes)"""
    #Read in data
    cbp=pd.read_csv(path,usecols=cols)
    
    #Create year variable
    if int(path[53:54]) < 3:
        yr='20'+path[53:55]
    else:
        yr='19'+path[53:55]
        
    #Replace [/-] in SIC and NAICS codes
    cbp['ind_code_temp']=DataFrame(cbp[ind_id]).applymap(lambda x: x.replace('/','0'))
    cbp['ind_code']=DataFrame(cbp['ind_code_temp']).applymap(lambda x: x.replace('-','0'))
    
    #Create new naics columns
    if int(yr) < 1998:
        cbp['naics02']=cbp['ind_code'].map(s87_n02_dict)
        cbp['naics07']=cbp['naics02'].map(n02_n07_dict)
    elif (int(yr) >= 1998) & (int(yr) < 2008):
        cbp['naics07']=cbp['ind_code'].map(n02_n07_dict)
    else:
        cbp['naics07']=cbp['ind_code']
        
    #Drop dead weight
    useless=['ind_code_temp']
    for var in useless:
        cbp.pop(var)
    
    """
    #Create lists for dictionary mapping full naics codes to super codes only
    naics_full=[]
    naics_super=[]
    for i in range(len(cbp[ind_id])):
        naics_full.append(cbp[ind_id][i])
        naics_super.append(cbp[ind_id][i][:2])
    
    #Create dictionary
    naics_dict=dict(zip(naics_full,naics_super))
    
    #Create super code grouping variable
    cbp['snaics']=cbp[ind_id].map(naics_dict)
    
    #Groupby state, county and super code
    scbp=cbp.groupby(['fipstate','fipscty','snaics']).sum()
    
    #Create year variable
    if int(path[53:54]) < 3:
        yr='20'+path[53:55]
        scbp['year']=yr
    else:
        yr='19'+path[53:55]
        scbp['year']=yr
    
    return scbp"""
    return cbp

#Test run
cbp97=compress_cbp('/home/choct155/Google Drive/Dissertation/Data/cbp/cbp97co.txt',sub_cols_97,'sic')
HTML(cbp97[:250].to_html())
```

The compression function appears to work for the time being.  We need to create separate file lists for SIC years (before and including 1997) and NAICS years (1998 and after).

```python
#Create list of SIC years
sic_yrs=[str(x) for x in range(86,98)]

#Create containers for SIC and NAICS files
sic_files=[]
naics_files=[]

#Allocate files appropriately
for f in f_list:
    if f[53:55] in sic_yrs:
        sic_files.append(f)
    else:
        naics_files.append(f)
```

Moment of truth, let's try to roll through the NAICS years and put them in a large dataframe.

```python
naics_data=pd.concat(compress_cbp(f,sub_cols98_,'naics') for f in naics_files)
```
