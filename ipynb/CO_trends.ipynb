## Population Trends

Let's first look at aggregate growth across all counties according to the CO demographic profiles.

```python
#Calculate population across all counties for each year
CO_pop_growth=trend.groupby(level='AUDIT_YEAR').sum()['CTY_POP']

#Isolate population for each county for each year
CO_cty_pop=trend['CTY_POP'].reset_index()

#Generate plot object
fig,ax=plt.subplots(2,figsize=(16,10))

#Plot population trend
ax[0].fill_between(CO_pop_growth.index,0,CO_pop_growth,alpha=.5,lw=2)

#Plot distribution of populations over time
sb.violinplot(CO_cty_pop['CTY_POP'],groupby=CO_cty_pop['AUDIT_YEAR'],ax=ax[1])

#Fix background
ax[0].grid('off')
ax[0].patch.set_facecolor('w')
ax[1].grid('off')
ax[1].patch.set_facecolor('w')

#Set limits
ax[0].set_xlim([1975,2009])
ax[1].set_ylim([0,CO_cty_pop['CTY_POP'].max()])

#Fix labels
##Titles
ax[0].set_title('Growth in Population Across All Counties')
ax[1].set_title('Distribution Population Across Counties by Year')
##Rotate years
plt.setp(plt.xticks()[1],rotation=30)
##Format yticklabels
ax[0].set_yticklabels(['{:,}'.format(int(val)) for val in ax[0].get_yticks()])
ax[1].set_yticklabels(['{:,}'.format(int(val)) for val in ax[1].get_yticks()])

plt.savefig('CO_pop_growth.png')
print CO_pop_growth.ix[1975],CO_pop_growth.ix[2009],CO_pop_growth.ix[2009]/float(CO_pop_growth.ix[1975])
```

# Correlation Between Population and Housing

Let's get pop/hu counts in the same space as median home values...

```python
#Read in median home value information
mhv_temp=pd.read_csv(workdir+'CO_cty_mhv.csv',header=None,names=['year','cty','mhv'],index_col=['year','cty'])

#Subset to 2010
mhv10=mhv_temp.ix[2010]

#Read 2010 Gazeteer file for CO counties
co_cty_gzt=pd.read_table('http://www.census.gov/geo/maps-data/data/docs/gazetteer/counties_list_08.txt')

#Create cty variable commensurate with mhv data
co_cty_gzt['cty']=co_cty_gzt['NAME'].str.replace(' County','')

#Set cty to index
co_cty_gzt.set_index('cty',inplace=True)

#Join mhv to gazette info
cty_base_cnt=co_cty_gzt[['POP10','HU10']].join(DataFrame(mhv10))

cty_base_cnt
```

...and plot

```python
#Subset to top 20 MHV
cty_base_cnt_top20mhv=cty_base_cnt.sort(columns=['mhv'])[:20]

#Generate plot object
fig,ax=plt.subplots(2,figsize=(15,15))

#Plot correlation
phv_scatter=ax[0].scatter(cty_base_cnt['POP10'],cty_base_cnt['HU10'],c=cty_base_cnt['mhv'],s=400,alpha=.4,cmap='RdBu')
phv_scatter=ax[1].scatter(cty_base_cnt_top20mhv['POP10'],cty_base_cnt_top20mhv['HU10'],c=cty_base_cnt_top20mhv['mhv'],\
                          s=400,alpha=.4,cmap='RdBu')

#Fix background
ax[0].grid('off')
ax[0].patch.set_facecolor('w')
ax[1].grid('off')
ax[1].patch.set_facecolor('w')

#Set limits
ax[0].set_xlim([0,700000])
ax[0].set_ylim([0,300000])
ax[1].set_xlim([0,300000])
ax[1].set_ylim([0,100000])

#Set labels
ax[0].set_title('Housing Count vs Population by County (2010)')
ax[0].set_xlabel('County Population')
ax[0].set_ylabel('County Housing Unit Count')
ax[0].set_xticklabels(['{:,}'.format(int(val)) for val in ax[0].get_xticks()])
ax[0].set_yticklabels(['{:,}'.format(int(val)) for val in ax[0].get_yticks()])

ax[1].set_title('Housing Count vs Population by County (2010) - Top 20 Counties by Median Home Value')
ax[1].set_xlabel('County Population')
ax[1].set_ylabel('County Housing Unit Count')
ax[1].set_xticklabels(['{:,}'.format(int(val)) for val in ax[1].get_xticks()])
ax[1].set_yticklabels(['{:,}'.format(int(val)) for val in ax[1].get_yticks()])

#Generate color bar
fig.subplots_adjust(right=.8)
cbar_ax=fig.add_axes([.85,.15,.05,.7])
fig.colorbar(phv_scatter,label='Median Home Value ($)',cax=cbar_ax)

plt.savefig('cty_pop_vs_hu2010.png')
```

## Income Distribution

We will need to combine the basic county level Census data with the shapefile information.  We can use GeoPandas for this without huge aesthetic losses.

```python
#Read in shapefile
co_shp=gp.GeoDataFrame.from_file(workdir+'spatial/US/co_county.shp')

#Read in census data
co_econ=pd.read_csv(workdir+'ACS_12_5YR_DP03_raw.csv')

#Create GEOID commensurate with shapefile representation
co_econ['GEOID']=co_econ['GEO.id2'].apply(lambda x: '0'+str(x))

#Set indices
co_shp.set_index('GEOID',inplace=True)
co_econ.set_index('GEOID',inplace=True)

#Join sets
co_map=gp.GeoDataFrame(co_shp.join(co_econ))

co_map
```

We need to read in a mapping of the relevant variables.

```python
#Read in mapping
var_map=pd.read_csv(workdir+'CO_ACS_econ_cty_2008_2012.csv',header=None,names=['var','desc'])

#Create dict
varmap=dict(zip(var_map['desc'],var_map['var']))

#Identify income related variables
sorted([desc for desc in varmap.keys() if 'MEDIAN' in desc.upper()])
```
