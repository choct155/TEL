# Transfer Functions

I currently have an application for a transfer function, which in this case means a method of smoothing the impact of data over time.  The context is public investment at a particular location.  Since public investmen is unlikely to have the full impact in $YEAR_0$, I would like to phase in the $YEAR_0$ investment over $n$ time periods.

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

d=DataFrame(np.arange(10)).rename(columns={0:'data'})

d
```

One method that is currently sufficient for my needs is to split the phase-in equally across years.  pandas can make this happen with the [rolling sum](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.stats.moments.rolling_sum.html#pandas.stats.moments.rolling_sum) method (albeit with assistance).

```python
window=4
d['cs']=pd.stats.moments.rolling_sum(d['data'],window,min_periods=1)*(1/float(window))

d
```

The example leaves a little to be desired with the chosen data stream.  Let's try something a bit more realistic, which turns on and off.

```python
invest=[0,0,0,100,100,0,0,200,0,0,50,0,0,100]
d2=DataFrame(invest).rename(columns={0:'data'})
d2['trans']=pd.stats.moments.rolling_sum(d2['data'],window,min_periods=1)*(1/float(window))
d2
```
