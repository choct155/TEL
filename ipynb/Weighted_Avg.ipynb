```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

d=DataFrame(np.arange(10).astype(float))
d.columns=['data']
d['cat']=['a']*3+['b']*3+['c']*2+['d']*2
d['pop']=[10]*3+[50,25,25,40,60,50,50]


print d.dtypes
d
```

```python
dg=d.groupby(d['cat'])

print 'STANDARD MEAN'
print dg.mean()

def wavg(group):
    d = group['data']
    w = group['pop']
    return (d * w).sum() / w.sum()

print '\nWEIGHTED MEAN'
print dg.apply(wavg)
```
