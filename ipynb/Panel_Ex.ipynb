# Panel Analysis Example

This script serves as an example/playground for panel data operations.  Initially, we will explore how pandas structures panel data.  The brief discussion in [McKinney's book](http://www.amazon.com/books/dp/1449319793) reads in panel data from an external source (Yahoo financial data), using the following syntax:
>panel_data=pd.Panel(dict((x,web.get_data_fred(x,start_date,end_date)) for x in x_list))

In this formulation, we have pulled data from the [St. Louis Fed](http://research.stlouisfed.org/fred2/) instead of Yahoo financial.  The x represents an indicator, and x_list represents ... a list of indicators.  We will unpack this call so that we can abstract to a more general formulation for panel data.

In the first cell, we simply get a sense of how much work the FRED pull is doing in this scheme.

```python
import pandas.io.data as web
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
pd.options.display.mpl_style='default'
from IPython.display import HTML

#Print one example feed from the FRED API
ex_feed=web.get_data_fred('DCPCPI','1/1/1990','12/31/2012')

#Examine data structure
print '***TYPE AND SHAPE OF OUTPUT***'
print type(ex_feed),ex_feed.shape
ex_feed.plot(kind='line',title='Per Capita Income in the District')
ex_feed.head()
```

As can be seen, one single-indicator call to FRED returns a time series.  The wrapper automatically houses this data in a DataFrame and indexes the data by timestamp.  If we were to wrap this call in a loop over a sequence of indicators, we would simply get several conceptually identical data structures, one for each indicator.  We could place these in a list for easy access...

```python
#Identify per capita income in DC, MD, and VA
dseries=['DCPCPI','MDPCPI','VAPCPI'] 

#Capture feeds for each indicator in list
series_list=[]
for ind in dseries:
    series_list.append(web.get_data_fred(ind,'1/1/1990','12/31/2012'))
    
#Print data
for series in series_list:
    print type(series),'\n',series.head()

#Plot data
pd.concat(series_list,axis=1).plot(kind='line')

```

This is where things get interesting.  pd.Panel actually takes a [dictionary](http://interactivepython.org/courselib/static/thinkcspy/Dictionaries/dictionaries.html) argument, with each key corresponding to a different group in the panel.  The dictionary constructor used in the McKinney example is actually a little unusual.  It relies on the fact that dictionaries can be built from [tuples](http://interactivepython.org/courselib/static/thinkcspy/Lists/lists.html#tuples-and-mutability).

A brief example will show what I mean.  We will construct the following dictionary:
>ex_dict={'a':[1,2,3],'b':[1,2,3],'c':[1,2,3]}

Disregard the fact that the values for each key are the same.  With the FRED data, they will actually be different by virtue of calling different indicators. (The indicator will be used to both name the key and function as the first argument in the web.get_fred_data pull.)

We will build the dictionary in two ways: verbose and compact.

```python
"""Verbose Build"""
print '***VERBOSE CONSTRUCTION***'
#Create a list of dictionary keys
v_keys=['a','b','c']

#Creat list of value arrays
v_values=[[1,2,3]]*3
print 'KEYS:',v_keys,'\nVALUES:',v_values

#Create tuples
v_tuple0=(v_keys[0],v_values[0])
v_tuple1=(v_keys[1],v_values[1])
v_tuple2=(v_keys[2],v_values[2])

#Put tuples in list
v_tuple_list=[v_tuple0,v_tuple1,v_tuple2]
print 'TUPLE LIST:',v_tuple_list

#Generate dictionary
v_dict=dict(v_tuple_list)
print 'DICTIONARY:',v_dict

"""Compact Build"""
print '\n***COMPACT CONSTRUCTION***'
#Create a list of dictionary keys
c_keys=['a','b','c']

#Creat list of value arrays
c_values=[[1,2,3]]*3
print 'KEYS:',c_keys,'\nVALUES:',c_values

#Create dictionary
c_dict=dict((c_keys[i],c_values[i]) for i in range(len(c_keys)))
print 'DICTIONARY:',c_dict
```

This method just embeds looping capability into a single line to provide more functionality per keystroke.  Anyway, getting back to the FRED data, here is the constructed dictionary...

```python
#Construct dictionary from feeds
feed=dict((series,web.get_data_fred(series,'1/1/1990','12/31/2012')) for series in dseries)
print feed
```

Now we can place the data in a Panel data structure.  Notice that examining a group only displays information from that group. This is because even though all of the columns show the same concept, they are not called the same thing.  Thus, they are treated as different variables.  We will fix this by renaming during the construction of the Panel set below.

```python
#Create Panel data structure
pdata=pd.Panel(feed)
print pdata

#Examine Maryland data
pdata['MDPCPI']
```

Now we should start thinking about what we might actually do from an analytical standpoint.  Let's say we are designating per capita income as our dependent variable.  What about regressors?  This script is more about working with data in pandas than model specification, so we are going to just grab a few variables that already sit in FRED.  Items that could affect per capita income:

>1. Business cycle via unemployment rate proxy (*UR)

>2. Population density (*POP - this is only population, but we can identify the land area)

>3. Government employment (*GOVTN)

Perhaps we want to make the case that health insurance coverage impacts earnings by preventing disruptive losses from derailing earning trajectories of workers.  Our weird variable of interest, then, is health insurance coverage percentage (*HICCOVPCT).  Note that the data for this variable go back only to 1999.

Given that the input for Panel is just a dictionary that holds DataFrames, we can construct this dictionary any way we like.  We will use a strategy that tries to minimize the editing scope if we decide we want to change variables or states.

```python
#Construct list of indicator suffixes (will be joined programmatically with states)
ind_suffix_list=['PCPI','UR','POP','GOVTN','HICCOVPCT']

#Construct state list
state_list=['DC','MD','VA']

#Construct land area list
land_area=[68.3,12407.0,42775.0]

#Create function to generate dictionary
def fred_dict(groups,indicators,start_date,end_date,comvar):
    """Function grabs FRED data and constructs a dictionary:

            'jurisdiction':[DataFrame holding associated variables]

        ***Note that this function aggregates to yearly values via mean().***
    """
    #Create container for group-specific DFs (will be bound to group names in output dict)
    df_list=[]
    i=0
    for group in groups:
        #Create container for within group DFs (for each indicator)
        subdf_list=[]
        #Create container for full indicator names (for use in renaming)
        indic_list=[]
        for indic in indicators:
            #Construct indicator name and throw it in indic_list
            ind=group+indic
            indic_list.append(ind)
            #Call data and throw it into the within group DF list
            subdf_list.append(web.get_data_fred(ind,start_date,end_date))
        #Concatenate to one DF per group
        df=pd.concat(subdf_list,axis=1)
        #Annualize data by averaging over each year
        df2=df.groupby(df.index.year).mean()
        #Create renaming dictionary
        rename_dict=dict(zip(indic_list,comvar))
        #Rename columns
        df3=df2.rename(columns=rename_dict)
        #Add in state land area for use in constructing population density
        df3['area']=land_area[i]
        i+=1
        #Calculate population density
        df3['POP_DENS']=df3['POP']/df3['area']
        #Drop population and area columns
        df3.pop('area')
        df3.pop('POP')
        #Throw group DFs into a values list for binding with keys
        df_list.append(df3)
    #Construct output dictionary
    df_dict=(dict(zip(groups,df_list)))
    return df_dict

#Generate Panel input
panel_input=fred_dict(state_list,ind_suffix_list,'1/1/1999','12/31/2011',ind_suffix_list)
print panel_input
```

Now we can create our Panel structure, and see if we can actually analyze this piece.  The Panel class is only a data structure.  Methods for analysis sit in a different location: ols.  ols can take Panel data structures as inputs.

```python
health=pd.Panel(panel_input)
print health
```

To ease our way into things, we can perform basic Ordinary Least Squares on one group at a time. To get the appropriate slice, we index along each axis (items, major, and minor).  In our data set, these correspond to (state,year,and variables) respectively.  

```python
#Slice down to states
dc=health.ix['DC']
md=health.ix['MD']
va=health.ix['VA']

#Create column slice lists
y_col='PCPI'
x_col=['GOVTN','HICCOVPCT','POP_DENS','UR']

#Run OLS
dc_res=pd.ols(y=dc[y_col], x=dc[x_col])
md_res=pd.ols(y=md[y_col], x=md[x_col])
va_res=pd.ols(y=va[y_col], x=va[x_col])
print dc_res,md_res,va_res
```

Now, these summaries look nice, but what if we want to actually do something with some of the output?  The output objects have methods of their own that extract elements of the analysis.  One may wish, for example, to compare the impact of each element across states.  

```python
#Create DF for comparison
##Throw results in a list
res_list=[dc_res,md_res,va_res]
##Create list of state names in the same order as results
res_state=['DC','MD','VA']
##Creat container for result parameters (so we can merge)
res_out=[]
for i in range(len(res_list)):
    #Grab parameters of interest
    out=[res_list[i].beta,res_list[i].p_value]
    #Create DF from parameters, renaming the columns appropriately
    res_df=pd.concat(out,axis=1).rename(columns={0:'beta',1:'pval'})
    #Remove intercept and rename index
    res_df2=res_df[res_df.index!='intercept']
    #Add state identifier
    res_df2['STATE']=res_state[i]
    #Output transposed parameter DFs to list
    res_out.append(res_df2)
#Concatenate to one output parameter DF
out_param=pd.concat(res_out)

#Add state to index
out_param.set_index('STATE',append=True,inplace=True)

out_param
```

Looks like we have a usable DF, and we can make some comparisons.  These comparisons would be easier to see graphically, so let's plot the data.  Ideally we could have color mapped to the p_value, so we could have a clear idea of the relative significance of each estimate.  This is fairly trivial with ggplot2 (in R), but using *rmagic* is not the subject of this script.  Consequently, we will make do with the sometimes clunky matplotlib.

The darker the color (light: green, medium: blue, dark: purple), the higher the degree of "statistical significance" based upon our p-value metric.

The results above (superficially) suggest that state-specific factors are quite important in determining the impact of our covariates on per capita income.  We, thus, have cause to explore the panel nature of these data.  

Note that we will be using our Panel structure for this section instead of indexing down to a single DataFrame.  This increases the dimensionality of our inputs.  When we ran OLS above, our dependent (y_ols) was a single Series, basically a 1-D array.  Our covariates (X_ols) were housed in a DataFrame, a 2-D array.  Now that we are performing panel analysis, our dependent (y_panel) will be a DataFrame of the same length as our previous dependent (y_ols).  However, it now has three columns, one for each group.  

```python
#y=dc['DCPCPI'], x=dc[['DCGOVTN','DCHICCOVPCT','DCPOP','DCUR']]
#dc=health.ix['DC',:,:5]
#md=health.ix['MD',:,5:10]
#va=health.ix['VA',:,10:]
print '***y_ols for DC***\n',health.ix['DC',:,'PCPI']
print '***y_panel***\n',health.ix[:,:,'PCPI']
```

Our new covariates (X_panel) are now housed in a 3-D array.  You can think of it as having three DataFrames (one for each group) stacked on top of one another. 

```python
print '***X_ols for DC***\n',type(health.ix['DC',:,['GOVTN','HICCOVPCT','POP_DENS','UR']]),'\n', \
                                  health.ix['DC',:,['GOVTN','HICCOVPCT','POP_DENS','UR']]
print '***X_panel (to_frame)***\n',type(health.ix[:,:,['GOVTN','HICCOVPCT','POP_DENS','UR']].to_frame()),'\n', \
                                        health.ix[:,:,['GOVTN','HICCOVPCT','POP_DENS','UR']].to_frame()
print '***X_panel***\n',health.ix[:,:,['GOVTN','HICCOVPCT','POP_DENS','UR']]
```

*The pandas ols class will recognize the shift in inputs, and alter the way it treats the data.* Furthermore, additional options (pool, entity_effects, time_effects, etc.) for panel analysis will become available.

First, let's try to abstract away from specific states, and explore a pooled model. **Note that the minor axis in the X Panel object (x_panel) must correspond with the column names in the y DataFrame (y_panel).  If this is not the case, trying to run the panel analysis will throw a cryptic error.** The one I saw was a math domain error (essentially claiming the instance of a figure value that could not be achieved, like a log operation returning a negative number), and the stack trace pointed to a specific line number rather than associated code (which I moved around when attempting to print values).  Save yourself the hassle and just make sure your axes are where they are supposed to be (this is the reason for .swapaxes() call below).

```python
y_panel=health.ix[:,:,'PCPI']
x_panel=health.ix[:,:,['GOVTN','HICCOVPCT','POP_DENS','UR']].swapaxes('items','minor')
hpool=pd.ols(y=y_panel, x=x_panel)
print hpool
```

```python
h_efe=pd.ols(y=y_panel, x=x_panel,entity_effects=True)
print h_efe
```

As can be seen, adding in fixed effects for the states changes the game substantially.  The impact of population density now dominates the space.  In fact, the impact is so aggressive it suggests a serious specification issue.  Nevertheless, refining the specification is not our current purpose (though it is notable that health insurance coverage remains significant).

Perhaps we are concerned about temporal events skewing results in a given year (or over time since we haven't deflated income!)...

```python
h_tfe=pd.ols(y=y_panel, x=x_panel,time_effects=True)
print h_tfe
```

These figures look quite a bit more reasonable, and in large part, accord with at least the expected sign. (GOVTN appears to be a toss up through the different specfications...)
