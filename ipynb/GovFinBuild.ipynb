### 1967, 1970-2006

Note that in practice, I will be reading in a very small portion of the data to test the build.  Ultimately the restriction on the number of rows will be removed.

Each year of data is actually split into three parts, such is the extensiveness of the variable list.  These variables are linked together via a set of merge keys, the identity of which we will utilize the data to determine.

```python
#Establish directory location
dirpath='/home/choct155/Google Drive/Dissertation/Data/Census/GovFinIndUnit/_IndFin_1967-2006/'

#Read in
if67a=pd.read_table(dirpath+'IndFin67a.Txt',sep=',',nrows=100)
if67b=pd.read_table(dirpath+'IndFin67b.Txt',sep=',',nrows=100)
if67c=pd.read_table(dirpath+'IndFin67c.Txt',sep=',',nrows=100)

#Find the columns by which the horizontal components may be joined
m_key=list(set(if67a.columns) & set(if67b.columns) & set(if67c.columns))
m_key
```

The length of time it took me to do that for the first time is both amusing and sad at the same time.  We need to join on this merge key.

```python
#Set indices in preparation for join
f67a=if67a.set_index(m_key)
f67b=if67b.set_index(m_key)
f67c=if67c.set_index(m_key)

#Join across variable sections
f67=f67a.join(f67b.join(f67c))

print f67.info()
f67[f67.columns[:10]].head()
```

This all looks generally successful.  Let's read in every year.  We can just loop through each finance file in the folder within each variable section, concatenate these together, and merge on `m_key`.

```python
#Capture relevant file names
file_a=glob.glob(dirpath+'IndFin*a.Txt')
file_b=glob.glob(dirpath+'IndFin*b.Txt')
file_c=glob.glob(dirpath+'IndFin*c.Txt')

#Create container for variable section-specific DFs
df_list=[]

#For each variable section file list...
for flist in [file_a,file_b,file_c]:
    #...create a container...
    file_list=[]
    #...and for each Individual Government Finance file in each file list...
    for f in flist:
        #...read in the data and throw it in the list
        file_list.append(pd.read_table(f,sep=',',nrows=100).set_index(m_key))
    #...vertically concatenate and throw it in df_list
    df_list.append(pd.concat(file_list))
   
#Join the variable section-specific DFs together on m_key
gfs67_06=df_list[0].join(df_list[1].join(df_list[2])).sortlevel(0)

gfs67_06[gfs67_06.columns[:10]].head()
```

The first thing to be noticed as that the clear winner on a merge key is `id`.  The second thing to notice is that different fiscal activities are arranged by code.  I know what you are thinking, surely there must be some dictionary that is accessible in a manner conducive to statistical programming.  Unfortunately, for some inexplicable reason, that does not appear to be the case.  We will have to build this monstrosity from scratch using the [2006 Classification Manual](http://www2.census.gov/govs/pubs/classification/2006_classification_manual.pdf), which is in PDF.  On the plus side, once, we build this dictionary, we can save it to disk and never deal with this again.

So, how are we going to do this?  One very reliable method would be to copy/paste from the classification manual.  It would be reliable in the sense that we would know before we used the dictionary that all elements were represented.  The downside is that the tedium would be overwhelming since there are 1000 codes or at least more than I want to enter (with respect to manual entry).  

A less reliable way would be to [convert the PDF to text](http://stackoverflow.com/questions/17098675/searching-text-in-a-pdf-using-python), and parse the file.  Then we could let Python do the heavy lifting, and clean up afterwards.  It turns out that each code is presented in the same common format:

'\nCode XXX [Description]\n'

This suggests that we should be able to pick these things out, split the strings into tuples, and convert our tuple array into a dictionary.  The first part was actually done in [BASH](http://en.wikipedia.org/wiki/Bash_(Unix_shell) (but it ought to be achievable in DOS as well).

    sudo pdftotext 2006_classification_manual.pdf 2006_class
    
Our new text version of the file is `2006_class`.  Let's pull out all lines that start with 'Code'.

```python
#Set classification information location
class_dir='/home/choct155/Google Drive/Dissertation/Data/Census/GovFinIndUnit/GovFin Tech Docs/'

#Open file
clss=open(class_dir+'2006_class','r')

#Create container for lines that start with 'Code'
code_list=[]

#For each line in 2006_class...
for line in clss:
    #...if the line starts with 'Code'...
    if line.startswith('Code'):
        code_list.append(line)
    else:
        pass
```

The cleaning process here can be done a bit more compactly, but making the process explicit and iterative may make things a bit easier to absorb from a pedagogical perspective.  [Regular expressions](http://en.wikipedia.org/wiki/Regular_expression) are a whole other thing.

The first thing we need to do is eliminate duplicates, which is easy enough with a `set()` reference.

```python
#Find unique Code lines
code_set=list(set(code_list))

#Compare lengths of original list and unique list
print 'LINES BEGINNING WITH Code:',len(code_list)
print 'UNIQUE LINES BEGINNING WITH Code:',len(code_set)
```

That covers explicit duplicates, but previous perusal of the classification manual revealed that some of these lines will just be continuations of the first reference.  We should remove these.

```python
#Show the continued elements
print [x for x in code_set if 'continued' in x]

#Subset to exclude continued elements
code_set2=[x for x in code_set if 'continued' not in x]
code_set2[:10]
```

As can be seen, some references are just headers that describe the information shown on the page.  We can eliminate these by keying on 'Validity'.

```python
#Show the Validity elements
print [x for x in code_set2 if 'Validity' in x]
print [x for x in code_set2 if 'valid' in x]

#Subset to exclude Validity elements
code_set3=[x for x in code_set2 if 'Validity' not in x]
code_set4=[x for x in code_set3 if 'valid' not in x]

code_set4[:10]
```

Some of the elements include a dash ('-') followed by definitional information.  The dash, however, is represented by `\xe2\x80\x93`.  Furthermore, apostrophes are represented by `\xe2\x80\x99`.

```python
#Replace escapes
code_set5=[x.replace('\xe2\x80\x93','-') for x in code_set4]
code_set6=[x.replace('\xe2\x80\x99','\'') for x in code_set5]

code_set6[:10]
```

Now we need to get rid of some string elements.  For starters, we do not need newline escapes (`\n`) at the end of each line.  We can also drop the code references and periods after the actual code.

```python
#Remove elements
code_set7=[x.replace('\n','') for x in code_set6]
code_set8=[x.replace('Code ','') for x in code_set7]
code_set9=[x.replace('Code: ','') for x in code_set8]
code_set10=[x.replace('.','') for x in code_set9]

code_set10[:10]
```

Now we are cooking with gas.  We need to convert this guy into a dictionary, but this must be done in such a way as to preserve an empty value space for keys that do not have a corresponding description from our file read.

```python
#Create containers for keys and values
key_list=[]
val_list=[]

#For entry element in code_set10...
for entry in code_set10:
    #...if the entry can be split...
    if len(entry.split(' ',1))>1:
        #...put the first position in the key list...
        key_list.append(entry.split(' ',1)[0])
        #...and the second position in the value list...
        val_list.append(entry.split(' ',1)[1])
    #...otherwise...
    else:
        #...put the entry in the key list...
        key_list.append(entry)
        #...and the value entry is left blank
        val_list.append('')

#Generate code dictionary mapping codes to descriptions
code_dict=dict(zip(key_list,val_list))
```
