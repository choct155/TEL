# County Business Patterns

The local economic activity information I will need for describing the economic bases of considered jurisdictions is probably best provided by [County Business Patterns](http://www.census.gov/econ/cbp/historical.htm).  I figure these data are of general utility and, given the evolution of SIC and NAICS, involve non-trivial manipulation to prepare for integration with other data.  For both of these reasons, this script will be a standalone.  

The link above points to the Census site's archive, which only goes back to 1986.  This should be sufficient, but if I need more, data going back to 1967 is hosted by the [National Archives](http://research.archives.gov/description/613576).  Retrieving it, however, is a bit more involved.  

In any case, we will be using the [BeautifulSoup4](http://www.crummy.com/software/BeautifulSoup/bs4/doc/) HTML parser (and [urllib2](http://docs.python.org/2/library/urllib2.html)) to retrieve the data.  The reason we are doing this is because there is no central location that houses all the data.  They are not even in the same directory on the FTP site.  This means that we cannot iterate over the data URLs directly because they are not sufficiently consistent.  However, the URLs for the CBP historical data pages *are consistently formatted*.  All that changes is the year.  Consequently, we can iterate over each of these URLs and extract the relevant URL from a link on each page.

There are multiple steps in this process:
>1. Establish an output file destination to hold the data;

>2. Identify the base URL holding the link to the desired CBP historical data page and open socket;

>3. Parse the content located at the URL above with Beautiful Soup;

>4. Loop through all the links (tag='a') on the page;

>5. Grab the URL attached to the link of interest (e.g. 'Complete County File');

>6. Open socket to this URL, and write to disk the data associated with said URL.

Once this has been accomplished, we can wrap this guy up in a function and run it over all years (1986-2011).

Now to try and create a function...

```python
def cbp_fetch(year,outdir,target):
    """Grabs target CBP object from a given year and downloads it to outdir"""
    
    #Establish output file name
    file_out='cbp'+str(year)+'.zip'
    
    #Establish output location
    out=os.path.join(outdir,file_out)
    
    #Construct URL for year-specific historical data page
    base='http://www.census.gov/econ/cbp/download/'+str(year)[2:]+'_data/index.htm'
    
    #Parse contents of year-specific historical data page
    soup=bs(urllib2.urlopen(base))
    
    #Loop through all links
    for link in soup.find_all('a'):
        #If the link text matches the target...
        if unicode(target).split()==unicode(link.string).split():
            print link['href']
            #...extract the href attribute (associated URL) and open socket
            response=urllib2.urlopen(link['href'])
            #Write contents to disk
            open(out,'wb').write(response.read())

#cbp_fetch(1986,'/home/choct155/data/cbp','Complete County File')
```

It may be that for at least the upfront work, the subdivisions within NAICS super codes is not all that important.  We want to be able to describe a broad view of industrial composition.

We can compress by grouping on the super codes.  Let's first get the character info out of the way by dumping extraneous flags and setting the index.

```python
print cbp10.dtypes
```

```python
#Drop dead weight
for var in ['empflag','emp_nf','qp1_nf','ap_nf']:
    cbp10.pop(var)

#Set index
cbp10.set_index(['fipstate','fipscty','naics'],inplace=True)

HTML(cbp10[:100].to_html())
```

We really only care about Colorado...

```python
cbp_co10=cbp10[cbp10.index.get_level_values(level='fipstate')==8]
cbp_co10
```

About 40,000 per year is much more manageable.  How can we cluster by NAICS?

```python
#Create dictionary mapping full naics codes to super codes only
naics_full=[]
naics_super=[]
for i in range(len(cbp_co10.index.get_level_values(level='naics'))):
    naics_full.append(cbp_co10.index.get_level_values(level='naics')[i])
    naics_super.append(cbp_co10.index.get_level_values(level='naics')[i][:2])

naics_dict=dict(zip(naics_full,naics_super))

#Assign temporary variable to enable dictionary mapping to a super code variable
cbp_co10['n_tmp']=cbp_co10.index.get_level_values(level='naics')

#Fill out naics values
cbp_co10['n_code']=cbp_co10['n_tmp'].replace(['-','/'],'0')

#Create super code variable
cbp_co10['nsuper']=cbp_co10['n_tmp'].map(naics_dict)

#Jettison temporary variable
#bp_co10.pop('n_tmp')
print cbp_co10.dtypes
HTML(cbp_co10[:20].to_html())
```

Note that performing Series operations on index objects does not always behave like one might expect.  This is the reason for assigning a temporary variable and then getting rid of it.  

Now we can groupby 'nsuper'.

```python
#Groupby county and super code
scbp_co10=cbp_co10.reset_index().groupby(['fipscty','nsuper']).sum()

#Drop state indicator
scbp_co10.pop('fipstate')

#Create year variable
scbp_co10['year']=2010

#Check out summary information
print scbp_co10.info

#Visually inspect output
HTML(scbp_co10[:20].to_html())
```
