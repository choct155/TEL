## Neighborhood Definition

To start, we need some data.  I am not entirely sure where the data in the book comes from, but I think it would perhaps be more useful to work with District data anyway.  I would use Colorado, but for the initial bit here, DC contains a more manageable number of census geographies.

The first task is to read in the shapefile and convert it to an object of class **nb**, the structure used by the **spdep** package to represent neighbor relationships between n observations.

```python
%%R

library(rgdal)
library(spdep)

#Read in data
dc_shp<-readOGR('/home/choct155/Google Drive/Dissertation/Data/spatial/DC','tl_2012_11_tract')
print(class(dc_shp))
cat('\n**********\n')

#Convert to nb
dc_nb<-poly2nb(dc_shp)
print(summary(dc_nb))
```

Both the original shapefile and the **nb** object confirm the presence of 179 Census tracts in the District (taken from the [2012 TIGER repository](http://www.census.gov/geo/maps-data/data/tiger-line.html) maintained by the Census).  It also contains information about the extent of linkages between tracts, the link number distribution, and the linkage tail values.

What is it that we are actually looking at?

```python
%%R -w 700 -h 800

#Plot shapefile
plot(dc_shp,border='grey60')

#Plot nb object
plot(dc_nb,coordinates(dc_shp), pch=19,cex=1,add=TRUE)
```

When overlaying the **nb** object on top of the DC map, it becomes apparent what we are doing here - literally connecting each tract with other tracts.  By default, it displays a 'queen' relationship scheme.  This means that neighbors need only share a single point to satisfy the contiguity condition.  If we required two shared vertices, we would see a 'rook' relationship scheme.  (And the chess analogue becomes clear...)

```python
%%R -w 700 -h 800

#Generate a second nb instance for rook display
dc_nb_rook<-poly2nb(dc_shp,queen=FALSE)

#Plot shapefile
plot(dc_shp,border='grey60')

#Plot nb object
plot(dc_nb_rook,coordinates(dc_shp), pch=19,cex=1,add=TRUE)
```

It is a noticeably less busy scheme, and avoids the link crossing in the 'queen' representation.

It is also possible to use centroids to capture what are known as graph-based neighbors.  Four apparently proper representations are the following:
>1. [Delauney Triangulation Neighbors](http://en.wikipedia.org/wiki/Delaunay_triangulation)

>2. [Sphere of Influence Neighbors](http://demonstrations.wolfram.com/SphereOfInfluenceGraphs/)

>3. [Gabriel Graph Neighbors](http://en.wikipedia.org/wiki/Gabriel_graph)

>4. [Relative Graph Neighbors](http://en.wikipedia.org/wiki/Relative_neighborhood_graph)

```python
%%R -w 1400 -h 1400

library(tripack)

#Simplify insertion of relevant data
coords<-coordinates(dc_shp)
IDs<-row.names(as(dc_shp,'data.frame'))

#Grab Delauney neighbors
dc_del<-tri2nb(coords,row.names=IDs)

#Grab SOI
dc_soi<-graph2nb(soi.graph(dc_del,coords),row.names=IDs)

#Grab Gabriel
dc_gab<-graph2nb(gabrielneigh(coords),row.names=IDs)

#Grab Relative
dc_rel<-graph2nb(relativeneigh(coords),row.names=IDs)

#Plot
par(mfrow=c(2,2))
plot(dc_shp,border='grey60')
plot(dc_del,coords,add=TRUE)
title(main="Delauney Triangulation")
plot(dc_shp,border='grey60')
plot(dc_soi,coords,add=TRUE)
title(main="Sphere of Influence")
plot(dc_shp,border='grey60')
plot(dc_gab,coords,add=TRUE)
title(main="Gabriel Graph")
plot(dc_shp,border='grey60')
plot(dc_rel,coords,add=TRUE)
title(main="Relative Graph")
```

Clearly each of these drives a categorically different set of relationships.  For our purposes, sphere of influence appears to be best suited to capture the kind of proximate relationships we seek to tease out.  Both Delauney and SOI provide at least one neighbor for all geographies and both are symmetric (if *i* is a neighbor of *j*, then *j* is a neighbor of *i*).  The latter cannot be said for the other two.  Furthermore, SOI does not suffer from the extremely long linkages picked up by Delauney.  Unless there is compelling evidence to the contrary, SOI is our go to in the graph neighborhood area.

Alternatively, if we are willing to relinquish the security of at least one neighbor for all geographies, we can employ distance-base measures.  The first considered method is k-nearest-neighbors.  The method has the advantage of adapting to density differentials in the analysis area and ensuring a consistent number of neighors.  However, it can contain asymmetries.

```python
%%R -w 1400 -h 700

#Generate different neighborhoods based upon number of neighbors
dc_knear1<-knn2nb(knearneigh(coords,k=1),row.names=IDs)
dc_knear2<-knn2nb(knearneigh(coords,k=2),row.names=IDs)
dc_knear4<-knn2nb(knearneigh(coords,k=4),row.names=IDs)

#Plot
par(mfrow=c(1,3))
plot(dc_shp,border='grey60')
plot(dc_knear1,coords,add=TRUE)
title(main="One Neighbor (k=1)")
plot(dc_shp,border='grey60')
plot(dc_knear2,coords,add=TRUE)
title(main="Two Neighbors (k=2)")
plot(dc_shp,border='grey60')
plot(dc_knear4,coords,add=TRUE)
title(main="Four Neighbors (k=4)")
```

The distances between neighbors calculated by these routines just sits in a vector and is thus available for exploration...

```python
%%R

#Unpack distances
kn1_dist<-unlist(nbdists(dc_knear1,coords))

#Check out distribution
hist(kn1_dist)

#Check out summary stats
print(summary(kn1_dist))
```

We could even go so far as to grab the maximum distance between neighbors, and then explore how many connections are made when our band is increased to given multiples of that maximum (using the *dnearneigh* method).  Take my word for it, the number of neighbors grows very quickly.  In fact, a major weakness of straight up distance bands is the inability to deal well with high variance in neighbor distances.

Finally, *nblag* provides the functionality to understand higher order neighbors (or those that are removed from direct contact, but do retain influence through influence on intermediaries).  The function takes an existing neighbor list and returns a list of lists, from first to maxlag order neighbors.

## Spatial Weights

While defining the neighborhood provides the information for basic binary style influence between proximate geographies, it is unreasonable to think that all geographies in the neighborhood exert identical influence.  This is where spatial weights come into play.  The weight matrix parses out capacity for influence amongst the constituent members of the neighborhod.  Envision a matrix of neighbors of dimension (i=10,j=10)...

```python
from sympy import *
%load_ext sympy.interactive.ipythonprinting

W=MatrixSymbol('W',10,10)
print W
Matrix(W)
```

Each element *Wij* represents the strength of influence of feature *i* on feature *j*, and vice versa.

These neighbors are established via the definitions above, and if two features are not neighbors, the influence is zero.  Weights are presented by objects of type **listw**.  **nb** objects can be converted directly via the aptly named *nb2listw* method, which standardizes the weights to sum to unity for each areal unit by default (conversion style *W*).  Note that *W* style just averages the impact of all neighbors, so the weight for each neighbor is the same.

```python
%%R

#Convert to listw
dc_lw_W<-nb2listw(dc_nb)
print('***Entire listw Summary***')
print(dc_lw_W)
cat('\n')

#Examine components
print('***Conversion Style***')
print(dc_lw_W$style)
cat('\n')
print('***Neighbor Information***')
print(dc_lw_W$neighbours)
cat('\n')
print('***Weights associated with first of 179 elements***')
print(dc_lw_W$weights[1])
```

In addition to the components of the object, we can also examine attributes...

```python
%%R 

print(attributes(dc_lw_W))
```

The practical effect of using *W* style conversion is a weighting scheme that gives substantially more weight to neighbors of those areal units that have only a few units in their neighborhood relative to those in densely populated neighborhoods.  The histogram below supports this contention.  There are many more instances of low weights because these belong to weight arrays of units in with many neighbors.

```python
%%R

#Look at distribution of weights
hist(unlist(dc_lw_W$weights))
```

But, if you look at the aggregate weights by areal unit (that is, the sum of their constituent weights), we can confirm this summing to unity for each and every areal unit.

```python
%%R

#Check summed weights by neighborhood
print(summary(sapply(dc_lw_W$weights,sum)))
```

In contrast, the binary conversion style (set with *B*) does pretty much the opposite.  Every valid neighbor carries a weight of 1, so now the weights for each areal unit differ by the number of neighbors.

```python
%%R

dc_lw_B<-nb2listw(dc_nb,style='B')
print('Summary Info about Individual Weights')
print(summary(unlist(dc_lw_B$weights)))
cat('\n')
print('Summary Info about Aggregated Weights by Areal Unit')
print(summary(sapply(dc_lw_B$weights,sum)))
```

Style *C* is like *B*, except the weights sum to the number of units rather than the number of neighbors (each neighbor still has the same weight value).  *U*, in similar fashion, simply sums to unity instead.

```python
%%R

dc_lw_C<-nb2listw(dc_nb,style='C')
print('Summary Info about Individual Weights for Style *C*')
print(summary(unlist(dc_lw_C$weights)))
cat('\n')
print('Summary Info about Aggregated Weights by Areal Unit for Style *C*')
print(summary(sapply(dc_lw_C$weights,sum)))
cat('\n')

dc_lw_U<-nb2listw(dc_nb,style='U')
print('Summary Info about Individual Weights for Style *U*')
print(summary(unlist(dc_lw_U$weights)))
cat('\n')
print('Summary Info about Aggregated Weights by Areal Unit for Style *U*')
print(summary(sapply(dc_lw_U$weights,sum)))
```

A 'variance-stablizing scheme' has been proposed that has less variation in weights than *W* (but more than [*B*,*C*,*D*]), and less variation in aggregate weights by areal unit than [*B*,*C*,*D*] (but more than *W*).  It was proposed in [Tiefelsdorf et al. (1999)](http://www.envplan.com/abstract.cgi?id=a310165), and it corresponds to style *S*.  It seeks to moderate the impact of style choice on the treatment of areal unit density.  Here is how the distribution shakes out for both individual weights and areal unit totals.

```python
%%R -w 1000

dc_lw_S<-nb2listw(dc_nb,style='S')

par(mfrow=c(1,2))
hist(unlist(dc_lw_S$weights),main="Individual Weights")
hist(sapply(dc_lw_S$weights,sum),main="Aggregate Weights by Areal Unit")
```

It is also possible to set weights more directly with *glist*.  If we felt that influence declines with distance, for instance, we could construct an inverse distance weighting scheme.  Recall that the distances of each neighbor can be extracted via *nbdist* for a given **nb** object.  Then we can invert these distances (scaled as desired), and apply them with style *B* (which is really just an on/off switch).

```python
%%R -w 1000

#Grab neighbor distances
dc_dsts<-nbdists(dc_nb,coords)

#Invert said distances
idw<-lapply(dc_dsts,function(x) 1/(x/1000))

#Apply general weights in binary fashion
dc_lw_idwB<-nb2listw(dc_nb,glist=idw,style='B')

par(mfrow=c(1,2))
hist(unlist(dc_lw_idwB$weights),main="Individual Weights (Inverse Distance)")
hist(sapply(dc_lw_idwB$weights,sum),main="Aggregate Weights by Areal Unit (Inverse Distance)")
```

This obviously opens the door for substantial customization.  

There are two things to watch out for:
>1. Areal Units without Neighbors (typically handled via the [*zero.policy*](http://rss.acs.unt.edu/Rdoc/library/spdep/html/nb2listw.html) option); and,

>2. Missing Data (typically handled via [*subset.listw*](http://www.inside-r.org/packages/cran/spdep/docs/subset.listw), which regenerates weights for a given subset, which may be acquired via [*complete.cases*](http://stat.ethz.ch/R-manual/R-patched/library/stats/html/complete.cases.html))

Finally, note that we can actually transform the list of arrays in the **listw** object into a matrix form for use elsewhere...

```python
%%R

dc_idw_mat<-listw2mat(dc_lw_idwB)
print(dc_idw_mat[1:5,1:5])
```

The above figures basically correspond to the following sub-section of the 179x179 weight matrix...

```python
W=MatrixSymbol('W',5,5)
Matrix(W)
```

If we were to try and simulate spatial autocorrelation via the following matrix...

```python
from sympy.abc import rho

#Create identity matrix
I=Identity(5)

(I-rho*W)
```

..where $\rho$ is the spatial autocorrelation coefficient, we would need to use the *invIrW* function.  We could use the weight matrix of our choosing to make this happen.

```python
%%R -w 1000

set.seed(987654)

#Grab number of census tracts
n<-length(dc_nb)

#Generate 'n' random Gaussian draws
uncorr_x<-rnorm(n)

#Set correlation coefficient
rho<-.5

#Simulate autocorrelation 
autocorr_x<-invIrW(dc_lw_W,rho) %*% uncorr_x

par(mfrow=c(1,2))
plot(uncorr_x)
plot(autocorr_x)
```

The second plot demonstrates a positive relationship (but I am not going to go through the trouble of adding a LOESS line).  In any event, it should also be noted that the *lag* method provides access to ... lags in **listw** objects.

## Spatial Autocorrelation

Now that we have the basics of spatial weight allocation in place, we can start to do something with it.  One of the biggest concerns in spatial analysis is dependency across features.  Consequently, we need some way of testing for autocorrelation.  First, it's important to understand the assumptions underlying the tests we will examine.  Our focus will be Moran's I, but these assumptions are quite general:
