## Modeling Approach

The opening salvo of "Statistical Preliminaries" is largely abstract, but it is probably of use to briefly describe the general approach.  Cressie and Wikle advocate a Hierarchical Modeling (HM) approach that relies heavily on [conditional distributions](http://en.wikipedia.org/wiki/Conditional_probability_distribution) and [Bayes' Theorem](http://en.wikipedia.org/wiki/Bayes%27_theorem).  To be precise, "in order to express probabilistically the uncertainty on the parameters, the HM is called a *Bayesian Hierarchical Model* (BHM)."  If the parameters are fixed but unknown, and must be estimated from the data, the approach is called the *Empirical Hierarchical Model* (EHM).

There are three basic components to the BHM:

**Data Model:** $[Z|Y,\theta]$

**Process Model:** $[Y|\theta]$

**Parameter Model:** $[\theta]$  where $\theta = \{\theta_D, \theta_P\}$

The $\theta$ subscripts $D$ & $P$ denote data model parameters and process parameters, respectively.  The former attempt to capture elements of the analytical protocol (which might be modeling choice, measurement error, etc.), while the latter capture elements of the data generating process (DGP).  These components provide a natural method of decomposing the joint distribution of $Z$ (the data), $Y$ (the DGP we wish to explain/predict), and $\theta$ (model parameters).  

$$[Z,Y,\theta] = [Z|Y,\theta][Y|\theta][\theta]$$

This is all well and good, but we want to estimate the process and model parameters as a function of the data.  Enter Bayes'Theorem:

$$[B|A] = \frac{[A|B][B]}{\int[A|B][B]dB} = \frac{[A|B][B]}{[A]}$$

Applying this concept to the BHM is straightforward.

$$[Y,\theta|Z] = \frac{[Z|Y,\theta][Y,\theta]}{\int\int [Z|Y,\theta][Y,\theta]dYd\theta]}
= \frac{[Z|Y,\theta][Y|\theta][\theta]}{\int\int [Z|Y,\theta][Y,\theta][Y|\theta][\theta]dYd\theta]}
=\frac{[Z|Y,\theta][Y|\theta][\theta]}{[Z]}$$

The above formulation ($[Y,\theta|Z]$) is generally known as the *posterior distribution*, or the output of the interaction between the analyst's prior beliefs and the observed data.

The EHM features two main components (as a consequence of the fixed parameter set $\theta$):

**Data Model:** $[Z|Y,\theta]$

**Process Model:** $[Y|\theta]$

In this framework, the posterior also reflects the parameters being taken as given:

$$[Y|Z,\theta] = \frac{[Z|Y,\theta][Y|\theta]}{[Z|\theta]}$$

This is all that I will say here on the "Statistical Preliminaries", but the book goes on to discuss optimal predictors and the benefits of Markov assumptions to simplify computation of joint probabilities.  From here, we move on to dynamic systems.

## Temporal Processes

In general, we are working with a process of the form $\{Y(r):r \in D_t\}$ where $r$ is some time index and $D_t$ is a subset of 1-d $\mathbb{R}$ space.  This model is sufficiently general to cover *deterministic* and *stochastic* processes, as well as different conceptions of time (*continuous, discrete,* and *temporal point*).

### Deterministic Dynamical Systems

Cressie and Wikle take a [state space](http://en.wikipedia.org/wiki/State_space) approach to modeling deterministic systems.  

"[T]he *state* of a system at time $t$ is described by a set of $p_\alpha$ variables, $\alpha_t=(\alpha_t(1),...,\alpha_t(p_\alpha))'$, which are sometimes known as the *state variables* or, collectively, as the *state process*. [...] [A]ny system whose state changes with time is a *dynamical system*. The evolution of the system through *state space* (or, equivalently, *phase space*) is often called a *trajectory*.  If the current state of the system, say $\alpha_t$, uniquely determines the future states $\alpha_{t+\tau} ; \tau \gt 0$, the system is referred to as a *deterministic dynamical system*.

The focus here is on discrete systems of the form $\alpha_t = M(\alpha_{t-1};\theta)$, where $M$ is a mapping function describing the temporal dynamic of the state process.  An example of such a model would be the *first-order affine dynamical system*, which is just a ton of words that refer to a basic linear model:

$$\alpha_t = \theta_0 + \theta_1 \alpha_{t-1}$$

#### Equilibrium Value/Fixed Point of a State System

We can use the linear model above to illustrate the concept of an *equilibrium value* of $\alpha_t$.  Technically an equilibrium value "is a number '$a$' such that $\alpha_t = a$ for all values of $t$ when $\alpha_0=a$."  In other words, at certain starting values, the system will hold steady over time.  

To cement this concept, Cressie and Wikle use the example of an animal population that starts at size $\alpha_0$.  This population grows at a rate of 5% annually, but also loses a fixed number of 30 members each year.  The system is thus characterized by the following model:

$$\alpha_t = 1.05\alpha_{t-1} - 30$$

The equilibrium value requires all values of $\alpha$ to be equivalent for all $t$, so the equation can be solved analytically.

$$\alpha(1.05-1)=30 \Rightarrow \alpha = 600$$

We can observe this dynamics of this system, and test the analytical result, with a straightforward function.

```python
#Define autoregressive process
def ar(alpha_0,rate,loss):
    alpha_new=(1+rate)*alpha_0 - loss
    return alpha_new

#Define starting conditions for equilibrium value
a=600
r=.05
l=30

#Define perturbations of initial value
pert=.01
a_low=(1-pert)*a
a_high=(1+pert)*a

#Initialize systems at different initial values of alpha
a_new=ar(a,r,l)
a_new_low=ar(a_low,r,l)
a_new_high=ar(a_high,r,l)

#Generate vectors for equilibrium, low, and high scenarios
##Create containers for data
a_list=[a,a_new]
a_low_list=[a_low,a_new_low]
a_high_list=[a_high,a_new_high]
##For 100 iterations...
for i in range(100):
    #...dynamically generate data with the above defined AR process and ...
    a_new=ar(a_new,r,l)
    a_new_low=ar(a_new_low,r,l)
    a_new_high=ar(a_new_high,r,l)
    #...capture data in containers for display
    a_list.append(a_new)
    a_low_list.append(a_new_low)
    a_high_list.append(a_new_high)
    
#Formulate into DF
alpha_df=DataFrame({'equil':a_list,
                    'low_start':a_low_list,
                    'high_start':a_high_list})

#Plot trajectories
plt.rcParams['figure.figsize']=15,5
alpha_df.plot(kind='line',title='Trajectory of AR(1) Process w/ Different Starting Values - Sensitivity = '+str(pert))
```

#### System Stability

While the rate of departure varies with the perturbation, it is clear that the equilibrium value is *unstable* because even when the perturbation is arbitrarily small, the perturbed system trends away from the equilibrium value over time.  A *stable* system, by contrast, would (within a given distance of the equilibrium value) trend towards the equilibrium value in the limit.

The relationship of a system to its equilibrium value can display a range of behavior.  Based upon the starting values, a system may approach or diverge exponentially, oscillate in a dampened manner towards a, or oscillate with increasing amplitude.  There are also systems that can oscillate in a stable fashion, neither diverging from nor converging on the equilibrium value.  An example of a dampening system is easily constructed from the following model:

   $$\alpha_t = -0.7\alpha_{t-1}$$

   where $\alpha_0=12$

```python
#Define starting conditions for equilibrium value
a = 12
r = -1.7
l = 0

#Initialize system
a_new=ar(a,r,l)

#Generate system trajectory vector
##Create container for data
a_list=[a,a_new]
##For 40 iterations...
for i in range(20):
    #...dynamically generate data with the above defined AR process and ...
    a_new=ar(a_new,r,l)
    #...capture data in container for display
    a_list.append(a_new)
    
#Plot resulting figure
Series(a_list).plot(kind='line',title='Example of a Process Characterized by Dampening Oscillation')
```

#### Higher-Order Systems

Staying within the univariate framework, it useful here to consider higher order systems.  In this context, this means only that we are estimating parameters for more than single lags.  From a practical standpoint, a lag operator ($L$) becomes useful.  (Note:  Cressie and Wikle refer to a backward shift operator ($B$), but the lag operator fits more with my experience and does nothing to damage the concept.)  Consequently, starting from the following model...

$$\alpha_t=\theta_1 \alpha_{t-1} + \theta_2 \alpha_{t-2}$$

...the same process can be expressed as the following:

$$(1 - \theta_1 L - \theta_2 L^2) \alpha_t = 0$$

The first component of the left side $(1 - \theta_1 L - \theta_2 L^2)$ is known as the *characteristic polynomial* of the system.  The roots of this expression describe the behavior of the system.

### Nonlinear Discrete Dynamical Systems

While there is a great deal of conceptual overlap with linear systems, there are some peculiarities in this area that beg exploration.  Cressie and Wikle focus on the logistic dynamical system, one example of which would be the following density-dependent population growth model:

$$\alpha_t = (1+\theta) \alpha_{t-1} - \frac{\theta_1}{\theta_0} \alpha_{t-1}^2, t=1,2,...$$

This is an example of the more general model $\alpha_t = M(\alpha_{t-1})$.  In general $a$ is attracting if $|M'(a)|>1$ and repelling if $|M'(a)|<1$ (the superscript denotes the first derivative with respect to $\alpha_t$).  In the case of the population growth model, the first derivative is $|M'(a)| = 1 + \theta_1 - 2\frac{\theta_1}{\theta_0}a$.  We can model the behavior of this system using a value of $a=4$.

```python
def pop_growth(a0,t1,t0):
    alpha_new=(1+t1)*a0 - (t1/t0)*a0**2.0
    return alpha_new

#Set starting alpha
alpha0=4.0

#Set model paramters for two scenarios
theta1_x=.25
theta1_y=-.25
theta0=10.0

#Initialize systems
alpha_new_x=pop_growth(alpha0,theta1_x,theta0)
alpha_new_y=pop_growth(alpha0,theta1_y,theta0)

#Create containers for data
alpha_x_list=[alpha0,alpha_new_x]
alpha_y_list=[alpha0,alpha_new_y]

#Generate data series and store in lists
for i in range(50):
    #Generate next iteration of dynamical process
    alpha_new_x=pop_growth(alpha_new_x,theta1_x,theta0)
    alpha_new_y=pop_growth(alpha_new_y,theta1_y,theta0)
    #Store in containers
    alpha_x_list.append(alpha_new_x)
    alpha_y_list.append(alpha_new_y)
    
#Plot series
fig,axes=plt.subplots(2)
Series(alpha_x_list).plot(kind='line',title='Growth Parameter 0 < $theta1$ < 2 Stabilizes at Carrying Capacity',ax=axes[0])
Series(alpha_y_list).plot(kind='line',title='Growth Parameter 0 > $theta1$ > -2 Suggests Extinction',ax=axes[1])
```

If $|M'(a)|=1$ and $|M''(a)|<0$ then the process is semistable from above.  Conversely, if $|M'(a)|=1$ and $|M''(a)|>0$ then the process is semistable from below.  

### Bifurcation

When the 'qualitative' behavior of a process shifts dramatically in response to small perturbations in the parameter space, the values at which these changes occur are known as bifurcation values.  This idea can be demonstrated with a one-parameter logistic equation:

$$\alpha_t = \theta \alpha_{t-1} (1 - \alpha_{t-1})$$

The first equilibrium value is trivial, $a=0$.  The second can be solved for by substituting $a$ for all instances of $\alpha$.

$$a = \theta a - \theta a^2$$

$$\theta a = \theta - 1$$

$$a = \frac{\theta - 1}{\theta}$$

We can locate the bifurcation point graphically by plotting the function twice (with both equilibrium points $a$) over the parameter space.  We will, somewhat arbitrarily, plot over the [0,3] interval in the $\theta$ space.

```python
#Initialize theta
theta=.001

#Initialize data index (the theta space container)
theta_list=[theta]

#Initialize a values
a_x=0
a_y=(theta-1)/theta

#Create containers
a_x_list=[a_x]
a_y_list=[a_y]

#Generate data series (definitions of a embedded) and toss them into lists
for i in range(0,30,1):
    theta+=(.1)
    theta_list.append(theta)
    a_x_list.append(a_x)
    a_y_list.append((theta-1)/theta)
    
#Plot both values of a over the parameter space
plt.figure()
Series(a_x_list,index=theta_list).plot(kind='line',label='a=0',linestyle='--',linewidth=5)
Series(a_y_list,index=theta_list).plot(kind='line',label='a=(theta-1)/theta',linewidth=5)
plt.xlim([.1,3.0])
plt.ylim([-1.5,1])
plt.legend(loc='best')
plt.title('A Transcritical Bifurcation Exists at theta=1')
```
