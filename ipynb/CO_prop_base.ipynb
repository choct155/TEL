```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
from IPython.display import HTML

#Enable R capability
%load_ext rmagic

#Set print length
pd.set_option('line_width',140)

#Set chart aesthetic
pd.options.display.mpl_style='default'

#Establish data directory
data_dir='/home/choct155/Google Drive/Dissertation/Data/'

#Read in data
# med_homeval=pd.read_csv(data_dir+'CO_med_hv_cty_1979_.csv',skiprows=[0,1,2,4,5,6,7,8,9,10,11,12],\
#                         delimiter='\t',parse_dates=True)
m=pd.read_csv(data_dir+'CO_med_hv_cty_1979_.csv',skiprows=[0,1,2,4,5,6,7,8,9,10,11,12])

m[m.columns[:6]].head(20)
```

Well that was unexpected... there is only one column of data.  Evidently Moody's lied to me, and the data are not comma delimited.  In fact, they are in a rather awkward arrangement (which can be seen when viewing the data as a text file).  Each row is tab-delimited, but all of the values in each row are grouped and displayed as a wrapped block of text, before moving on to the next row.

This is certainly awkward, but by no means insurmountable.  We need to parse out the different columns by splitting on the tabs and creating an ndarray, which flows naturally into a DataFrame.  A bit of trial and error was used to determine the appropriate index calls to get at the data hidden in this weird format.

```python
#Initiate container with headers
m_array=[m.columns[0].split('\t')]

#For each row in m...
for i in range(len(m)):
    #...split the value on tabs, and append the resulting list to the above container
    m_array.append(m.ix[i].str.split('\t')[0])

#Inspect date covereage
dates=[]
#For each row...
for row in m_array:
    #...throw the first element in the dates list
    dates.append(row[0])

print dates[:5]
print dates[-5:]
```

So, it looks like we have full coverage and can transition into a DataFrame.  Note that there are many empty columns on the back end, but they are easy enough to dispose of from the DataFrame object.

```python
#Convert to DF (first element of m_array contains header info)
mhv_temp=DataFrame(m_array[1:])

#Clean up column names (drop County (CO))
##Define function to replace multiple elements
def replace_all(s,dct):
    for i,j in dct.iteritems():
        s=s.replace(i,j)
    return s
##Define substrings to be removed
replace_dict={'(CO)':'',
              'County':''}
##Create container for new county names
new_cty_list=[]
##For each county in the header array...
for cty in m_array[0]:
    #...replace the relevant substrings
    new_cty_list.append(replace_all(cty,replace_dict).strip())
##Set new county names as column names
mhv_temp.columns=new_cty_list

#Drop extraneous columns
mhv=mhv_temp[mhv_temp.columns[:65]]

#Extract year
mhv['year']=mhv.Geography.str[:4]

#Drop Geography
mhv.pop('Geography')

mhv[mhv.columns[:14]].head()
```

We need to average these data for each year and reshape these data into a panel.

```python
#Average over years
mhv_a=mhv.astype(float).groupby('year').mean().applymap(lambda x: x*1000).stack()

mhv_a.to_csv(data_dir+'CO_cty_mhv.csv')

mhv_a.head()
```
