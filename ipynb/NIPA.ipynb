We will also need recession bars to place trajectories in context.  For this purpose we have pulled a time series of binary, NBER-based, [recession indicators](http://research.stlouisfed.org/fred2/series/USREC#) from the [FRED](http://research.stlouisfed.org/fred2/) repository hosted by the St. Louis Fed branch.

```python
#Read in data
rec=pd.read_csv(data_dir+'nber_recessions_fred.csv',parse_dates='observation_date')

#Convert date to datetime
rec['date']=rec['observation_date'].apply(lambda x: pd.to_datetime(x))

#Extract year and month
rec['year']=rec['date'].apply(lambda x: x.year)
rec['month']=rec['date'].apply(lambda x: x.month)

#Calculate month fraction and add to year for starts and stops
rec['stop']=(rec['month']/12.)+rec['year']
rec['start']=(rec['month']/12.-(1/12.))+rec['year']

rec.head()
```

## Current Government Expenditures as a Percentage of GDP

```python
#Define colors
exp_color='#1F0F87'
rev_color='#BD1722'

#Generate plot object 
fig,ax=plt.subplots(figsize=(16,5))

#Add recession bars
ax.fill_between(rec['stop'],0,rec['USREC']*100.,lw=0,color='#4E4A52',alpha=.3)

#Plot total government revenues and expenditures as a % of GDP
((nipa['tot_gov_rec']/nipa['gdp'])*100.).plot(kind='line',ax=ax,color=rev_color,lw=3,label='Receipts',
                                       title='Total Government Revenue and Expenditures as a % of GDP')
((nipa['tot_gov_exp']/nipa['gdp'])*100.).plot(kind='line',ax=ax,color=exp_color,lw=3,label='Expenditures',
                                       title='Total Government Revenue and Expenditures as a % of GDP')

#Remove grid
ax.grid('off')

#Fix axes
def simpleaxis(ax):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()
simpleaxis(ax)
ax.set_xlim([1929,2014])
ax.set_ylim([0,40])

#Fix labels
ax.set_xlabel('Year')
ax.set_ylabel('Volume as a % of GDP')
ax.legend(loc='best')

# plt.savefig('/home/choct155/dissertation/TEL/out/figures/tot_gov_GDP_perc.png',dpi=500)
plt.savefig(out_dir+'tot_gov_GDP_perc.png',dpi=500)
```

## Property Tax

```python
#Read in data
lre=pd.read_csv(data_dir+'nipa_loc_rev_exp.csv')

#Strip spaces
lre['Concept']=lre['Concept'].apply(lambda x: x.strip())

#Set index
lre.set_index('Concept',inplace=True)

#Transpose
lre2=lre.T

#Capture subset
lre_sub=lre2[['Property taxes','Current transfer receipts','Current receipts']]

#Define property receipts as proportion of total current receipts
lre_sub['prop']=lre_sub['Property taxes']/lre_sub['Current receipts']

#Define transfer receipts as proportion of total current receipts
lre_sub['transfer']=lre_sub['Current transfer receipts']/lre_sub['Current receipts']

#Capture percentage change
lre_sub['prop_change']=lre_sub['Property taxes'].pct_change()

lre_sub.head(10)
```

```python
#Generate plot object 
fig,ax=plt.subplots(figsize=(16,5))

#Add recession bars
ax.fill_between(rec['stop'],0,rec['USREC'],lw=0,color='#4E4A52',alpha=.3)

#Plot property taxes as % of all local current receipts
lre_sub['prop'].plot(kind='line',ax=ax,color='#857846',lw=5,label='Property Tax')
lre_sub['transfer'].plot(kind='line',ax=ax,color='#465385',lw=5,label='Transfers')

#Fix background
ax.patch.set_facecolor('w')
ax.grid('off')

#Fix labels
ax.set_xlabel('Year')
ax.set_ylabel('Proportion of Total Current Receipts')
ax.legend(loc='best')

#Fix axis
simpleaxis(ax)

plt.savefig(out_dir+'nipa_prop_transfer.svg')
```
