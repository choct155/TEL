# Exploring Data with Simulation

Among other things, simulation has demonstrated the capacity to provide insight into small sample sizes.  To this end, this will work through some examples to provide basic concepts.

The initial problem is determining the fairness of a flipped coin after only 8 flips (6H:2T).  To evaluate this, we are going to simulate a sequence of tosses at multiple levels of *p* (where *p*=.5 is a balanced coin).

```python
import random
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

plt.rcParams['figure.figsize']=10,10

#Define trial structure
repeats,tosses=60,8

#Define function representing a toss of heads
def heads(tosses,p):
    h=0
    #For each toss, heads occurs if the random number is below p
    for x in range(0,tosses):
        if random.random() < p: 
            h+=1
    return h

#Create containers to capture results for plotting
prob=[]
h_t=[]

#Initialize probability value
p=0

#Roll through each probability value at .05 intervals and grab number of heads
while p<1.01:
    for t in range(0,repeats):
        #print p,'\t',heads(tosses,p)
        prob.append(p)
        h_t.append(heads(tosses,p))
    p+=.05

#Set jitter parameters to deal with overplotting
jitter=.01
jitter2=.2
prob_jitter=[p+random.uniform(-jitter,jitter) for p in prob]
ht_jitter=[ht+random.uniform(-jitter2,jitter2) for ht in h_t]

#Plot results
plt.scatter(prob_jitter,ht_jitter,alpha=.2,facecolors='None')
plt.xlim([0-jitter,1+jitter])
plt.ylim([0-jitter2,8+jitter2])
```

What about this bootstrapping business?  One of the issues we have in the District (like most places) is the sparse nature of PUMS data.  Perhaps we could resample to get more out of this data?  Let's take a look at an example.  We are going to test resampling against a known distribution.  The data (n=25) are taken from a standard Gaussian distribution with mean zero and standard deviation of one.  We know at the outset that the standard error (standard deviation divided by the square of the number of observations) is 1/5.

```python
plt.rcParams['figure.figsize']=5,5
mean = 0
variance = 1
sigma = sqrt(variance)
x = np.linspace(-3,3,100)
plt.plot(x,plt.mlab.normpdf(x,mean,sigma))
```

Now to draw random data from a Gaussian distribution...

```python
gaus=np.random.normal(0,1,25)
print(gaus)
```

From this sample we need to pull 100 samples with replacement of 25 observations each.  We'll calculate and capture the mean of all 100 samples in list for plotting.

```python
gaus_mean=[]
for i in range(100):
    gaus_sample=[]
    for j in range(25):
        index=np.random.randint(0,25)
        gaus_sample.append(gaus[index])
    gaus_mean.append(np.mean(gaus_sample))
print gaus_mean
```

Let's try to throw all of this together on a plot to see what we have.

```python
#Enable kernel density estimation
from scipy.stats import gaussian_kde

#Create vector of length gaus_mean
y=[0.1]*len(gaus_mean)
jit=.04
y_jitter=[yi+random.uniform(-jit,jit) for yi in y]

#Create height vector of length gaus
y2=[0]*len(gaus)
jit2=.01
y2_jitter=[yi+random.uniform(-jit2,jit2) for yi in y2]

#Generate kernel densities for both the sample and bootstrap data
gdens=gaussian_kde(gaus)
gmdens=gaussian_kde(gaus_mean)

#Generate horizontal info
x_g=np.linspace(-5,5,200)

fig,axes=plt.subplots()
plt.rcParams['figure.figsize']=10,10
axes.plot(x,plt.mlab.normpdf(x,mean,sigma),label='Normal Distribution')
axes.scatter(gaus_mean,y_jitter,alpha=.2,facecolors='none',label='Bootstrap Means')
axes.scatter(gaus,y2_jitter,facecolor='g',label='Random Data')
axes.plot(x_g,gdens(x_g),label='KDE Data')
axes.plot(x_g,gmdens(x_g),label='KDE Bootstrap')
```

```python
from scipy.stats.kde import gaussian_kde
from scipy.stats import norm
from numpy import linspace,hstack
from pylab import plot,show,hist

# creating data with two peaks
sampD1 = norm.rvs(loc=-1.0,scale=1,size=300)
sampD2 = norm.rvs(loc=2.0,scale=0.5,size=300)
samp = hstack([sampD1,sampD2])

# obtaining the pdf (my_pdf is a function!)
my_pdf = gaussian_kde(samp)

# plotting the result
x = linspace(-5,5,100)
plot(x,my_pdf(x),'r') # distribution function
hist(samp,normed=1,alpha=.3) # histogram
```
