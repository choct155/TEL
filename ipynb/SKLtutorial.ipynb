### k-Nearest Neighbor Classification (kNN)

This is perhaps the simplest classification method.  The tutorial defines the method as follows:

>[G]iven a new observation `X_test`, find in the training set (i.e. the data used to train the estimator) the observation with the closest feature vector.

Coming from a bit more of a spatial bent, this seems to me an awkward way of stating the process.  It seems clearer to me to say that new observations should belong to the dominant class in their local training neighborhood.  Then again, maybe that's a confusing explanation for other folks.

To implement kNN, or any other learning algorithm, the data must first be split into two groups: train and test.  With the **iris** data, we can split the data in half to serve these purposes.  One might ask, what if the order of the rows in the data is meaningful?  This issue can be mitigated by running the data through a random permutation.

```python
#Set seed 
np.random.seed(0)

#Permute the data index
indices=np.random.permutation(len(iris_X))

#Capture training set
iris_X_train=iris_X[indices[:-10]]
iris_y_train=iris_y[indices[:-10]]

#Capture test set
iris_X_test=iris_X[indices[-10:]]
iris_y_test=iris_y[indices[-10:]]
```

kNN is implemented here with the **KNeighborsClassifier** object.

```python
knn=neighbors.KNeighborsClassifier(algorithm='auto', leaf_size=30, metric='minkowski',\
                                   n_neighbors=5, p=2, weights='uniform')
type(knn)
```

This is our first opportunity to see a particular implementation of the general **estimator** structure from above.  In the previous cell we set the classifier parameters, and now we will fit the data.

```python
#Fit the data
knn.fit(iris_X_train,iris_y_train)
```

Now we can compare the values predicted by the classifer with the values of the actual data from the test set.

```python
#Print actual data
print iris_y_test

#Print predicted values
print knn.predict(iris_X_test)

#Print difference
print iris_y_test - knn.predict(iris_X_test)
```

Only one miss seems like a pretty good hit rate given the small size of the training set.  Still, it's a bit abstract.  We can go a step further to visualize the pattern we are attempting to capture.  It's actually possible to plot classification regions with the data overlaid with the aid of **np.meshgrid**.  This method basically increases the resolution of $n$ vectors (by setting the step size) in pursuit of building an $n$-dimensional coordinate array.  The initial output is comprised of $n$ ndarrays that each capture one position in individual coordinates.  These ndarrays can be flattened (with **np.ravel**) and zipped to capture the full list of coordinates as tuples in a 1-D vector.  An example of this operation can be seen below.

```python
#Establish step size
step=.5

#Create coordinate matrices
new_a,new_b=np.meshgrid(np.arange(0,2,step),\
                        np.arange(0,1,step))
print '**First Dimension**'
print new_a
print '\n**Second Dimension**'
print new_b
print '\n**List of Coordinates**'
print zip(new_a.ravel(), new_b.ravel())
```

The coordinate array from our new vectors allows for finer control over any boundaries we might want, so let's get back to the task at hand.

Note that we will be plotting in 2-D space, so we will again use only sepal length and width to define our axes.  This means that while in the context of our data *y* indicates the target values (flower classes), in the context of the region, *y_min* and *y_max* refer to the dimensions of the second data dimension (sepal width).

```python
#Create color maps
cmap_light=matplotlib.colors.ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = matplotlib.colors.ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

#Define the region to be colored
x_min, x_max = iris_X_sub[:, 0].min() - 1, iris_X_sub[:, 0].max() + 1
y_min, y_max = iris_X_sub[:, 1].min() - 1, iris_X_sub[:, 1].max() + 1

#Set step size in mesh
h=.02

#Create coordinate matrices
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),\
                     np.arange(y_min, y_max, h))
```

Now that we have fine resolution of individual positions in the sample space, we will fit the subset of data (again, sepal length and width), and use this data as a training set to predict the class of each of the individual sample space positions we have just defined.  We are defining the local neighborhood as having a size of 15 neighbors.

To plot this business, we want to use the shape of the coordinate system, which we could capture from either *xx* or *yy*. Each coordinate will be colored using the **plt.pcolormesh** method, which appears to be analogue to the psuedocolor function **plt.pcolor**.  The difference is the former takes dimension-specific ndarrays and a value overlay, whereas the latter just uses the values in a single ndarray.

```python
#Reshape the predicted position classes
Z=Z.reshape(xx.shape)

#Generate figure object
plt.figure(figsize=(12,8))

#Map colors to positions
plt.pcolormesh(xx,yy,Z,cmap=cmap_light)

#Plot training points
plt.scatter(iris_X_sub[:,0],iris_X_sub[:,1],c=iris_y,cmap=cmap_bold)

#Set limits
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())

#Set title
plt.title('3-Class classification (k = 15, weights = Uniform')
```

### [The Curse of Dimensionality](http://en.wikipedia.org/wiki/Curse_of_dimensionality)

I don't think the explanation offered in the tutorial is particularly clear, so we will deviate rely a bit more heavily on the general information provided by [Wikipedia](http://en.wikipedia.org/wiki/Main_Page).  Basically, the curse of dimensionality occurs when an increase in the number of dimensions causes an increase in the parameter space at such a rapid rate that the available data becomes sparse.  This sparsity can be thought of as decreasing the frequency with which observations vary.  Variation in this sense is not referring to the magnitude of the variation, but rather whether or not variation occurs at all.

To use a very simple (and unrealistic) example, suppose one wanted to test whether flipping a light switch would cause a light to turn on.  We already know that the switch is not a sure thing, and merely increases the probability of the light illuminating.  The greater the number of tests that result in non-events (the light not turning on or off), the more data is required to establish that the switch has a non-zero probability of turning on the light (if that is indeed the case in the underlying data generating process).  In a nutshell, we need variation in the data and the target to identify the relationship between them.  The more variation we have, the greater the precision with which we can identify the relationship.  The sparsity driven by high dimensionality discussed above is an enemy of this identification.

Within the context of machine learning, the implication of high-dimensionality is an increase in the amount of training samples that are fit to the target.  If the training set is fixed, predictive power reduces with each additional dimension.  This is, apparently, known as the Hughes Effect.

#### [Ridge Regression](http://en.wikipedia.org/wiki/Tikhonov_regularization)

Least squares is a very widely used estimator because it is the Best Linear Unbiased Estimator even when the errors are not Gaussian (by virtue of the Gauss-Markov Theorem).  However, by virtue of specifying it as the best among unbiased estimators, one is lead to question whether the variance around estimates can be reduced if the unbiased property can be relaxed.  This would be useful when one does not have enough data in each dimension to drive down the standard error around an estimate.  In fact, there are techniques that are able to capture more precise estimates, one of which is ridge regression.  (An excellent, brief discussion of the similarity between ridge regression, principal components analysis, and bayesian estimation can be found on [Open Mind](http://tamino.wordpress.com/2011/02/12/ridge-regression/).)  

In a nutshell, ridge regression is more sensitive to drivers of target variation.  It achieves this, however, at the expense of being able to provide an effect estimate that reflects the true impact of whatever is being estimated.  Said differently, it increases [precision at the expense of accuracy](http://en.wikipedia.org/wiki/Accuracy_and_precision).

The following example can shed some light on what this means.  First we define a train target set that maps 1-to-1 to the train data, and a test data set that extends beyond the boundaries of the train set.

```python
#Define train set
ridge_X=np.c_[.5,1].T
ridge_y=[.5,1]

#Define test set
test=np.c_[0,2].T

print 'Train Data\n',ridge_X
print '\nTrain Target\n',ridge_y
print '\nTest Data\n',test
```

Now we will generate six variations on `ridge_X`, fit regression objects with the variations and `ridge_y`, and plot `test` against the predictions yielded by the same regression objects.

```python
#Generate plot object
plt.figure(figsize=(12,8))

#Set random seed
np.random.seed(0)

#Instantiate regression object
regr=linear_model.LinearRegression()

#Create container for coefficients
reg_coef_list=[]

#For six iterations...
for _ in range(6):
    #...generate a variant of ridge_X...
    this_X=.1*np.random.normal(size=(2,1)) + ridge_X
    #...fit the variant and ridge_y...
    regr.fit(this_X,ridge_y)
    #...plot test against the regression objects predictions...
    plt.plot(test,regr.predict(test))
    #...overlay scatters of ridge_y vs the variant of ridge_X...
    plt.scatter(this_X,ridge_y,s=3)
    #...and capture the coefficient
    reg_coef_list.append(regr.coef_)
```

Let's unpack this a bit.  Basically, we have explored the following regression model.

$$y = \beta X + \epsilon$$  

realized as 

`ridge_y = reg.coef_ * this_X` plus assumed $iid$ errors.

The slope of each of the lines in the plot above is the coefficient of X.  Note the wide variation in coefficients from (and this is key) small random normal deviations on the train data.  These are peturbations in the train data that would be legitimately captured by the error term.

If we were to perform the analagous procedure using ridge regression...

```python
#Generate plot object
plt.figure(figsize=(12,8))

#Set random seed
np.random.seed(0)

#Instantiate regression object
ridge_reg=linear_model.Ridge(alpha=.1)

#Create container for coefficients
ridge_coef_list=[]

#For six iterations...
for _ in range(6):
    #...generate a variant of ridge_X...
    this_X=.1*np.random.normal(size=(2,1)) + ridge_X
    #...fit the variant and ridge_y...
    ridge_reg.fit(this_X,ridge_y)
    #...plot test against the regression objects predictions...
    plt.plot(test,ridge_reg.predict(test))
    #...overlay scatters of ridge_y vs the variant of ridge_X...
    plt.scatter(this_X,ridge_y,s=3)
    #...and capture the coefficient
    ridge_coef_list.append(ridge_reg.coef_)
```

...we can see that our estimates do not vary as much.  How different are the two sets of coefficients?

```python
#Generate DF from coefficients
ridge_v_reg=DataFrame({'ridge':[x[0] for x in ridge_coef_list],
                       'reg':[x[0] for x in reg_coef_list]})

#Set plot size
plt.rcParams['figure.figsize']=12,8

#Generate boxplot
ridge_v_reg.boxplot()
plt.title('Ridge vs. OLS Coefficients');
```

The boxplot above captures the essence of the implications of the ridge technique.  We know that the train target and train data have a 1-to-1 direct (that is, with no proportional scaling) mapping.  Therefore, the population parameter is 1.  OLS, our BLUE estimator, shows a median estimate that is very close to this value.  There is, however, relatively substantial variation around the median estimate.

The ridge regression shows dramatically less variation around its median estimate, *but the estimate is quite biased*.  The ridge median estimate is quite close to half the magnitude of the true population parameter.  The `alpha` parameter in the **Ridge** object determines how much we would like to suppress variance in the estimate.

#### Classification

For classification, linear regression is not an appropriate tool because it would give too much weight to extreme values (data that is 'far from the decision frontier').  The conventional approach for categorical responses in a machine learning context is the same one used in classical econometrics:  logistic regression.

$$y = \text{sigmoid}(X \beta - \text{offset}) + \epsilon = \frac{1}{1 + exp(-X \beta + \text{offset})} + \epsilon$$

[Support Vector Machines](http://en.wikipedia.org/wiki/Support_vector_machine) (SVMs) are an alternative method of deterministic classification.  In a nutshell, they look for samples of the data that enable the building of a plane that maximizes the margin between two classes.

Finally, [kernel classification](http://en.wikipedia.org/wiki/Kernel_trick) is a variant of SVM that provides flexibility in capturing classes that are not linearly separable in the feature space.

## Model Selection:  Choosing Estimators and Their Parameters

The goal in machine learning is to maximize *predictive* accuracy.  When performing an analysis, the goal is to avoid overfitting the data and to acquire the best estimate of the underlying population parameter possible.  If do not split the data into training and test cases, the threat of overfitting looms large.

Let us look again at the SKL's digit data.  Here we will find the score after one train/test split.

```python
#Extract data and target from the digits Bunch object
X_digits=digits.data
y_digits=digits.target

#Create a support vector machine object
svc=svm.SVC(C=1,kernel='linear')

#Fit the data and extract the score
svc.fit(X_digits[:-100],y_digits[:-100]).score(X_digits[-100:],y_digits[-100:])
```

To increase the predictive accuracy, we can simply increase the number times we train and test.  Splitting the data into groups for this process is called **KFold** cross-validation.  When we get multiple scores, we have a better sense of the accuracy of our model.

We can fold the data and target above into three groups.

```python
#Fold the data into three groups
X_folds=np.array_split(X_digits,3)
y_folds=np.array_split(y_digits,3)

X_folds
```

Let's explore the distribution of scores we get when testing across these three groups.  With each scoring iteration, we are pulling $\frac{2}{3}$ as the train set, and testing against the remaining $\frac{1}{3}$.  When scores are consistently high across groups, it gives us confidence in the predictive accuracy of our model.

```python
#Create container for scores
score_list=[]

#For three iterations
for i in range(3):
    #...create a list copy of data and target...
    X_train=list(X_folds)
    y_train=list(y_folds)
    #...pull out test set arrays from a third of the folded data and target...
    X_test=X_train.pop(i)
    y_test=y_train.pop(i)
    #...combine the remaining two thirds into the training set...
    X_train=np.concatenate(X_train)
    y_train=np.concatenate(y_train)
    #...fit and score the data for the given split...
    score=svc.fit(X_train,y_train).score(X_test,y_test)
    #...and capture the score
    score_list.append(score)
    
#Plot scores
plt.rcParams['figure.figsize']=8,6
Series(score_list).plot(kind='bar',title='KFold Cross-Validation Scores - k=3')
print '**Scores**\n',score_list
```

This is a bit of a pain to write each time, so SKL has implemented some convenience functions.  In this case, once the **KFold** object has been generated, the relevant information can be extracted succinctly with a list comprehension.

```python
#Generate KFold object
k_fold=cross_validation.KFold(len(X_digits),n_folds=3)

#Capture scores from same model
new_score_list=[svc.fit(X_digits[train],y_digits[train]).score(X_digits[test],y_digits[test]) for train,test in k_fold]

#Plot scores
Series(new_score_list).plot(kind='bar',title='KFold Cross-Validation Scores - k=3')
print '**Scores**\n',score_list
```

In fact, if we are concerned only with the scores themselves, there is a method dedicated to this task.

```python
#Capture scores
even_newer_scores=list(cross_validation.cross_val_score(svc, X_digits, y_digits, cv=k_fold, n_jobs=-1))

#Plot scores
Series(even_newer_scores).plot(kind='bar',title='KFold Cross-Validation Scores - k=3')
print '**Scores**\n',even_newer_scores
```

In addition to `KFold(n,k)`, there are other cross-validation generators as well:

+ `StratifiedKFold(y,k)`:  preserves class ratios/label distribution within each fold
+ `LeaveOneOut(n)`:  pops just one observation out instead of evenly split folds
+ `LeaveOneLabelOut(labels)`:  pops out a group based upon a categorical identifier

### Grid-Search & Cross-Validated Estimators

This all well and good, but this all occurs after we have developed our parameter estimates.  What if we want to employ these techniques to help us identify the appropriate estimate?

#### Grid-Search

It turns out SKL provides a class designed to pick the parameter estimates that maximize the cross-validation score.  Going back to the SVM example, one of the parameters of interest is the kernel coefficient.  We can feed a **GridSearchCV** object with a list of possible values for $\gamma$ (the kernel coefficient), and extract which is optimal (based upon the resultant cross-validation score).

```python
#Define parameter input dictionary
gamma_dict=dict(gamma=np.logspace(-6,2,10))

#Instantiate GridSearchCV object
gscv=grid_search.GridSearchCV(estimator=svc,param_grid=gamma_dict,n_jobs=-1)

#Fit data
gscv.fit(X_digits[:1000],y_digits[:1000])
```

From here, we can ask all kinds of questions.  

+ What was the best score?
+ What was the best value of $\gamma$?
+ What did the distribution of scores look like?

```python
#Extract best score
print '***Best Score***\n',gscv.best_score_

#Extract best gamma value
print '\n***Best Gamma Value***\n',gscv.best_estimator_.gamma

#Create containers for gamma, mean, and std
gamma_list=[]
mean_list=[]
std_list=[]

#For each value of gamma...
for i in range(len(gscv.grid_scores_)):
    #...capture gamma, mean, and std
    gamma_list.append(gscv.grid_scores_[i].parameters['gamma'])
    mean_list.append(gscv.grid_scores_[i].cv_validation_scores.mean())
    std_list.append(gscv.grid_scores_[i].cv_validation_scores.std())

#Generate DF from extracted data
gamma_performance=DataFrame({'gamma':gamma_list,
                             'mean':mean_list,
                             'std':std_list})
print '\n***Gamma Performance***'
gamma_performance
```

To be perfectly honest, I don't know why the scores don't vary with $\gamma$.  I will have to get into the weeds a bit more with SVM to understand if this is really an issue.

In any case, we can also run two cross-validation loops in parallel:  one for gamma selection and one measuring predictive performance.  The result are unbiased estimates of scores on new data.
